```python
import traceback
import os
import time
import requests
import numpy as np
import pandas as pd
import pandas_ta
import yfinance as yf
from datetime import datetime, timedelta
import pywt
import antropy as ant
import lightgbm as lgb
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit # Removed train_test_split as TimeSeriesSplit is more appropriate here
from sklearn.metrics import accuracy_score, classification_report, log_loss, roc_auc_score # ADDED roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest, mutual_info_classif, f_classif, VarianceThreshold # ADDED VarianceThreshold
from transformers import PatchTSTConfig, PatchTSTForPrediction, AutoProcessor


if hasattr(pd.DataFrame, 'ta') is False and pandas_ta is not None:
    try:
        pd.DataFrame.ta = pandas_ta.Core(df=None) # More direct registration
        print("pandas_ta DataFrame accessor registered globally.")
    except Exception as e:
        print(f"Could not globally register pandas_ta accessor: {e}")

optuna_available = False
try:
    import optuna
    optuna_available = True
    print("Optuna imported successfully.")
except ImportError:
    print("Optuna not found. LightGBM hyperparameter optimization with Optuna will be skipped.")

torch_available = False
# torchdyn_available = False # Not used with DISABLE_NEURAL_ODE = True
# NeuralODE = None
DISABLE_NEURAL_ODE = True
# DISABLE_AUTOFORMER = True # This was for HuggingFace Autoformer, custom AutoformerPredictor is different
try:
    import torch
    import torch.nn as nn # Import nn
    import torch.optim as optim # Import optim
    torch_available = True
    print("PyTorch imported successfully.")
    if torch.cuda.is_available():
        print(f"PyTorch CUDA available: True, Version: {torch.version.cuda}")
        print(f"Using PyTorch on GPU: {torch.cuda.get_device_name(0)}")
    else:
        print("PyTorch CUDA available: False.")

    # Neural ODE part remains disabled as per original
    if not DISABLE_NEURAL_ODE:
        pass # Original logic was here, but it's disabled
    else:
        print("Neural ODE features are DISABLED by configuration.")
except ImportError:
    print("PyTorch not found. Neural ODE and Custom Transformer features will be SKIPPED.")


sktime_available = False
ClaSPSegmentation = None
# plot_series = None # Not explicitly used, can be removed if sktime plotting isn't called
# skbase_available = False # skbase is a dependency of sktime often, not directly used here
try:
    import sktime
    # print(f"Attempting to import sktime specific modules for sktime version: {sktime.__version__}")
    from sktime.annotation.clasp import ClaSPSegmentation
    # from sktime.utils.plotting import plot_series # Uncomment if used
    sktime_available = True
    print(f"Sktime {sktime.__version__} imported successfully (ClaSPSegmentation ready).")
except ImportError as e_sktime_import:
    print(f"Sktime modules import failed: {e_sktime_import}. ClaSP Regime detection will be skipped.")
except Exception as e_sktime_general:
    print(f"A general error occurred during sktime import or setup: {e_sktime_general}")


dowhy_available = False
CausalModel = None
nx = None # networkx - defined when dowhy is imported
# causalml_available = False # Not used as it wasn't found in user's log for critical path
try:
    import dowhy
    from dowhy import CausalModel
    import networkx as nx # Import networkx here
    dowhy_available = True
    print(f"DoWhy {dowhy.__version__} and NetworkX {nx.__version__} imported successfully.")
    # CausalML check removed as it wasn't critical and not found in user logs
except ImportError:
    print("DoWhy or NetworkX not found. Causal Discovery will be skipped.")


onnx_available = False
ort_available = False # onnxruntime
skl2onnx_available = False
onnxmltools_available = False
FloatTensorType = None
try:
    import onnx
    onnx_available = True
    import onnxruntime as ort
    # ort_available = True # Implied by successful import
    import skl2onnx
    from skl2onnx.common.data_types import FloatTensorType # Import here
    # skl2onnx_available = True # Implied
    import onnxmltools
    # onnxmltools_available = True # Implied
    print("ONNX, ONNXRuntime, skl2onnx, and onnxmltools imported successfully.")
    if hasattr(onnxmltools, '__version__'):
          print(f"Onnxmltools version: {onnxmltools.__version__}")
except ImportError as e:
    print(f"One or more ONNX components not found: {e}. ONNX features will be skipped.")

print("\nAll libraries and modules conditional imports attempted.")

# --- Constants ---
TWELVE_DATA_API_KEY = "b6dbb92e551a46f2b20de27540aeef0a" # IMPORTANT: Replace with your actual key
API_KEY = TWELVE_DATA_API_KEY # Ensure this uses the above variable
DEFAULT_SYMBOL = "MSFT"
START_DATE = (datetime.now() - timedelta(days=3*365)).strftime('%Y-%m-%d')
END_DATE = datetime.now().strftime('%Y-%m-%d')
AUTOFORMER_MODEL_PATH = "autoformer_predictor_trained.pth" # Path for saving/loading trained model

# --- Function Definitions ---

class AutoformerPredictor:
    def __init__(self, input_len=60, pred_len=5, d_model=64, n_heads=8, num_encoder_layers=2, num_decoder_layers=1, dim_feedforward=256):
        self.device = 'cuda' if torch.cuda.is_available() and torch_available else 'cpu'
        self.model = None
        self.input_len = input_len
        self.pred_len = pred_len
        self.d_model = d_model # Store for saving/loading context if needed

        if torch_available:
            print(f"ðŸ”§ Initializing Custom AutoformerPredictor on device: {self.device}")
            self.model = nn.Transformer(
                d_model=d_model, nhead=n_heads, num_encoder_layers=num_encoder_layers,
                num_decoder_layers=num_decoder_layers, dim_feedforward=dim_feedforward,
                activation='gelu', batch_first=False # PyTorch Transformer expects (S, N, E)
            ).to(self.device)
            # Input embedding: projects single feature per time step to d_model
            self.enc_embedding = nn.Linear(1, d_model).to(self.device)
            self.dec_embedding = nn.Linear(1, d_model).to(self.device)
            # Output projection: projects d_model back to single feature per time step
            self.projection = nn.Linear(d_model, 1).to(self.device)
        else:
            print("PyTorch not available. AutoformerPredictor cannot be initialized.")

    def _generate_square_subsequent_mask(self, sz):
        mask = (torch.triu(torch.ones(sz, sz)) == 1).transpose(0, 1)
        mask = mask.float().masked_fill(mask == 0, float('-inf')).masked_fill(mask == 1, float(0.0))
        return mask.to(self.device)

    def forward(self, src, tgt): # src: (S_enc, N, 1), tgt: (S_dec, N, 1)
        if not torch_available or self.model is None: return None

        src_embedded = self.enc_embedding(src) # (S_enc, N, d_model)
        tgt_embedded = self.dec_embedding(tgt) # (S_dec, N, d_model)


        tgt_mask = self._generate_square_subsequent_mask(tgt.size(0))

        output = self.model(src_embedded, tgt_embedded, tgt_mask=tgt_mask) # (S_dec, N, d_model)
        output = self.projection(output) # (S_dec, N, 1)
        return output

    def train_model(self, train_loader, epochs=10, learning_rate=0.001):
        if not torch_available or self.model is None:
            print("Cannot train: PyTorch or model not available.")
            return
        if not train_loader:
            print("Cannot train: No data loader provided.")
            return

        print(f"Starting AutoformerPredictor training for {epochs} epochs...")
        self.model.train()
        optimizer = optim.Adam(self.model.parameters(), lr=learning_rate)
        criterion = nn.MSELoss()

        for epoch in range(epochs):
            epoch_loss = 0
            for i, (batch_src, batch_tgt_input, batch_tgt_output) in enumerate(train_loader):
                # Shapes: batch_src (N, S_enc, 1), batch_tgt_input (N, S_dec, 1), batch_tgt_output (N, S_dec, 1)
                batch_src = batch_src.float().to(self.device).permute(1,0,2) # (S_enc, N, 1)
                batch_tgt_input = batch_tgt_input.float().to(self.device).permute(1,0,2) # (S_dec, N, 1)
                batch_tgt_output = batch_tgt_output.float().to(self.device).permute(1,0,2) # (S_dec, N, 1)

                optimizer.zero_grad()
                predictions = self.forward(batch_src, batch_tgt_input) # (S_dec, N, 1)

                if predictions is None:
                    print(f"Epoch {epoch+1}, Batch {i+1}: Forward pass returned None. Skipping batch.")
                    continue

                loss = criterion(predictions, batch_tgt_output)
                loss.backward()
                optimizer.step()
                epoch_loss += loss.item()
            
            avg_epoch_loss = epoch_loss / len(train_loader) if len(train_loader) > 0 else float('nan')
            print(f"Epoch {epoch+1}/{epochs}, Average Loss: {avg_epoch_loss:.4f}")
        print("Training finished.")
        self.save_model(AUTOFORMER_MODEL_PATH)


    def predict(self, series_data): # series_data is historical data
        if not torch_available or self.model is None:
            print("Cannot predict: PyTorch or model not available.")
            return np.array([np.nan] * self.pred_len)

        self.model.eval()
        with torch.no_grad():
            if isinstance(series_data, pd.Series):
                series_data = series_data.values
            if isinstance(series_data, list):
                series_data = np.array(series_data)

            if not isinstance(series_data, np.ndarray):
                print(f"Unsupported series data type for predict: {type(series_data)}")
                return np.array([np.nan] * self.pred_len)

            # Ensure series_data is 1D
            if series_data.ndim > 1:
                series_data = series_data.squeeze()
            if series_data.ndim == 0: # scalar
                 print(f"Series data is scalar, cannot process.")
                 return np.array([np.nan]*self.pred_len)


            # Normalization of the input series (encoder input)
            mean_val = series_data[-self.input_len:].mean() if len(series_data) >= self.input_len else series_data.mean()
            std_val = series_data[-self.input_len:].std() if len(series_data) >= self.input_len else series_data.std()
            std_val = std_val if std_val > 1e-8 else 1e-8 # Avoid division by zero

            if len(series_data) < self.input_len:
                padding_val = series_data[0] if len(series_data) > 0 else 0 # Simple padding
                padding = np.full(self.input_len - len(series_data), padding_val)
                norm_series_np = (np.concatenate([padding, series_data]) - mean_val) / std_val
            else:
                norm_series_np = (series_data[-self.input_len:] - mean_val) / std_val
            
            src_tensor = torch.tensor(norm_series_np, dtype=torch.float32).unsqueeze(0).unsqueeze(-1) # (N=1, S_enc, 1)
            src_tensor = src_tensor.permute(1,0,2).to(self.device) # (S_enc, N=1, 1) for nn.Transformer

            # Auto-regressive decoding
            # Start with a zero or a start token for the decoder input
            decoder_input_np = np.zeros((1, 1, 1)) # (N, S_dec_step=1, 1)
            decoder_input = torch.tensor(decoder_input_np, dtype=torch.float32).permute(1,0,2).to(self.device) # (S_dec_step=1, N, 1)
            
            predictions_normalized = []

            for _ in range(self.pred_len):
                prediction_step = self.forward(src_tensor, decoder_input) # (S_dec_step, N, 1)
                
                if prediction_step is None: # Should not happen if model is available
                    return np.array([np.nan] * self.pred_len)


                last_pred_step = prediction_step[-1:, :, :] # Get the last time step (1, N, 1)
                predictions_normalized.append(last_pred_step.squeeze().cpu().item()) # Store normalized prediction
                

                if len(predictions_normalized) < self.pred_len : # Only if we need more predictions

                    new_input_step_embedded = self.dec_embedding(last_pred_step) # Embed the last prediction

                    
                    current_tgt_sequence_normalized = torch.tensor(predictions_normalized, dtype=torch.float32).unsqueeze(0).unsqueeze(-1) # (N=1, current_len, 1)
                    current_tgt_sequence_normalized = current_tgt_sequence_normalized.permute(1,0,2).to(self.device) # (current_len, N=1, 1)
                    decoder_input = current_tgt_sequence_normalized # Update decoder input for next step
                    if decoder_input.shape[0] > self.pred_len: # Should not happen due to loop limit
                        decoder_input = decoder_input[:self.pred_len,:,:]


            denorm_preds = (np.array(predictions_normalized) * std_val) + mean_val
            return denorm_preds

    def save_model(self, path):
        if not torch_available or self.model is None: return
        # Save all necessary components: model state, embeddings, projection
        torch.save({
            'model_state_dict': self.model.state_dict(),
            'enc_embedding_state_dict': self.enc_embedding.state_dict(),
            'dec_embedding_state_dict': self.dec_embedding.state_dict(),
            'projection_state_dict': self.projection.state_dict(),
            'input_len': self.input_len,
            'pred_len': self.pred_len,
            'd_model': self.d_model
        }, path)
        print(f"AutoformerPredictor model saved to {path}")

    def load_model(self, path):
        if not torch_available:
            print("Cannot load model: PyTorch not available.")
            return False
        if not os.path.exists(path):
            print(f"Model file not found at {path}. Initializing new model.")
            return False # Indicates that a new model will be used / needs training

        checkpoint = torch.load(path, map_location=self.device)

        self.input_len = checkpoint.get('input_len', self.input_len)
        self.pred_len = checkpoint.get('pred_len', self.pred_len)


        self.model = nn.Transformer( # Recreate model structure
                d_model=checkpoint.get('d_model', self.d_model), # Use saved d_model if available
                nhead=self.model.nhead if self.model else 8, # Assuming nhead is fixed or stored
                num_encoder_layers=len(self.model.encoder.layers) if self.model else 2,
                num_decoder_layers=len(self.model.decoder.layers) if self.model else 1,
                dim_feedforward=self.model.encoder.layers[0].linear2.out_features if self.model and self.model.encoder.layers else 256,
                activation='gelu', batch_first=False
            ).to(self.device)
        self.enc_embedding = nn.Linear(1, checkpoint.get('d_model', self.d_model)).to(self.device)
        self.dec_embedding = nn.Linear(1, checkpoint.get('d_model', self.d_model)).to(self.device)
        self.projection = nn.Linear(checkpoint.get('d_model', self.d_model), 1).to(self.device)

        self.model.load_state_dict(checkpoint['model_state_dict'])
        self.enc_embedding.load_state_dict(checkpoint['enc_embedding_state_dict'])
        self.dec_embedding.load_state_dict(checkpoint['dec_embedding_state_dict'])
        self.projection.load_state_dict(checkpoint['projection_state_dict'])
        self.model.eval() # Set to evaluation mode
        print(f"AutoformerPredictor model loaded from {path}")
        return True # Successfully loaded


def create_sequences(data, input_len, pred_len):
    """ Helper function to create input and target sequences for Autoformer training """
    X, y_input, y_output = [], [], []
    for i in range(len(data) - input_len - pred_len + 1):
        X.append(data[i:(i + input_len)])
        y_input.append(data[(i + input_len -1):(i + input_len + pred_len -1)]) # Decoder input uses last known + future predictions
        y_output.append(data[(i + input_len):(i + input_len + pred_len)])   # What we want to predict
    return np.array(X), np.array(y_input), np.array(y_output)


def train_or_load_autoformer(df_series, input_len=60, pred_len=5, force_train=False):
    """
    Trains the AutoformerPredictor model or loads it if a saved version exists.
    Returns an instance of the trained/loaded AutoformerPredictor.
    """
    if not torch_available:
        print("PyTorch not available, cannot train or load Autoformer.")
        return None

    autoformer_forecaster = AutoformerPredictor(input_len=input_len, pred_len=pred_len)
    
    if not force_train and os.path.exists(AUTOFORMER_MODEL_PATH):
        print(f"Loading existing Autoformer model from {AUTOFORMER_MODEL_PATH}")
        if autoformer_forecaster.load_model(AUTOFORMER_MODEL_PATH):
            return autoformer_forecaster
        else:
            print(f"Failed to load model from {AUTOFORMER_MODEL_PATH}. Proceeding to train a new one.")
            force_train = True # Force training if loading failed

    if force_train:
        print("Training a new AutoformerPredictor model...")
        # Normalize the series for training
        scaler = StandardScaler()
        series_scaled = scaler.fit_transform(df_series.values.reshape(-1, 1))

        # Create sequences
        # Note: For simplicity, decoder input here is just shifted target.
        # More sophisticated approaches might use a start token and rely on model to generate.
        X_seq, y_seq_dec_in, y_seq_out = create_sequences(series_scaled.flatten(), input_len, pred_len)

        if len(X_seq) == 0:
            print("Not enough data to create sequences for Autoformer training. Skipping training.")
            return autoformer_forecaster # Return un-trained model

        # Reshape for PyTorch: (N, S, F) where N is batch_size, S is sequence_len, F is num_features
        X_seq = X_seq.reshape(X_seq.shape[0], X_seq.shape[1], 1)
        y_seq_dec_in = y_seq_dec_in.reshape(y_seq_dec_in.shape[0], y_seq_dec_in.shape[1], 1)
        y_seq_out = y_seq_out.reshape(y_seq_out.shape[0], y_seq_out.shape[1], 1)

        # Create DataLoader
        dataset = torch.utils.data.TensorDataset(
            torch.from_numpy(X_seq).float(),
            torch.from_numpy(y_seq_dec_in).float(),
            torch.from_numpy(y_seq_out).float()
        )
        # Use a small batch size for demonstration. Tune as needed.
        train_loader = torch.utils.data.DataLoader(dataset, batch_size=16, shuffle=True)

        autoformer_forecaster.train_model(train_loader, epochs=20) # epochs=20 for quicker demo
        # autoformer_forecaster.save_model(AUTOFORMER_MODEL_PATH) # save_model is called in train_model
    else:
         print("Autoformer training skipped (force_train=False and no existing model or load failed). Model is un-trained.")


    return autoformer_forecaster


def fetch_twelve_data(symbol, api_key, start_date_str=None, end_date_str=None):
    base_url = "https://api.twelvedata.com/time_series"
    params = {
        "symbol": symbol, "interval": "1day", "apikey": api_key, "format": "JSON",
        "outputsize": 5000 # Request more data if possible
    }
    if start_date_str: params["start_date"] = start_date_str
    if end_date_str: params["end_date"] = end_date_str
    print(f"Fetching data for {symbol} from Twelve Data (interval=1day, from {start_date_str} to {end_date_str})...")
    try:
        response = requests.get(base_url, params=params, timeout=30)
        response.raise_for_status()
        data = response.json()
    except requests.exceptions.RequestException as e:
        print(f"Request failed for {symbol}: {e}"); return None
    except ValueError as e: # Includes JSONDecodeError
        print(f"Failed to parse JSON for {symbol}: {e}. Response: {response.text[:200]}..."); return None

    if isinstance(data, dict) and (data.get("status") == "error" or "values" not in data):
        print(f"API Error for {symbol} (Code: {data.get('code')}): {data.get('message', 'Unknown error')}"); return None
    if not isinstance(data, dict) or "values" not in data or not data["values"]: # Check if list is empty
        print(f"No data values for {symbol} for the period, or unexpected data format."); return None

    df = pd.DataFrame(data["values"]).rename(columns={'datetime': 'date'})
    # Ensure OHLCV columns are float64. This is crucial for pandas_ta.
    for col in ['open', 'high', 'low', 'close', 'volume']:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce').astype('float64')
        else: # If essential columns like 'close' are missing after fetch
            if col in ['open', 'high', 'low', 'close']:
                print(f"Critical column '{col}' missing from fetched data. Cannot proceed.")
                return None
            df[col] = 0.0 # Default for volume if missing, though not ideal

    if 'date' not in df.columns:
        print("Critical Error: 'date' column missing after renaming."); return None

    df.index = pd.to_datetime(df['date'])
    df.drop(columns=['date'], inplace=True)
    df.sort_index(inplace=True)
    # Drop rows if key price data is NaN after conversion
    df.dropna(subset=[col for col in ['open', 'high', 'low', 'close'] if col in df.columns], inplace=True)

    if df.empty:
        print(f"No data remaining for {symbol} after processing."); return None
    print(f"Successfully fetched/processed {len(df)} data points for {symbol}."); return df

def add_technical_indicators(df):
    df_feat = df.copy()
    if not hasattr(df_feat.ta, 'rsi'): # Check if ta methods are available
        print("pandas_ta not available or not correctly registered on DataFrame. Skipping TIs."); return df_feat
    print("Adding technical indicators...")
    try:
        # Ensure required columns are present and float, if not already guaranteed
        ohlcv_cols = ['open', 'high', 'low', 'close', 'volume']
        for col in ohlcv_cols:
            if col not in df_feat.columns:
                print(f"Warning: Column '{col}' not found for TIs. Some TIs may fail or be inaccurate.")
                # Create a dummy column if essential for pandas_ta to not crash, though results would be meaningless
                if col in ['high', 'low', 'open', 'close']: df_feat[col] = df_feat['close'] # Risky, but for structure
                elif col == 'volume': df_feat[col] = 0.0

        c, h, l, v = 'close', 'high', 'low', 'volume' # Explicitly use these main columns

        df_feat.ta.rsi(close=df_feat[c], length=14, append=True) # Default name: RSI_14
        df_feat.ta.rsi(close=df_feat[c], length=9, append=True, col_names='RSI_9')
        df_feat.ta.rsi(close=df_feat[c], length=25, append=True, col_names='RSI_25')

        df_feat.ta.macd(close=df_feat[c], fast=12, slow=26, signal=9, append=True)
        df_feat.ta.macd(close=df_feat[c], fast=5, slow=15, signal=9, append=True, col_names=('MACD_5_15_9', 'MACDh_5_15_9', 'MACDs_5_15_9'))

        for p in [10, 20, 50, 100, 200]:
            df_feat.ta.sma(close=df_feat[c], length=p, append=True) # Default name: SMA_p
            df_feat.ta.ema(close=df_feat[c], length=p, append=True) # Default name: EMA_p

        df_feat.ta.bbands(close=df_feat[c], length=20, std=2, append=True)

        if all(x in df_feat.columns for x in [h,l,c]):
            df_feat.ta.atr(high=df_feat[h], low=df_feat[l], close=df_feat[c], length=14, append=True)
            df_feat.ta.adx(high=df_feat[h], low=df_feat[l], close=df_feat[c], length=14, append=True)
            df_feat.ta.stoch(high=df_feat[h], low=df_feat[l], close=df_feat[c], append=True)
            df_feat.ta.willr(high=df_feat[h], low=df_feat[l], close=df_feat[c], append=True)
            df_feat.ta.cci(high=df_feat[h], low=df_feat[l], close=df_feat[c], append=True)

        if all(x in df_feat.columns for x in [h,l,c,v]):
             # Ensure volume is float type to potentially avoid MFI dtype issues
            df_feat[v] = df_feat[v].astype('float64')
            df_feat.ta.mfi(high=df_feat[h], low=df_feat[l], close=df_feat[c], volume=df_feat[v], append=True)

        # Sanitize all column names after adding all TIs
        df_feat.columns = df_feat.columns.str.replace('[^A-Za-z0-9_]+', '', regex=True)
        print("Technical indicators added.")
    except Exception as e:
        print(f"Error adding TIs: {e}\n{traceback.format_exc()}")
    return df_feat

def add_optimized_features(df, price_col='close', volume_col='volume'):
    print("Adding optimized statistical and ratio features...")
    df_new = df.copy()
    df_new['returns'] = df_new[price_col].pct_change()
    safe_price = df_new[price_col].replace(0, np.nan) # Avoid log(0)
    safe_price_shifted = df_new[price_col].shift(1).replace(0, np.nan)
    df_new['log_returns'] = np.log(safe_price / safe_price_shifted) # Use safe versions

    for window in [5, 10, 20, 50]: # Added 50d window
        df_new[f'volatility_{window}'] = df_new['log_returns'].rolling(window).std() # Use log_returns for volatility
        df_new[f'skew_{window}'] = df_new['log_returns'].rolling(window).skew()
        df_new[f'kurtosis_{window}'] = df_new['log_returns'].rolling(window).kurt()

    if volume_col in df_new.columns and price_col in df_new.columns:
        rolling_mean_volume = df_new[volume_col].rolling(20).mean().replace(0, np.nan) # Avoid division by zero
        df_new['volume_ratio'] = df_new[volume_col] / rolling_mean_volume
        df_new['price_volume'] = df_new[price_col] * df_new[volume_col]
        df_new['volume_change'] = df_new[volume_col].pct_change()

    if all(col in df_new.columns for col in ['high', 'low', 'close']):
        safe_low = df_new['low'].replace(0, np.nan)
        safe_high = df_new['high'].replace(0, np.nan)
        safe_close = df_new['close'].replace(0, np.nan) # Use current close for intraday range denom

        df_new['high_low_ratio'] = df_new['high'] / safe_low
        df_new['close_to_high_ratio'] = safe_close / safe_high # Renamed for clarity
        df_new['close_to_low_ratio'] = safe_close / safe_low   # Renamed for clarity
        df_new['intraday_range_norm'] = (df_new['high'] - df_new['low']) / safe_close # Normalized by close

    # Ensure standard column names from pandas_ta for RSI and MACD
    # Example: RSI_14, MACD_12_26_9, MACDs_12_26_9
    if 'RSI_14' in df_new.columns:
        df_new['RSI_signal'] = 0
        df_new.loc[df_new['RSI_14'] < 30, 'RSI_signal'] = 1  # Oversold
        df_new.loc[df_new['RSI_14'] > 70, 'RSI_signal'] = -1 # Overbought

    # Default MACD column names from pandas_ta are MACD_12_26_9 and MACDs_12_26_9
    if 'MACD_12_26_9' in df_new.columns and 'MACDs_12_26_9' in df_new.columns:
        df_new['MACD_signal_line_cross'] = (df_new['MACD_12_26_9'] > df_new['MACDs_12_26_9']).astype(int)
    return df_new

def add_wavelet_features(df, column='close', wavelet='mexh', scales_range=(1, 32), num_scales_to_extract=5):
    df_feat = df.copy()
    print(f"Adding CWT for '{column}' using wavelet '{wavelet}'...")
    if not torch_available: # pywt is a dependency of torch, but explicitly check pywt
         if pywt is None: print("PyWavelets not available."); return df_feat
    
    signal = df_feat[column].values
    if len(signal) < scales_range[1] + 5:
        print(f"Signal length {len(signal)} too short for CWT. Skipping."); return df_feat
    
    actual_max_scale = min(scales_range[1], len(signal) // 2 - 1)
    if actual_max_scale < scales_range[0]:
        print("Max scale too small after constraint. Skipping CWT."); return df_feat
        
    scales = np.arange(scales_range[0], actual_max_scale + 1)
    if len(scales) == 0:
        print("No valid scales for CWT. Skipping."); return df_feat
        
    try:
        coefficients, _ = pywt.cwt(signal, scales, wavelet) # Coeffs shape: (num_scales, len_signal)
        coeffs_df = pd.DataFrame(coefficients.T, index=df_feat.index, columns=[f"cwt_scale_{s}" for s in scales])
        
        df_feat[f'{column}_cwt_mean'] = coeffs_df.mean(axis=1)
        df_feat[f'{column}_cwt_std'] = coeffs_df.std(axis=1)
        

        s_indices_to_extract = np.linspace(0, len(scales)-1, min(num_scales_to_extract, len(scales)), dtype=int)

        for s_idx in s_indices_to_extract:
            actual_scale_val = scales[s_idx] # Get the actual scale value
            # Use the column name from coeffs_df which corresponds to this scale
            col_name_for_scale = f"cwt_scale_{actual_scale_val}"
            if col_name_for_scale in coeffs_df.columns:
                 df_feat[f'{column}_cwt_energy_s{actual_scale_val}'] = coeffs_df[col_name_for_scale]**2
            else: # Should not happen if logic is correct
                 print(f"Warning: Could not find {col_name_for_scale} in CWT coefficients during energy extraction.")

        print("Wavelet features added.")
    except Exception as e:
        print(f"Error adding wavelet features: {e}\n{traceback.format_exc()}")
    return df_feat

def add_entropy_features(df, column='close', window=40):
    df_feat = df.copy()
    print(f"Adding Entropy for '{column}' (window={window})...")
    if ant is None: print("Antropy not available."); return df_feat
    if len(df_feat) < window + 15: # Need enough data points for rolling window
        print(f"Data too short for entropy (window={window}). Skipping."); return df_feat
    try:
        sig = df_feat[column].astype(float) # Ensure float for antropy functions
        # Sample Entropy - robust to NaNs if they are handled before lambda
        df_feat[f'{column}_entropy_sample'] = sig.rolling(window=window, min_periods=window).apply(
            lambda x: ant.sample_entropy(x.dropna()) if x.dropna().shape[0] >= window//2 else np.nan, raw=False # Ensure enough non-NaNs
        )
        # Spectral Entropy - requires no NaNs in window
        df_feat[f'{column}_entropy_spectral'] = sig.rolling(window=window, min_periods=window).apply(
            lambda x: ant.spectral_entropy(x.dropna(), sf=1.0, method='welch', nperseg=min(x.dropna().shape[0], window // 2 if window // 2 > 0 else 1))
                      if x.dropna().shape[0] == window else np.nan, raw=False # Process only if full non-NaN window
        )
        print("Entropy features added.")
    except Exception as e:
        print(f"Error adding entropy features: {e}\n{traceback.format_exc()}")
    return df_feat


def add_advanced_technical_features(df, price_col='close', high_col='high', low_col='low', volume_col='volume'):
    print("Adding advanced TIs...")
    df_new = df.copy()
    if not hasattr(df_new.ta, 'mom'): print("pandas_ta not available or not registered. Skipping advanced TIs."); return df_new
    try:
        if not all(c in df_new.columns for c in [price_col, high_col, low_col]):
            print("Missing OHLC for advanced TIs. Skipping."); return df_new

        df_new.ta.mom(close=df_new[price_col], append=True) # MOM_10
        df_new.ta.roc(close=df_new[price_col], append=True) # ROC_10
        df_new.ta.natr(high=df_new[high_col], low=df_new[low_col], close=df_new[price_col], append=True) # NATR_14
        df_new.ta.aroon(high=df_new[high_col], low=df_new[low_col], append=True) # AROOND_14, AROONU_14, AROONOSC_14
        
        # The STC warning about Series.__getitem__ is usually internal to pandas-ta.
        # Ensure inputs are Series of floats.
        df_new.ta.stc(close=df_new[price_col].astype(float), tclength=23, fast=50, slow=100, factor=0.5, append=True,
                      col_names=('STC_23_50_05', 'STCD_23_50_05', 'STCK_23_50_05'))

        if volume_col in df_new.columns:
            df_new.ta.pvol(close=df_new[price_col], volume=df_new[volume_col].astype(float), append=True) # PVOL
            df_new.ta.cmf(high=df_new[high_col], low=df_new[low_col], close=df_new[price_col],
                          volume=df_new[volume_col].astype(float), append=True) # CMF_20

        df_new.columns = df_new.columns.str.replace('[^A-Za-z0-9_]+', '', regex=True) # Sanitize names
        print("Advanced TIs added.")
    except Exception as e:
        print(f"Error adding advanced TIs: {e}\n{traceback.format_exc()}")
    return df_new

def add_transformer_features_conceptual(df, column='close', sequence_length=20):
    df_feat = df.copy() # Work on a copy
    print(f"Adding Transformer-inspired conceptual features for '{column}' (sequence_length={sequence_length})...")

    if len(df_feat) < sequence_length + 5: # Need enough data
        print(f"Data too short for conceptual transformer features ({len(df_feat)} points). Skipping."); return df_feat

    feature_col_base = f"{column}_trans_seq"
    feature_col_names = [f'{feature_col_base}_mean', f'{feature_col_base}_std',
                         f'{feature_col_base}_trend', f'{feature_col_base}_volatility', f'{feature_col_base}_autocorr1']
    for col_name in feature_col_names:
        df_feat[col_name] = np.nan # Initialize columns

    try:
        data_series = df_feat[column].values
        # Create rolling windows (sequences)
        windows = np.lib.stride_tricks.sliding_window_view(data_series, sequence_length) # More efficient

        means, stds, trends, volatilities, autocorrs = [], [], [], [], []

        for seq in windows:
            # Normalize each sequence (z-score)
            mean_val, std_val = np.mean(seq), np.std(seq)
            if std_val > 1e-8: # Avoid division by zero for flat sequences
                norm_seq = (seq - mean_val) / std_val
            else:
                norm_seq = np.zeros_like(seq) # Flat sequence becomes all zeros

            means.append(np.mean(norm_seq)) # Mean of normalized sequence (should be ~0)
            stds.append(np.std(norm_seq))   # Std of normalized sequence (should be ~1 if not flat)

            if len(norm_seq) > 1:
                try:
                    slope = np.polyfit(np.arange(len(norm_seq)), norm_seq, 1)[0]
                    trends.append(slope if not np.isnan(slope) else 0.0)
                except (np.linalg.LinAlgError, ValueError, TypeError):
                    trends.append(0.0)

                diff_norm_seq = np.diff(norm_seq)
                volatilities.append(np.std(diff_norm_seq) if len(diff_norm_seq) > 0 else 0.0)

                if len(norm_seq) >= 2:
                    s1, s2 = norm_seq[:-1], norm_seq[1:]
                    if len(s1) >= 1 and np.std(s1) > 1e-8 and np.std(s2) > 1e-8: # check for variance
                        try:
                            autocorr = np.corrcoef(s1, s2)[0, 1]
                            autocorrs.append(autocorr if not np.isnan(autocorr) else 0.0)
                        except Exception: autocorrs.append(0.0)
                    else: autocorrs.append(0.0)
                else: autocorrs.append(0.0)
            else: # Should not happen if sequence_length > 1
                trends.append(0.0); volatilities.append(0.0); autocorrs.append(0.0)


        idx_start = sequence_length - 1
        df_feat.loc[df_feat.index[idx_start:idx_start+len(means)], f'{feature_col_base}_mean'] = means
        df_feat.loc[df_feat.index[idx_start:idx_start+len(stds)], f'{feature_col_base}_std'] = stds
        df_feat.loc[df_feat.index[idx_start:idx_start+len(trends)], f'{feature_col_base}_trend'] = trends
        df_feat.loc[df_feat.index[idx_start:idx_start+len(volatilities)], f'{feature_col_base}_volatility'] = volatilities
        df_feat.loc[df_feat.index[idx_start:idx_start+len(autocorrs)], f'{feature_col_base}_autocorr1'] = autocorrs
        
        print("Conceptual Transformer-inspired features added.")
    except Exception as e:
        print(f"Error adding conceptual Transformer features: {e}\n{traceback.format_exc()}")
    return df_feat


def add_clasp_regimes(df, column='close', period=20, n_cps=3): # period and n_cps are ClaSP params
    df_reg = df.copy()
    print(f"Detecting regimes using ClaSPSegmentation for {column}...")
    if not sktime_available or ClaSPSegmentation is None:
        print("ClaSPSegmentation not available. Using simple volatility regime detection as fallback.")
        return detect_regimes_simple(df, column) # Fallback to simpler method

    if column not in df_reg.columns or df_reg[column].isnull().all():
        print(f"'{column}' not found or all NaN. Skipping ClaSP regime detection."); df_reg['regime'] = 0; return df_reg
    
    series_for_clasp = df_reg[column].dropna()
    if len(series_for_clasp) < period * 2: # ClaSP needs sufficient data
        print(f"Series length {len(series_for_clasp)} too short for ClaSP with period {period}. Using simple fallback.")
        return detect_regimes_simple(df, column)

    try:
        clasp = ClaSPSegmentation(period_length=period, n_cps=n_cps)
        found_cps = clasp.fit_predict(series_for_clasp) # Returns change point indices

        # Create regime labels based on change points
        # This is a simplified way to assign regimes; more sophisticated logic could be used
        df_reg['regime'] = 0 # Default regime
        current_regime = 0
        last_cp = 0
        for cp in found_cps:
            # Assign regime to segment between last_cp and cp
            # series_for_clasp.index are original indices from df_reg[column].dropna()
            segment_indices = series_for_clasp.index[last_cp:cp]
            df_reg.loc[segment_indices, 'regime'] = current_regime
            current_regime = (current_regime + 1) % (n_cps + 1) # Cycle through regimes
            last_cp = cp
        # Assign regime to the last segment
        segment_indices = series_for_clasp.index[last_cp:]
        df_reg.loc[segment_indices, 'regime'] = current_regime
        
        # Forward fill regimes for original NaNs in the input column, then backfill
        df_reg['regime'] = df_reg['regime'].ffill().bfill().fillna(0)

        print(f"ClaSP regimes identified. Distribution:\n{df_reg['regime'].value_counts(normalize=True).sort_index()*100} %")
    except Exception as e:
        print(f"Error during ClaSPSegmentation: {e}. Using simple volatility regime detection as fallback.")
        return detect_regimes_simple(df, column)
    return df_reg


def detect_regimes_simple(df, column='close', window=20): # Added window param
    df_reg = df.copy()
    print(f"Detecting regimes (simplified volatility-based) for {column}...")
    if column not in df_reg.columns:
        print(f"'{column}' not found. Skipping regime detection."); df_reg['regime_simple'] = 0; return df_reg

    # Use log returns for volatility calculation for better statistical properties
    returns = np.log(df_reg[column] / df_reg[column].shift(1)).dropna()
    rolling_vol = returns.rolling(window=window, min_periods=window//2).std() # Ensure min_periods
    df_reg['regime_simple'] = 0 # Default: medium volatility

    if not rolling_vol.dropna().empty:
        vol_low_thresh = rolling_vol.quantile(0.33)
        vol_high_thresh = rolling_vol.quantile(0.67)
        df_reg.loc[rolling_vol.index[rolling_vol <= vol_low_thresh], 'regime_simple'] = 1  # Low vol
        df_reg.loc[rolling_vol.index[rolling_vol > vol_high_thresh], 'regime_simple'] = 2   # High vol
    else:
        print("Not enough data for volatility percentile calculation. Defaulting simple regimes to 0.")
    # Fill any NaNs in regime_simple (e.g., at the beginning due to rolling window)
    df_reg['regime_simple'] = df_reg['regime_simple'].fillna(method='bfill').fillna(0)
    print(f"Simple Regimes (0:Med,1:Low,2:High): {df_reg['regime_simple'].value_counts(normalize=True).sort_index()*100} %")
    return df_reg

def balanced_target_definition(df, column='close', periods=5, lower_q_thresh=0.45, upper_q_thresh=0.55): # Tunable params
    df_t = df.copy()
    print(f"Balanced target definition for '{column}' over {periods} periods...")
    if column not in df_t.columns:
        print(f"'{column}' not found for target. Defaulting target."); df_t['target'] = 0; return df_t

    df_t[column] = pd.to_numeric(df_t[column], errors='coerce')
    # Future log return is generally preferred for target definition
    df_t['future_log_return_target'] = np.log(df_t[column].shift(-periods) / df_t[column])
    
    valid_returns = df_t['future_log_return_target'].dropna()
    df_t['target'] = 0 # Default class (e.g., Hold/Neutral)

    if len(valid_returns) > 20: # Need enough data for meaningful quantiles
        lower_q_val = valid_returns.quantile(lower_q_thresh)
        upper_q_val = valid_returns.quantile(upper_q_thresh)

        df_t.loc[df_t['future_log_return_target'] < lower_q_val, 'target'] = 0 # Returns below lower quantile -> Class 0 (e.g. Sell/Hold short)
        df_t.loc[df_t['future_log_return_target'] > upper_q_val, 'target'] = 1 # Returns above upper quantile -> Class 1 (e.g. Buy/Hold long)

    else:
        print("Not enough valid returns for quantile-based target balancing. Using default target 0.")

    df_t.drop(columns=['future_log_return_target'], inplace=True, errors='ignore')
    print(f"Target distribution:\n{df_t['target'].value_counts(normalize=True, dropna=False)*100}")
    return df_t


def discover_causal_structure(df_features, target_col='target', max_feats=10, symbol=""):
    print(f"\nDiscovering causal structure for {symbol} using DoWhy...")
    if not dowhy_available or CausalModel is None:
        print("DoWhy not available for causal discovery."); return None
    
    df_c = df_features.copy()
    if target_col not in df_c.columns or df_c[target_col].isnull().all():
        print(f"Target '{target_col}' missing or all NaN in causal discovery."); return None
    
    df_c[target_col] = pd.to_numeric(df_c[target_col], errors='coerce')
    
    # Select numeric, non-target, non-all-NaN columns
    cand_cols = [c for c in df_c.columns if pd.api.types.is_numeric_dtype(df_c[c]) and c != target_col and df_c[c].notnull().any()]
    if not cand_cols:
        print("No valid numeric candidate cause columns found for causal discovery."); return None

    df_subset_for_causal = df_c[cand_cols + [target_col]].copy()
    df_subset_for_causal.replace([np.inf, -np.inf], np.nan, inplace=True)
    
    # Scaling features for causal model to handle potentially large differences in magnitudes
    # This helps if linear regression is used for effect estimation.
    if cand_cols: # only scale if there are candidate columns
        scaler_causal = StandardScaler()
        df_subset_for_causal[cand_cols] = scaler_causal.fit_transform(df_subset_for_causal[cand_cols])
    
    df_subset_for_causal.dropna(inplace=True) # DoWhy typically needs complete data
    
    if df_subset_for_causal.empty or target_col not in df_subset_for_causal.columns or df_subset_for_causal[target_col].nunique() < 1:
        print("Causal discovery: not enough data or target variation after cleaning/scaling."); return None
        
    # Refined potential_causes list (can be dynamically selected too)
    potential_causes = ['RSI_14', 'MACDh_12_26_9', 'ADX_14', 'ATR_14', 'close_cwt_mean', 'close_cwt_std',
                        'close_entropy_sample', 'close_entropy_spectral', 'regime_simple', # Using simple regime here
                        'BBP_200_20', 'BBB_200_20', # Example pandas-ta default bband names
                        'volatility_20', 'log_returns'] # Added log_returns
                        # Check actual column names produced by pandas_ta if they changed
    
    graph_feats = [c for c in potential_causes if c in df_subset_for_causal.columns and c != target_col and df_subset_for_causal[c].nunique() > 1]

    if not graph_feats:
        print("Predefined causal graph_feats not suitable or not found after scaling/cleaning, selecting top varying features.")
        num_to_select = min(max_feats, len(cand_cols))
        if num_to_select > 0:
             graph_feats = df_subset_for_causal[cand_cols].var().nlargest(num_to_select).index.tolist()
        else:
             print("No candidate columns to select from for graph_feats fallback."); return None
    
    if not graph_feats: print("Causal discovery: no suitable graph features after all checks."); return None
        
    final_df_for_causal_model = df_subset_for_causal[graph_feats + [target_col]].copy()
    if final_df_for_causal_model.empty or final_df_for_causal_model.shape[0] < 20 or final_df_for_causal_model[target_col].nunique() < 1:
        print("Causal discovery: final DataFrame for model is too small or target has no variation."); return None
    
    print(f"DoWhy will use graph features: {graph_feats} for Outcome: {target_col}")

    treatment_var = graph_feats[0] # DoWhy needs at least one 'treatment' for model init
    common_causes_for_graph = [f for f in graph_feats if f != treatment_var]


    graph_str_list = [f'"{f}" -> "{target_col}"' for f in graph_feats]

    graph_str = "digraph { " + "; ".join(graph_str_list) + " }"

    print(f"Generated Causal Graph:\n{graph_str}")
    try:

        model = CausalModel(data=final_df_for_causal_model,
                            treatment=treatment_var, # Placeholder for model creation
                            outcome=target_col,
                            graph=graph_str)
        print("DoWhy CausalModel created (using placeholder treatment for init)."); return model
    except Exception as e:
        print(f"DoWhy CausalModel error: {e}\n{traceback.format_exc()}"); return None


def causal_feature_selection(df_full_features, target_col='target', max_causal_feats=10): # Limit number of features selected by this
    print("Performing Causal Feature Selection with DoWhy...")
    if not dowhy_available or CausalModel is None:
        print("DoWhy not available. Skipping causal feature selection."); return []

    # Prepare data specifically for this function: scale numeric features, drop NaNs
    df_clean = df_full_features.copy()
    df_clean[target_col] = pd.to_numeric(df_clean[target_col], errors='coerce')
    
    numeric_features_for_causal = [col for col in df_clean.columns if col != target_col and pd.api.types.is_numeric_dtype(df_clean[col]) and df_clean[col].nunique() > 1]
    if not numeric_features_for_causal:
        print("No suitable numeric features for Causal FS."); return []

    # Scale these numeric features
    scaler_cfs = StandardScaler()
    df_clean[numeric_features_for_causal] = scaler_cfs.fit_transform(df_clean[numeric_features_for_causal])
    df_clean.replace([np.inf, -np.inf], np.nan, inplace=True)
    df_clean.dropna(subset=numeric_features_for_causal + [target_col], inplace=True) # Drop rows with NaNs in features or target

    if df_clean.empty or target_col not in df_clean.columns or df_clean[target_col].nunique() < 1:
        print("Target missing, no variation, or data empty after cleaning for Causal FS."); return []
    
    # Re-select features after potential row drops from dropna
    final_features_for_causal = [f for f in numeric_features_for_causal if f in df_clean.columns and df_clean[f].nunique() >1]
    if not final_features_for_causal :
        print("No features remaining after final cleaning for Causal FS."); return []


    print(f"Starting causal effect estimation for {len(final_features_for_causal)} features.")
    feature_effects = {}
    for feature_as_treatment in final_features_for_causal:
        try:

            current_data_for_model = df_clean[final_features_for_causal + [target_col]]

            graph_str_single_treatment_list = [f'"{feature_as_treatment}" -> "{target_col}"']

            
            model = CausalModel(data=current_data_for_model, # Use current_data to ensure consistency if any feature was all NaN and dropped
                                treatment=feature_as_treatment,
                                outcome=target_col,
                                graph=None) 

            identified_estimand = model.identify_effect(proceed_when_unidentifiable=True)
            # Using linear regression. For non-linear effects, other estimators like DML would be better.
            estimate = model.estimate_effect(identified_estimand,
                                             method_name="backdoor.linear_regression",
                                             test_significance=False,
                                             method_params={'force_univariate': True}) # Force univariate if estimating one by one without common causes in graph
            
            if estimate is not None and hasattr(estimate, 'value') and not np.isnan(estimate.value):
                feature_effects[feature_as_treatment] = abs(estimate.value)
        except Exception as e:
            # print(f"Causal FS error for {feature_as_treatment}: {e}") # Verbose
            if "is not in the digraph" in str(e) or "is not in G" in str(e) or "More than one treatment variable" in str(e) :
                 print(f"Node/Graph error for {feature_as_treatment} in Causal FS: {e}")
            pass

    if not feature_effects:
        print("Causal feature ranking did not yield any valid effects after scaling and cleaning.")
    else:
        sorted_features = sorted(feature_effects.items(), key=lambda x: x[1], reverse=True)
        print(f"Causal features by absolute effect strength (top {min(5, len(sorted_features))} of {len(sorted_features)}): {sorted_features[:min(5, len(sorted_features))]}")
        return sorted_features[:max_causal_feats] # Return top N
    return []


def prepare_ml_data(df, target_col='target', test_split_size=0.15, min_test_samples=50):
    print("Preparing ML data...")
    if target_col not in df.columns:
        print(f"Target column '{target_col}' missing. Cannot prepare ML data."); return None, None, None, None, None


    cols_to_drop_base = ['open', 'high', 'low', 'close', 'volume', 'returns'] # Drop 'returns' if 'log_returns' is the main one
    cols_to_drop_dynamic = [c for c in df.columns if 'target_' in c and c != target_col] + \
                           [c for c in df.columns if 'future_return' in c] # General future return leakage
    
    all_cols_to_drop = list(set(cols_to_drop_base + cols_to_drop_dynamic))
    X = df.drop(columns=[c for c in all_cols_to_drop if c in df.columns] + [target_col], errors='ignore')
    y = df[target_col].copy()

    # Ensure target is suitable before proceeding
    if y.isnull().all():
        print("Target column is all NaN. Cannot prepare ML data."); return None, None, None, None, None
    valid_target_mask = y.notna()
    X = X.loc[valid_target_mask]
    y = y.loc[valid_target_mask]

    if X.empty or y.empty:
        print("X or y is empty after filtering for valid target. Cannot prepare ML data."); return None, None, None, None, None


    X.replace([np.inf, -np.inf], np.nan, inplace=True)

    if X.isnull().any().any():
        print(f"Warning: NaNs found in X before splitting. Count per column:\n{X.isnull().sum()[X.isnull().sum()>0]}")
        X = X.fillna(X.mean()) # Impute with mean if any NaNs slip through (or use ffill/bfill)
        if X.isnull().any().any(): # If mean imputation fails (e.g. all-NaN column)
             print("Still NaNs after mean imputation. Dropping rows with any NaNs in X.")
             X.dropna(axis=0, how='any', inplace=True)
             y = y.loc[X.index] # Align y with X after dropping rows

    if X.empty or y.empty:
        print("X or y is empty after NaN handling in prepare_ml_data."); return None, None, None, None, None

    print(f"Data shape after NaN handling in prepare_ml_data: X={X.shape}, y={y.shape}")
    if len(X) < min_test_samples * 2: # Ensure enough data for a meaningful train/test split
        print(f"Not enough data ({len(X)} rows) for a robust split. Min required ~{min_test_samples*2}."); return None, None, None, None, None


    n_samples = len(X)
    # Ensure test_size_abs is at least min_test_samples, and train_size is also reasonable
    test_size_abs = max(min_test_samples, int(n_samples * test_split_size))
    if n_samples - test_size_abs < min_test_samples : # if train size becomes too small
        test_size_abs = n_samples - min_test_samples # Adjust test to keep min train
        if test_size_abs < 1: # still not enough
            print(f"Cannot make a meaningful split with {n_samples} samples and min_test_samples {min_test_samples}"); return None,None,None,None,None

    train_size = n_samples - test_size_abs
    
    X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]
    y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]

    if X_train.empty or y_train.empty or X_test.empty or y_test.empty:
        print("Train or test set is empty after split."); return None, None, None, None, None
    print(f"Train shapes: X_train={X_train.shape}, y_train={y_train.shape}; Test shapes: X_test={X_test.shape}, y_test={y_test.shape}")

    numeric_cols_xtrain = X_train.select_dtypes(include=np.number).columns
    if not numeric_cols_xtrain.empty:
        scaler = StandardScaler()
        X_train_scaled_np = scaler.fit_transform(X_train[numeric_cols_xtrain])
        X_train_scaled = pd.DataFrame(X_train_scaled_np, columns=numeric_cols_xtrain, index=X_train.index)
        for col in X_train.columns.difference(numeric_cols_xtrain): X_train_scaled[col] = X_train[col] # Add back non-numeric

        # Scale X_test using the scaler fitted on X_train
        numeric_cols_xtest = X_test.select_dtypes(include=np.number).columns
        # Ensure only columns present in both train (for scaler fitting) and test are transformed
        cols_to_scale_in_test = [col for col in numeric_cols_xtrain if col in numeric_cols_xtest]
        if cols_to_scale_in_test:
            X_test_scaled_np = scaler.transform(X_test[cols_to_scale_in_test])
            X_test_scaled = pd.DataFrame(X_test_scaled_np, columns=cols_to_scale_in_test, index=X_test[cols_to_scale_in_test].index)
            for col in X_test.columns.difference(cols_to_scale_in_test): X_test_scaled[col] = X_test[col] # Add back others
        else:
            print("No common numeric columns between train and test for scaling, or test has no numeric. Test set not scaled by fitted scaler.")
            X_test_scaled = X_test.copy() # Return unscaled test if no common numeric cols
        
        return X_train_scaled, X_test_scaled, y_train, y_test, scaler
    else:
        print("No numeric columns in X_train for scaling. Returning unscaled data.");
        return X_train, X_test, y_train, y_test, None


def lgbm_objective(trial, X_train, y_train, X_val, y_val, base_params):
    """ Optuna objective function for LightGBM """
    params = {
        'objective': base_params['objective'],
        'metric': base_params['metric'],
        'num_class': base_params.get('num_class'), # Handles multiclass
        'boosting_type': 'gbdt',
        'n_estimators': trial.suggest_int('n_estimators', 100, 1000, step=50),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1, log=True),
        'num_leaves': trial.suggest_int('num_leaves', 10, 50),
        'max_depth': trial.suggest_int('max_depth', -1, 10), # -1 for no limit
        'min_child_samples': trial.suggest_int('min_child_samples', 5, 50),
        'feature_fraction': trial.suggest_float('feature_fraction', 0.5, 1.0), # like colsample_bytree
        'bagging_fraction': trial.suggest_float('bagging_fraction', 0.5, 1.0), # like subsample
        'bagging_freq': trial.suggest_int('bagging_freq', 0, 7), # 0 means disable bagging
        'reg_alpha': trial.suggest_float('reg_alpha', 1e-3, 1.0, log=True), # L1
        'reg_lambda': trial.suggest_float('reg_lambda', 1e-3, 1.0, log=True), # L2
        'random_state': 42,
        'verbose': -1,
        'n_jobs': -1,
        'class_weight': base_params.get('class_weight', 'balanced')
    }
    # Remove num_class if it's None (for binary classification)
    if params['num_class'] is None: del params['num_class']


    model = lgb.LGBMClassifier(**params)
    model.fit(X_train, y_train,
              eval_set=[(X_val, y_val)],
              eval_metric=base_params['metric'], # Use the primary metric
              callbacks=[lgb.early_stopping(30, verbose=False)])
    

    if base_params['objective'] == 'binary':
        y_proba_val = model.predict_proba(X_val)[:, 1]
        # return roc_auc_score(y_val, y_proba_val) # Maximize AUC
        return log_loss(y_val, y_proba_val) # Minimize log_loss
    # For multiclass:
    else:
        y_proba_val = model.predict_proba(X_val)
        return log_loss(y_val, y_proba_val) # Minimize log_loss


def optimize_lgbm_hyperparameters(X_train, y_train, base_params, n_trials=50, validation_ratio=0.2):
    if not optuna_available:
        print("Optuna not available. Skipping hyperparameter optimization.")
        return optimized_lightgbm_params() # Return default fixed params

    print(f"Optimizing LightGBM hyperparameters with Optuna ({n_trials} trials)...")
    

    train_idx, val_idx = next(TimeSeriesSplit(n_splits=int(1/validation_ratio), 
                                               test_size=int(len(X_train)*validation_ratio)).split(X_train))
    
    X_train_opt, X_val_opt = X_train.iloc[train_idx], X_train.iloc[val_idx]
    y_train_opt, y_val_opt = y_train.iloc[train_idx], y_train.iloc[val_idx]

    study = optuna.create_study(direction='minimize') # Minimize log_loss (or maximize AUC)
    study.optimize(lambda trial: lgbm_objective(trial, X_train_opt, y_train_opt, X_val_opt, y_val_opt, base_params),
                   n_trials=n_trials,
                   callbacks=[lambda study, trial: print(f"Trial {trial.number} finished with value: {trial.value:.4f} and params: {trial.params}") if trial.number % 10 == 0 else None])

    print(f"Best trial for LightGBM: Value={study.best_value}, Params={study.best_params}")
    
    # Combine best Optuna params with fixed ones not tuned
    final_params = {**base_params, **study.best_params}
    return final_params


def train_lightgbm_model(X_train, y_train, X_test, y_test, optimized_params=None): # Accept optimized_params
    print("Training LightGBM model...")
    if X_train.empty or y_train.empty:
        print("X_train or y_train is empty. Skipping LightGBM training."); return None, None

    y_train_squeezed = y_train.squeeze()
    y_test_squeezed = y_test.squeeze()
    unique_labels_train = sorted(y_train_squeezed.unique())
    num_classes = len(unique_labels_train)

    if num_classes <= 1:
        print(f"Only {num_classes} class(es) in y_train. Skipping LightGBM training."); return None, None

    # Define base parameters (objective, metric, etc. that Optuna might not tune or needs as input)
    base_params = {
        'objective': 'multiclass' if num_classes > 2 else 'binary',
        'metric': 'multi_logloss' if num_classes > 2 else 'binary_logloss', # Can also use 'auc' for binary
        'random_state': 42,
        'verbose': -1,
        'n_jobs': -1,
        'class_weight': 'balanced' # Good default for potentially imbalanced classes
    }
    if base_params['objective'] == 'multiclass':
        base_params['num_class'] = num_classes

    if optimized_params is None: # If no pre-optimized params, run Optuna
        print("No optimized parameters provided, attempting to run Optuna...")
        params_to_use = optimize_lgbm_hyperparameters(X_train, y_train_squeezed, base_params.copy(), n_trials=25) # Reduced trials for demo
    else: # Use provided optimized params
        print("Using provided_optimized parameters.")
        # Ensure all necessary base params are there if optimized_params doesn't include them
        params_to_use = {**base_params, **optimized_params}


    # Map labels to 0-indexed integers for LightGBM
    label_map = {label: i for i, label in enumerate(unique_labels_train)}
    y_train_mapped = y_train_squeezed.map(label_map)
    y_test_mapped = y_test_squeezed.map(label_map).fillna(-1).astype(int) # Handle labels in test not in train map

    # Compute sample weights for training (optional, if class_weight='balanced' is not enough)
    sample_weights = None
    # ... (sample_weights computation logic can be added here if needed) ...

    model = lgb.LGBMClassifier(**params_to_use)
    
    # Prepare eval_set ensuring mapped test labels
    valid_eval_indices = y_test_mapped != -1
    eval_set_data = (X_test.loc[valid_eval_indices], y_test_mapped[valid_eval_indices]) if valid_eval_indices.any() else None

    if eval_set_data:
        model.fit(X_train, y_train_mapped, sample_weight=sample_weights,
                  eval_set=[eval_set_data],
                  eval_metric=params_to_use.get('metric', 'logloss'), # Ensure metric is passed
                  callbacks=[lgb.early_stopping(30, verbose=False)])
    else: # Fit without eval_set if no valid test samples for mapping
        print("Warning: No valid mapped test samples for early stopping. Fitting on full training data.")
        model.fit(X_train, y_train_mapped, sample_weight=sample_weights)

    y_pred_mapped = model.predict(X_test)
    y_proba = model.predict_proba(X_test)

    if not valid_eval_indices.any():
        print("No valid test samples for evaluation after mapping."); return model, None # Return model, but no metrics

    acc = accuracy_score(y_test_mapped[valid_eval_indices], y_pred_mapped[valid_eval_indices])
    print(f"\nðŸŽ¯ Accuracy on mapped test data: {acc:.4f}")

    if params_to_use['objective'] == 'binary' and y_proba.shape[1] == 2:
        try:
            auc = roc_auc_score(y_test_mapped[valid_eval_indices], y_proba[valid_eval_indices][:, 1])
            print(f"ðŸ“Š AUC on mapped test data: {auc:.4f}")
        except ValueError as e_auc:
            print(f"AUC Calculation Error: {e_auc}") # e.g. if only one class in y_test_mapped[valid_eval_indices]
    
    print("\nClassification Report (on mapped and valid test labels):")
    try:
        print(classification_report(y_test_mapped[valid_eval_indices], y_pred_mapped[valid_eval_indices],
                                    labels=sorted(y_train_mapped.unique()), # ensure correct labels
                                    zero_division=0))
    except Exception as e_cr:
        print(f"Classification Report Error: {e_cr}")

    feat_imp_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': model.feature_importances_}).sort_values(by='Importance', ascending=False)
    print("\nTop 10 features:\n", feat_imp_df.head(10))
    return model, feat_imp_df


def plot_feature_importance(feature_importance_df, top_n=20, symbol_for_plot=""):
    if feature_importance_df is None or feature_importance_df.empty:
        print("No feature importance data to plot.")
        return
    plt.figure(figsize=(12, max(6, min(top_n, len(feature_importance_df)) * 0.45)))
    # Use a consistent palette like "viridis" or "mako_r"
    sns.barplot(x='Importance', y='Feature', data=feature_importance_df.head(top_n), palette="viridis")
    plt.title(f'Top {top_n} Feature Importances for {symbol_for_plot} (LightGBM)', fontsize=16)
    plt.tight_layout()
    # Consider saving plot instead of showing if running non-interactively
    # plt.savefig(f"{symbol_for_plot}_feature_importance.png")
    plt.show()


def export_lgbm_to_onnx(lgbm_model, X_sample_df, file_path="lgbm_model.onnx", target_opset=12):
    print(f"\nExporting LightGBM model to ONNX: {file_path} (target opset={target_opset})")
    if not all([onnx_available, skl2onnx_available, onnxmltools_available, FloatTensorType is not None]):
        print("One or more ONNX libraries missing. Skipping ONNX export."); return None
    if lgbm_model is None or X_sample_df is None or X_sample_df.empty:
        print("LGBM Model or sample data for ONNX export is empty. Skipping."); return None

    try:
        # Ensure X_sample_df has features in the same order as training
        initial_type = [('float_input', FloatTensorType([None, X_sample_df.shape[1]]))]
        # Use onnxmltools for LightGBM conversion
        converted_model = onnxmltools.convert_lightgbm(lgbm_model, initial_types=initial_type, target_opset=target_opset)
        
        with open(file_path, "wb") as f:
            f.write(converted_model.SerializeToString())
        print(f"Model successfully exported to ONNX: {file_path}")
        
        # Verify the ONNX model
        onnx.checker.check_model(file_path)
        print("ONNX model check successful.")
        return file_path
    except Exception as e:
        print(f"Error exporting LightGBM to ONNX: {e}\n{traceback.format_exc()}")
        print("Attempting to save model as pickle instead.")
        try:
            import pickle
            pkl_path = file_path.replace('.onnx', '.pkl')
            with open(pkl_path, 'wb') as pf:
                pickle.dump(lgbm_model, pf)
            print(f"Model saved as pickle: {pkl_path}")
            return pkl_path
        except Exception as ep:
            print(f"Failed to save model as pickle: {ep}"); return None


def simple_feature_selection_fallback(X_train, y_train, max_features=20): # y_train not used but kept for API consistency
    print("Using simple variance-based feature selection fallback...")
    if X_train.empty: return pd.DataFrame() # Return empty DataFrame if no features

    # Ensure X_train contains only numeric data for var()
    X_train_numeric = X_train.select_dtypes(include=np.number)
    if X_train_numeric.empty:
        print("No numeric features in X_train for variance selection.")
        # Fallback to just taking all columns if no numeric, or first few
        selected_features = X_train.columns[:max_features].tolist()
        return pd.DataFrame({'Feature': selected_features})


    variance_scores = X_train_numeric.var().sort_values(ascending=False)
    num_features_to_select = min(max_features, len(variance_scores))
    selected_features = variance_scores.head(num_features_to_select).index.tolist()
    
    # Add back any non-numeric columns if they were part of original X_train and prioritized selection logic expects them
    # This part might need refinement based on how non-numeric features are handled overall.
    # For now, this fallback focuses on numeric variance.
    
    return pd.DataFrame({'Feature': selected_features, 'Score': variance_scores.head(num_features_to_select).values})


def prioritized_feature_selection(X_train, y_train, causal_ranking, max_features=20):
    print("Prioritized feature selection using Causal Ranking and Mutual Information...")
    if X_train.empty or y_train.empty: return pd.DataFrame()

    top_causal_features = []
    if causal_ranking: # causal_ranking is a list of (feature, score) tuples
        # Take top N/2 from causal, ensure they are in X_train columns
        num_causal_to_select = min(len(causal_ranking), max_features // 2) 
        for feat, score in causal_ranking[:num_causal_to_select]:
            if feat in X_train.columns:
                top_causal_features.append(feat)
    
    remaining_slots = max_features - len(top_causal_features)
    stat_selected_features = []

    if remaining_slots > 0:
        # Features not already selected by causal ranking
        features_for_stat_selection = [f for f in X_train.columns if f not in top_causal_features]
        
        if features_for_stat_selection:
            X_remaining_for_stat = X_train[features_for_stat_selection]
            y_train_squeezed = y_train.squeeze()

            if y_train_squeezed.nunique() > 1 and not X_remaining_for_stat.empty:
                try:
                    # Ensure X_remaining_for_stat only has numeric types for mutual_info_classif
                    X_remaining_numeric = X_remaining_for_stat.select_dtypes(include=np.number)
                    if not X_remaining_numeric.empty:
                        num_stat_to_select = min(remaining_slots, X_remaining_numeric.shape[1])
                        selector_mi = SelectKBest(mutual_info_classif, k=num_stat_to_select)
                        selector_mi.fit(X_remaining_numeric, y_train_squeezed)
                        stat_selected_features = X_remaining_numeric.columns[selector_mi.get_support()].tolist()
                    else:
                        print("No numeric features left for statistical selection (mutual info).")
                except Exception as e_mi:
                    print(f"Error in mutual_info_classif for prioritized selection: {e_mi}. No stat features added by MI.")
            else:
                print("Not enough target variance or no remaining features for MI-based stat selection.")
        else:
            print("No remaining features for statistical selection after causal.")

    final_selected_features = list(dict.fromkeys(top_causal_features + stat_selected_features)) # Combine and keep unique, preserving order

    if not final_selected_features and not X_train.empty:
        print("No features selected by prioritized methods, falling back to simple variance selection.")
        return simple_feature_selection_fallback(X_train, y_train, max_features)
    
    print(f"Prioritized features selected: {len(final_selected_features)}. Top 5: {final_selected_features[:5]}")
    return pd.DataFrame({'Feature': final_selected_features})


def optimized_lightgbm_params(): # This provides a base set if Optuna is skipped.
    return {
        'boosting_type': 'gbdt', 'num_leaves': 31, 'learning_rate': 0.05,
        'n_estimators': 200, 'feature_fraction': 0.8, 'bagging_fraction': 0.8,
        'bagging_freq': 5, 'min_child_samples': 20, 'reg_alpha': 0.01,
        'reg_lambda': 0.01, 'random_state': 42, 'verbose': -1, 'n_jobs': -1,
        'class_weight': 'balanced'
        # Objective, metric, num_class are set dynamically in train_lightgbm_model
    }


def ultimate_forecasting_workflow(symbol, df_raw, autoformer_model_instance, prediction_length=5):
    print(f"\n--- ðŸš€ Ultimate Forecasting (Using AutoformerPredictor): {symbol} ---")

    if autoformer_model_instance is None or not hasattr(autoformer_model_instance, 'predict'):
        print(f"Autoformer model instance not provided or invalid for {symbol}. Cannot forecast.")
        return {'status': 'failed', 'reason': 'Invalid Autoformer model instance'}

    close_prices = df_raw['close'].copy()
    if len(close_prices) < autoformer_model_instance.input_len: # Check against model's actual input_len
        print(f"Not enough historical data ({len(close_prices)} points) for Autoformer input_len ({autoformer_model_instance.input_len}).")
        # Fallback: Return NaNs or a simple naive forecast
        last_price_fallback = df_raw['close'].iloc[-1] if not df_raw.empty else np.nan
        return {
            'forecast': np.array([np.nan]*prediction_length),
            'method': "AutoformerPredictor (Insufficient Data)",
            'last_price': last_price_fallback,
            'status': "Insufficient data for Autoformer",
            'avg_forecast': np.nan, 'direction': "N/A", 'magnitude': np.nan,
            'confidence_metric_value': np.nan, 'confidence_label': "N/A",
            'advanced_analysis': False
        }

    forecast_result = autoformer_model_instance.predict(close_prices)

    if forecast_result is not None and isinstance(forecast_result, np.ndarray) and not np.isnan(forecast_result).all():
        last_price = df_raw['close'].iloc[-1]
        avg_prediction = np.nanmean(forecast_result)

        if np.isnan(avg_prediction):
            direction = "â†”ï¸ UNCERTAIN (NaN prediction)"
            magnitude = float('nan')
        elif avg_prediction > last_price: direction = "ðŸ“ˆ UP"
        elif avg_prediction < last_price: direction = "ðŸ“‰ DOWN"
        else: direction = "âž¡ï¸ FLAT"
        
        magnitude = abs((avg_prediction - last_price) / last_price * 100) if last_price != 0 and not np.isnan(avg_prediction) else (float('inf') if not np.isnan(avg_prediction) else float('nan'))
        forecast_std = np.nanstd(forecast_result)

        if np.isnan(forecast_std) or np.isnan(last_price): confidence_label = "âšª Unknown (NaN values)"
        elif forecast_std == 0: confidence_label = "ðŸŸ¡ Medium (Flat Forecast)" # Model predicts a flat line
        elif abs(last_price) > 1e-6 and forecast_std / abs(last_price) < 0.01 : confidence_label = "ðŸŸ¢ High (Low Relative Std)" # Low relative std
        elif abs(last_price) > 1e-6 and forecast_std / abs(last_price) < 0.03 : confidence_label = "ðŸŸ¡ Medium (Moderate Relative Std)"
        else: confidence_label = "ðŸ”´ Low (High Relative Std or price near zero)"

        print(f"âœ… AutoformerPredictor Forecast: {forecast_result}")
        print(f"ðŸ“Š Analysis: {direction} {magnitude:.1f}% - Confidence: {confidence_label} (Forecast_Std: {forecast_std:.4f})")
        
        return {
            'forecast': forecast_result, 'method': "Custom AutoformerPredictor",
            'last_price': last_price, 'avg_forecast': avg_prediction,
            'direction': direction, 'magnitude': magnitude,
            'confidence_metric_value': forecast_std, 'confidence_label': confidence_label,
            'advanced_analysis': True, 'status': "Completed"
        }
    else:
        print(f"Forecasting with AutoformerPredictor failed for {symbol} or returned all NaNs.")
        last_price_fallback = df_raw['close'].iloc[-1] if not df_raw.empty else np.nan
        return {
            'status': 'failed', 'reason': 'AutoformerPredictor failed or returned all NaNs',
            'forecast': np.array([np.nan]*prediction_length) if prediction_length else np.array([]),
            'method': "Custom AutoformerPredictor (Failed)",
            'last_price': last_price_fallback
        }



def run_patchtst_foundation_forecast(
    symbol_name: str,
    historical_data_df: pd.DataFrame,
    prediction_length: int,
    context_length: int = 96, # Default context, often 96 for ETTh1
    model_checkpoint: str = "ibm-research/patchtst-etth1-pretrain", # <<< CORRECTED MODEL ID
    patch_length: int = 16 # Default patch length, often 16 for ETTh1
):
    print(f"\n---  Forecasting for {symbol_name} using PatchTST Foundation Model ({model_checkpoint}) ---")

    if not torch_available or 'PatchTSTForPrediction' not in globals():
        print("Torch or PatchTST from transformers not available. Skipping foundation model forecast.")
        return {
            "status": "failed",
            "reason": "Torch or Transformers components missing",
            "method": "PatchTST (Skipped)",
            "forecast": np.array([np.nan] * prediction_length)
        }

    # Ensure 'close' column exists
    if 'close' not in historical_data_df.columns:
        print(f"Critical 'close' column missing in historical_data_df for {symbol_name}. Skipping forecast.")
        return {
            "status": "failed",
            "reason": "'close' column missing",
            "method": f"PatchTST ({model_checkpoint}) (Skipped)",
            "forecast": np.array([np.nan] * prediction_length)
        }
    close_prices = historical_data_df['close'].values.astype(np.float32)

    if len(close_prices) < context_length:
        print(f"Not enough historical data ({len(close_prices)} points) for PatchTST context_length ({context_length}). Skipping.")
        return {
            "status": "failed",
            "reason": "Insufficient data for PatchTST context",
            "method": f"PatchTST ({model_checkpoint}) (Skipped)",
            "forecast": np.array([np.nan] * prediction_length),
            "last_price": close_prices[-1] if len(close_prices) > 0 else np.nan
        }

    try:
        # Load config, model, and processor using the corrected model_checkpoint
        # The from_pretrained method will fetch the correct configuration for this model
        config = PatchTSTConfig.from_pretrained(
            model_checkpoint,
            prediction_length=prediction_length, # This will be overridden in the model's generation config if needed
            context_length=context_length,     # This should align with model's pre-training or be a valid input size
            patch_length=patch_length          # This should align with model's pre-training
        )
        # It's often better to let the model load its default config for prediction_length during pretraining
        # and then use generate's prediction_length. But for consistency in some models, explicitly setting is okay.
        
        model = PatchTSTForPrediction.from_pretrained(model_checkpoint, config=config)
        processor = AutoProcessor.from_pretrained(model_checkpoint)

        device = "cuda" if torch.cuda.is_available() else "cpu"
        model.to(device)

        # Prepare input for the model
        # Use the last `context_length` points from the historical data
        past_values = close_prices[-context_length:]
        
        inputs = processor(
            past_values, # Should be a list or 1D numpy array of past values
            return_tensors="pt"

        )
        inputs = {k: v.to(device) for k, v in inputs.items()}

        with torch.no_grad():

            generated_outputs = model.generate(
                past_values=inputs["past_values"],

                prediction_length = prediction_length # Specify how many steps to predict
            )
        
        # The output structure can vary slightly by model; consult docs or inspect `generated_outputs`
        # For PatchTSTForPrediction, it's typically in `prediction_outputs`
        forecast_values = generated_outputs.prediction_outputs.cpu().numpy().squeeze()
        
        # Ensure forecast_values has the desired prediction_length
        if forecast_values.ndim == 0: 
            forecast_values = np.array([forecast_values.item()] * prediction_length)
        elif forecast_values.shape[0] < prediction_length:
             print(f"Warning: PatchTST generated {forecast_values.shape[0]} steps, expected {prediction_length}. Padding...")
             forecast_values = np.pad(forecast_values, (0, prediction_length - forecast_values.shape[0]), 'edge')
        elif forecast_values.shape[0] > prediction_length:
            forecast_values = forecast_values[:prediction_length]


        last_actual_price = close_prices[-1]
        avg_prediction = np.nanmean(forecast_values)

        direction = "N/A"
        magnitude = float('nan')
        confidence_label = "N/A"
        forecast_std = np.nanstd(forecast_values)

        if not np.isnan(avg_prediction) and not np.isnan(last_actual_price):
            if avg_prediction > last_actual_price: direction = "ðŸ“ˆ UP"
            elif avg_prediction < last_actual_price: direction = "ðŸ“‰ DOWN"
            else: direction = "âž¡ï¸ FLAT"
            if last_actual_price != 0:
                magnitude = abs((avg_prediction - last_actual_price) / last_actual_price * 100)
            else:
                magnitude = float('inf') if avg_prediction != 0 else 0.0
        
        if np.isnan(forecast_std) or np.isnan(last_actual_price):
            confidence_label = "âšª Unknown (NaN values)"
        elif forecast_std == 0 and not np.isnan(last_actual_price): # Check last_actual_price too
            confidence_label = "ðŸŸ¡ Medium (Flat Forecast)"
        elif abs(last_actual_price) > 1e-9 and (forecast_std / abs(last_actual_price)) < 0.01 : # Use a small epsilon for last_actual_price
            confidence_label = "ðŸŸ¢ High (Low Relative Std)"
        elif abs(last_actual_price) > 1e-9 and (forecast_std / abs(last_actual_price)) < 0.05 :
            confidence_label = "ðŸŸ¡ Medium (Moderate Relative Std)"
        else:
            confidence_label = "ðŸ”´ Low (High Relative Std or price near zero)"


        print(f"âœ… PatchTST Forecast: {forecast_values}") # Print the final shaped forecast
        print(f"ðŸ“Š Analysis: {direction} {magnitude:.1f}% - Confidence: {confidence_label} (Forecast_Std: {forecast_std:.4f})")

        return {
            "status": "Completed",
            "forecast": forecast_values,
            "method": f"PatchTST ({model_checkpoint})",
            "last_price": last_actual_price,
            "avg_forecast": avg_prediction,
            "direction": direction,
            "magnitude": magnitude,
            "confidence_metric_value": forecast_std,
            "confidence_label": confidence_label,
            "advanced_analysis": True
        }

    except Exception as e:
        print(f"Error during PatchTST foundation model forecasting for {symbol_name}: {e}")
        error_trace = traceback.format_exc()
        print(error_trace) # Print the full traceback for debugging
        return {
            "status": "failed",
            "reason": str(e),
            "trace": error_trace, # Include traceback in results for easier debugging
            "method": f"PatchTST ({model_checkpoint}) (Error)",
            "forecast": np.array([np.nan] * prediction_length),
            "last_price": historical_data_df['close'].iloc[-1] if not historical_data_df.empty and 'close' in historical_data_df else np.nan
        }

# --- MAIN WORKFLOW FUNCTION ---
def run_full_workflow(symbol=DEFAULT_SYMBOL, start_date_str=START_DATE, end_date_str=END_DATE, api_key_val=API_KEY,
                      force_train_autoformer=False, run_optuna_lgbm=False): # Added control flags
    print(f"\n{'='*40}\nðŸš€ ENHANCED WORKFLOW FOR: {symbol}\n{'='*40}")
    
    default_return = {
        "symbol": symbol, "status": "Workflow Started", "raw_data_shape": (0,0),
        "featured_data_shape": (0,0), "X_train_shape": (0,0), "X_test_shape": (0,0),
        "selected_features_count": 0, "selected_feature_names": [], "scaler_object": None,
        "ml_model_object": None, "lgbm_feature_importance": None, "causal_model_object": None,
        "causal_feature_ranking": [], "onnx_model_path": None, "forecasting_results": None,
        "lgbm_optimized_params": None
    }

    df_raw = fetch_twelve_data(symbol, api_key_val, start_date_str=start_date_str, end_date_str=end_date_str)
    if df_raw is None or df_raw.empty:
        default_return["status"] = "Data Fetching Failed"
        return default_return
    default_return["raw_data_shape"] = df_raw.shape

    price_c, high_c, low_c, vol_c = 'close', 'high', 'low', 'volume'
    target_col = 'target' # Define target column name

    print(f"\n--- ðŸ”§ Feature Engineering: {symbol} ---")
    df_f = df_raw.copy()
    df_f = add_technical_indicators(df_f)
    df_f = add_optimized_features(df_f, price_col=price_c, volume_col=vol_c)
    df_f = add_wavelet_features(df_f, column=price_c)
    df_f = add_entropy_features(df_f, column=price_c, window=40)
    df_f = add_advanced_technical_features(df_f, price_col=price_c, high_col=high_c, low_col=low_c, volume_col=vol_c)
    df_f = add_transformer_features_conceptual(df_f, column=price_c, sequence_length=20)
    
    # Interaction features
    if 'RSI_14' in df_f.columns and 'ADX_14' in df_f.columns:
        df_f['RSI_ADX_interaction'] = df_f['RSI_14'] * df_f['ADX_14'] / 100.0 # Ensure float division
    if 'ATR_14' in df_f.columns and 'volatility_20' in df_f.columns: # Assuming volatility_20 is from log_returns
        volatility_safe = df_f['volatility_20'].replace(0, np.nan)
        df_f['ATR_vol_ratio'] = df_f['ATR_14'] / volatility_safe
    default_return["featured_data_shape"] = df_f.shape

    print(f"\n--- Data Cleaning (Inf/NaN Handling & Imputation): {symbol} ---")
    df_f.replace([np.inf, -np.inf], np.nan, inplace=True)
    numeric_cols_to_impute = df_f.select_dtypes(include=np.number).columns
    if not numeric_cols_to_impute.empty:
        nan_counts_before = df_f[numeric_cols_to_impute].isnull().sum().sum()
        if nan_counts_before > 0:
            print(f"NaNs before imputation in numeric columns: {nan_counts_before}")
            # df_f[numeric_cols_to_impute] = df_f[numeric_cols_to_impute].fillna(method='ffill').fillna(method='bfill').fillna(0)
            # More robust: impute with mean/median or interpolate
            df_f[numeric_cols_to_impute] = df_f[numeric_cols_to_impute].interpolate(method='linear', limit_direction='both', axis=0)
            df_f[numeric_cols_to_impute] = df_f[numeric_cols_to_impute].fillna(method='bfill').fillna(method='ffill').fillna(0) # Fallback for edges
            print(f"NaNs after imputation: {df_f[numeric_cols_to_impute].isnull().sum().sum()}")
        else: print("No NaNs found in numeric features to impute.")
    else: print("No numeric columns found for imputation.")


    print(f"\n--- ðŸ“Š Regime Detection: {symbol} ---")
    # Option 1: Simple Volatility based (current default)
    df_f = detect_regimes_simple(df_f, column=price_c)
    # Option 2: ClaSP based (more advanced, can be enabled)
    # df_f = add_clasp_regimes(df_f, column=price_c, period=30, n_cps=3) # Example ClaSP call

    print(f"\n--- ðŸŽ¯ Target Definition: {symbol} ---")
    df_f = balanced_target_definition(df_f, column=price_c, periods=5) # Using 5 periods ahead

    print(f"\n--- Causal Discovery & Ranking: {symbol} ---")
    causal_model_obj = None
    causal_feat_ranking = []
    if df_f is not None and not df_f.empty and target_col in df_f.columns and df_f[target_col].nunique(dropna=True) > 1:
        # Pass df_f which has all features, scaling is done inside causal functions
        causal_model_obj = discover_causal_structure(df_f.copy(), target_col=target_col, symbol=symbol)
        if causal_model_obj: # Check if model object was created
             # Pass df_f for feature selection, it will be cleaned and scaled inside
            causal_feat_ranking = causal_feature_selection(df_f.copy(), target_col=target_col)
    else:
        print("Skipping Causal Discovery & Ranking due to invalid target or data.")
    default_return["causal_model_object"] = causal_model_obj
    default_return["causal_feature_ranking"] = causal_feat_ranking if causal_feat_ranking else []


    print(f"\n--- ML Preparation & Feature Selection: {symbol} ---")
    X_tr, X_te, y_tr, y_te, scaler_obj = None, None, None, None, None
    sel_feat_names = []
    current_status_ml = "ML Prep Incomplete"

    if df_f is None or df_f.empty or target_col not in df_f.columns or df_f[target_col].isnull().all():
        current_status_ml = "ML Prep Failed - DataFrame empty or target all NaN"
    elif df_f[target_col].nunique(dropna=True) <= 1: # Crucial check for classification
        current_status_ml = f"ML Prep Skipped - Target has {df_f[target_col].nunique(dropna=True)} unique non-NaN values"
    else:
        X_tr_p, X_te_p, y_tr_p, y_te_p, scaler_obj_p = prepare_ml_data(df_f.copy(), target_col=target_col)
        default_return["scaler_object"] = scaler_obj_p # Store scaler from prepare_ml_data

        if X_tr_p is not None and not X_tr_p.empty and y_tr_p is not None and not y_tr_p.empty and y_tr_p.nunique() > 1:
            scores_df = prioritized_feature_selection(X_tr_p.copy(), y_tr_p.copy(), causal_feat_ranking, max_features=25) # Increased max_features
            
            sel_feat_names_intermediate = scores_df['Feature'].tolist() if scores_df is not None and not scores_df.empty else X_tr_p.columns.tolist()
            if not sel_feat_names_intermediate: sel_feat_names_intermediate = X_tr_p.columns.tolist()[:25] # Fallback to first 25 if all else fails
            
            X_tr_sel = X_tr_p[[col for col in sel_feat_names_intermediate if col in X_tr_p.columns]]
            X_te_sel = X_te_p[[col for col in sel_feat_names_intermediate if col in X_te_p.columns]]

            if not X_tr_sel.empty:
                numeric_cols_xtr_sel = X_tr_sel.select_dtypes(include=np.number).columns
                if not numeric_cols_xtr_sel.empty:
                    variance_selector = VarianceThreshold(threshold=0.001) # Lower threshold slightly
                    variance_selector.fit(X_tr_sel[numeric_cols_xtr_sel])
                    sel_feat_names_vt = X_tr_sel[numeric_cols_xtr_sel].columns[variance_selector.get_support()].tolist()
                    
                    if not sel_feat_names_vt: # If VT removes all
                        print("âš ï¸ All numeric features removed by VarianceThreshold. Using features before VT.")
                        sel_feat_names = sel_feat_names_intermediate
                    else:
                        sel_feat_names = sel_feat_names_vt
                    
                    # Keep non-numeric selected features if any (though unlikely after typical TIs)
                    non_numeric_sel_intermed = [f for f in sel_feat_names_intermediate if f not in numeric_cols_xtr_sel and f in X_tr_p.columns]
                    sel_feat_names = list(dict.fromkeys(sel_feat_names + non_numeric_sel_intermed)) # Combine and unique

                else: # No numeric features after initial selection
                    print("No numeric features in X_tr_sel for VarianceThreshold. Using features before VT.")
                    sel_feat_names = X_tr_sel.columns.tolist()

                X_tr = X_tr_p[sel_feat_names].copy() # Use original X_tr_p and selected names
                X_te = X_te_p[sel_feat_names].copy() # Use original X_te_p and selected names
                y_tr, y_te = y_tr_p, y_te_p # y remains aligned with X_tr_p, X_te_p
                
                # Re-scale AFTER final feature selection
                if not X_tr.empty and scaler_obj_p is not None: # scaler_obj_p was fit on X_tr_p
                    final_numeric_cols_tr = X_tr.select_dtypes(include=np.number).columns
                    if not final_numeric_cols_tr.empty:
                        # Important: Create a NEW scaler and fit on the FINAL X_train's numeric columns
                        final_scaler = StandardScaler()
                        X_tr[final_numeric_cols_tr] = final_scaler.fit_transform(X_tr[final_numeric_cols_tr])
                        default_return["scaler_object"] = final_scaler # Store the final scaler
                        
                        if not X_te.empty:
                            final_numeric_cols_te = [col for col in final_numeric_cols_tr if col in X_te.columns]
                            if final_numeric_cols_te:
                                X_te[final_numeric_cols_te] = final_scaler.transform(X_te[final_numeric_cols_te])
                    current_status_ml = "ML Prep Completed"
                elif X_tr.empty:
                    current_status_ml = "ML Prep Failed - X_tr empty after feature selection"
                else: # Scaler was None or X_tr empty
                    current_status_ml = "ML Prep Completed (Data not re-scaled after final selection)"


            else: current_status_ml = "ML Prep Failed - X_tr_sel empty after initial feature selection"
        else: current_status_ml = "ML Prep Failed - Initial data from prepare_ml_data insufficient or target has only one class"
    
    default_return["status"] = current_status_ml
    default_return["X_train_shape"] = X_tr.shape if X_tr is not None else (0,0)
    default_return["X_test_shape"] = X_te.shape if X_te is not None else (0,0)
    default_return["selected_feature_names"] = sel_feat_names
    default_return["selected_features_count"] = len(sel_feat_names)
    # Scaler object is already stored from prepare_ml_data or the final re-scaling step

    # --- LightGBM Model Training ---
    lgbm_model_obj, lgbm_feat_imp, onnx_file_path = None, None, None
    lgbm_final_params = None

    if current_status_ml == "ML Prep Completed" and X_tr is not None and not X_tr.empty and y_tr is not None and not y_tr.empty and y_tr.nunique() > 1:
        print(f"\n--- Model Training & Eval: {symbol} ---")
        try:
            params_for_lgbm = None
            if run_optuna_lgbm: # If flag is true, run optimization
                 # Base params for Optuna to know objective/metric
                unique_labels_tr = sorted(y_tr.squeeze().unique())
                num_classes_tr = len(unique_labels_tr)
                optuna_base_params = {
                    'objective': 'multiclass' if num_classes_tr > 2 else 'binary',
                    'metric': 'multi_logloss' if num_classes_tr > 2 else 'binary_logloss',
                    'num_class': num_classes_tr if num_classes_tr > 2 else None,
                    'class_weight': 'balanced'
                }
                lgbm_final_params = optimize_lgbm_hyperparameters(X_tr, y_tr.squeeze(), optuna_base_params, n_trials=30) # 30 trials for demo
            else: # Use default fixed params
                lgbm_final_params = optimized_lightgbm_params() # Fetches predefined defaults

            default_return["lgbm_optimized_params"] = lgbm_final_params
            lgbm_model_obj, lgbm_feat_imp = train_lightgbm_model(X_tr, y_tr, X_te, y_te, optimized_params=lgbm_final_params) # Pass the params

            if lgbm_model_obj:
                default_return["status"] = "Completed" # Overall status
                default_return["ml_model_object"] = lgbm_model_obj
                default_return["lgbm_feature_importance"] = lgbm_feat_imp
                if lgbm_feat_imp is not None and not lgbm_feat_imp.empty:
                    plot_feature_importance(lgbm_feat_imp, top_n=20, symbol_for_plot=symbol)
                
                if X_te is not None and not X_te.empty and not X_te.iloc[[0]].empty:
                     # Ensure sample for ONNX has correct columns if X_te was modified by scaling
                    onnx_sample_data = X_te.iloc[[0]][sel_feat_names] if sel_feat_names and all(f in X_te.columns for f in sel_feat_names) else X_te.iloc[[0]]
                    onnx_file_path = export_lgbm_to_onnx(lgbm_model_obj, onnx_sample_data, file_path=f"{symbol}_lgbm_model.onnx")
                    default_return["onnx_model_path"] = onnx_file_path
            else:
                default_return["status"] = "Completed (LGBM Training Failed or Skipped)"
        except Exception as e_train:
            print(f"LightGBM Model training/evaluation error: {e_train}\n{traceback.format_exc()}")
            default_return["status"] = "Completed (LGBM Training Error)"
    else:
        print(f"âš ï¸ Skipping LightGBM Training for {symbol}: {current_status_ml} or invalid/empty training data.")
        default_return["status"] = current_status_ml if "Failed" in current_status_ml or "Skipped" in current_status_ml else "ML Training Skipped"

    # --- Autoformer Forecasting Model ---
    print(f"\n--- Autoformer Model Initialization & Forecasting: {symbol} ---")
    # Train (if needed) or load the Autoformer model
    # Pass only the 'close' price series for Autoformer training/prediction
    autoformer_predictor_instance = train_or_load_autoformer(df_raw['close'], force_train=force_train_autoformer)

    if autoformer_predictor_instance:
        forecast_output = ultimate_forecasting_workflow(symbol, df_raw.copy(), autoformer_predictor_instance, prediction_length=5)
        default_return["forecasting_results"] = forecast_output
    else:
        print("Autoformer predictor instance could not be initialized/trained/loaded. Skipping forecasting.")
        default_return["forecasting_results"] = {"status": "failed", "reason": "Autoformer instance unavailable"}


    print(f"\n--- Foundation Model Forecasting (PatchTST): {symbol} ---")
    
    USE_FOUNDATION_MODEL_FORECAST = True # Set this flag as needed

    if USE_FOUNDATION_MODEL_FORECAST:
        foundation_forecast_output = run_patchtst_foundation_forecast(
            symbol_name=symbol,
            historical_data_df=df_raw, 
            prediction_length=5,       
            context_length=96,         # Default for ETTh1, check model card if using different checkpoint
            model_checkpoint="ibm-research/patchtst-etth1-pretrain", # <<< ENSURE CONSISTENCY OR RELY ON FUNCTION DEFAULT
            patch_length=16            # Default for ETTh1
        )
        default_return["forecasting_results"] = foundation_forecast_output
    else:
        # Fallback to your previous custom Autoformer logic if needed
        print("Skipping Foundation Model Forecast based on USE_FOUNDATION_MODEL_FORECAST flag.")

        default_return["forecasting_results"] = {"status": "skipped", "reason": "Foundation model disabled, no alternative specified here"}


    print(f"\n{'='*40}\nWORKFLOW FOR {symbol} CONCLUDED (Final Status: {default_return['status']})\n{'='*40}")
    return default_return


# --- MAIN EXECUTION BLOCK ---
if __name__ == "__main__":
    print("ðŸš€ Starting ENHANCED ML Trading System Workflow...")


    # Configure API Key
    if API_KEY == "YOUR_API_KEY_HERE" or not API_KEY:
        print("="*30 + " IMPORTANT " + "="*30)
        print("PLEASE SET YOUR TWELVE_DATA_API_KEY in the script.")
        print("Workflow will likely fail for data fetching.")
        print("="*70)
        # exit() # Optionally exit if API key is critical and not set

    stocks_to_analyze = ["MSFT", "AAPL"] # Add more symbols as needed
    all_run_results = {}


    FORCE_TRAIN_AUTOFORMER_FLAG = False # Set to True for initial training run or re-training

    RUN_OPTUNA_FOR_LGBM_FLAG = False # Set to True for HPO, False for faster runs with defaults

    for i, stock_symbol in enumerate(stocks_to_analyze):
        print(f"\nðŸŽ¯ Processing {stock_symbol} ({i+1}/{len(stocks_to_analyze)})...")
        try:
            stock_run_data = run_full_workflow(
                symbol=stock_symbol,
                start_date_str=START_DATE,
                end_date_str=END_DATE,
                api_key_val=API_KEY,
                force_train_autoformer=FORCE_TRAIN_AUTOFORMER_FLAG,
                run_optuna_lgbm=RUN_OPTUNA_FOR_LGBM_FLAG
            )
            all_run_results[stock_symbol] = stock_run_data
        except Exception as e_main_loop:
            print(f"FATAL ERROR processing {stock_symbol}: {e_main_loop}\n{traceback.format_exc()}")
            all_run_results[stock_symbol] = {"symbol": stock_symbol, "status": f"Workflow Crashed: {e_main_loop}"}

        if i < len(stocks_to_analyze) - 1:
            print(f"Pausing for a few seconds before processing next stock...")
            time.sleep(10) # API rate limiting consideration

    print("\n\n--- ðŸ“ˆ DETAILED RESULTS SUMMARY ---")
    for symbol_res_key, res_item in all_run_results.items():
        print(f"\n--- Results for: {symbol_res_key} ---")
        if res_item and isinstance(res_item, dict):
            print(f"  Overall Status: {res_item.get('status', 'N/A')}")
            print(f"  Raw Data Shape: {res_item.get('raw_data_shape', 'N/A')}")
            print(f"  Featured Data Shape: {res_item.get('featured_data_shape', 'N/A')}")
            print(f"  X_train Shape: {res_item.get('X_train_shape', 'N/A')}")
            print(f"  Selected Features Count: {res_item.get('selected_features_count', 'N/A')}")
            # print(f"  Selected Feature Names: {res_item.get('selected_feature_names', 'N/A')[:5]}...") # Print first 5

            if res_item.get('lgbm_optimized_params'):
                print(f"  LGBM Tuned Params (first few): {{ {', '.join([f'{k}: {v}' for k,v in list(res_item['lgbm_optimized_params'].items())[:3]])} ... }}")

            if res_item.get('ml_model_object'):
                print(f"  LightGBM Model Trained: Yes")
                if res_item.get('lgbm_feature_importance') is not None:
                    top_feat_lgbm = res_item['lgbm_feature_importance'].head(3)
                    print(f"  Top LGBM Features: {[(row['Feature'], row['Importance']) for index, row in top_feat_lgbm.iterrows()]}")
            else:
                print(f"  LightGBM Model Trained: No or Failed/Skipped")

            forecast_details = res_item.get('forecasting_results')
            if forecast_details and isinstance(forecast_details, dict) and forecast_details.get('status') != 'failed':
                print(f"  Forecasting Method: {forecast_details.get('method', 'N/A')}")
                print(f"  Forecast Values (next {len(forecast_details.get('forecast', []))} days): {forecast_details.get('forecast', 'N/A')}")
                print(f"  Predicted Direction: {forecast_details.get('direction', 'N/A')}")
                print(f"  Predicted Magnitude: {forecast_details.get('magnitude', 'N/A'):.2f}%" if isinstance(forecast_details.get('magnitude'), (int, float)) else "N/A")
                print(f"  Forecast Confidence: {forecast_details.get('confidence_label', 'N/A')} (Std: {forecast_details.get('confidence_metric_value', 'N/A'):.4f})" if isinstance(forecast_details.get('confidence_metric_value'), (int,float)) else "N/A")
            elif forecast_details:
                print(f"  Forecasting Status: Failed ({forecast_details.get('reason', 'Unknown')})")
            else:
                print("  Forecasting Data: Not available or workflow did not reach forecasting stage.")
        else:
            print(f"  No valid results dictionary for {symbol_res_key}")

    print(f"\nðŸŽ‰ Enhanced Analysis Run Complete!")

    # Installation guide remains useful
    print("\n" + "="*60)
    print("ðŸ’¡ Recommendation for Further Forecasting Enhancement:")
    print("="*60)
    print("For more advanced forecasting models beyond the custom Autoformer stub:")
    print("1. Fully implement and thoroughly train the `AutoformerPredictor`.")
    print("2. Explore libraries like Darts, Prophet, NeuralProphet, PyTorch-Forecasting, or Sktime's forecasting estimators.")
    print("   (Ensure they are installed: pip install darts prophet neuralprophet pytorch-forecasting sktime)")
    print("="*60)
```
