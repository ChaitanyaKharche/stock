```python
import traceback
import os
import time
import requests
import numpy as np
import pandas as pd
import pandas_ta
import yfinance as yf
from datetime import datetime, timedelta
import pywt
import antropy as ant
import lightgbm as lgb
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import accuracy_score, classification_report, log_loss, roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest, mutual_info_classif, f_classif, VarianceThreshold

# Foundational Model Imports
from transformers import PatchTSTConfig, PatchTSTForPrediction # No AutoProcessor needed if constructing manually

# --- Global Flags and Initializations for Optional Libraries ---
optuna_available = False
torch_available = False
sktime_available = False
ClaSPSegmentation = None
dowhy_available = False
CausalModel = None
nx = None # For networkx
onnx_available = False
skl2onnx_available = False
onnxmltools_available = False
FloatTensorType = None


if hasattr(pd.DataFrame, 'ta') is False and pandas_ta is not None:
    try:
        pd.DataFrame.ta = pandas_ta.Core(df=None)
        print("pandas_ta DataFrame accessor registered globally.")
    except Exception as e:
        print(f"Could not globally register pandas_ta accessor: {e}")

try:
    import optuna
    optuna_available = True
    print("Optuna imported successfully.")
except ImportError:
    print("Optuna not found. LightGBM hyperparameter optimization with Optuna will be skipped.")

try:
    import torch
    import torch.nn as nn
    import torch.optim as optim
    torch_available = True
    print("PyTorch imported successfully.")
    if torch.cuda.is_available():
        print(f"PyTorch CUDA available: True, Version: {torch.version.cuda}")
        print(f"Using PyTorch on GPU: {torch.cuda.get_device_name(0)}")
    else:
        print("PyTorch CUDA available: False.")
    print("Neural ODE features are DISABLED by configuration (per original script).")
except ImportError:
    print("PyTorch not found. Custom Transformer and Foundation Model features will be SKIPPED.")

try:
    import sktime
    from sktime.annotation.clasp import ClaSPSegmentation
    sktime_available = True
    print(f"Sktime {sktime.__version__} imported successfully (ClaSPSegmentation ready).")
except ImportError as e_sktime_import:
    print(f"Sktime modules import failed: {e_sktime_import}. ClaSP Regime detection will be skipped.")
except Exception as e_sktime_general:
    print(f"A general error occurred during sktime import or setup: {e_sktime_general}")

try:
    import dowhy
    from dowhy import CausalModel
    import networkx as nx
    dowhy_available = True
    print(f"DoWhy {dowhy.__version__} and NetworkX {nx.__version__} imported successfully.")
except ImportError:
    print("DoWhy or NetworkX not found. Causal Discovery will be skipped.")

try:
    import onnx
    onnx_available = True
    import onnxruntime as ort
    import skl2onnx
    skl2onnx_available = True # Set flag here
    from skl2onnx.common.data_types import FloatTensorType # Assign to global
    import onnxmltools
    onnxmltools_available = True # Set flag here
    print("ONNX, ONNXRuntime, skl2onnx, and onnxmltools imported successfully.")
    if hasattr(onnxmltools, '__version__'):
          print(f"Onnxmltools version: {onnxmltools.__version__}")
except ImportError as e:
    print(f"One or more ONNX components not found: {e}. ONNX features will be skipped.")

print("\nAll libraries and modules conditional imports attempted.")

# --- Constants ---
TWELVE_DATA_API_KEY = "b6dbb92e551a46f2b20de27540aeef0a"
API_KEY = TWELVE_DATA_API_KEY
DEFAULT_SYMBOL = "MSFT"
START_DATE = (datetime.now() - timedelta(days=3*365)).strftime('%Y-%m-%d')
END_DATE = datetime.now().strftime('%Y-%m-%d')
AUTOFORMER_MODEL_PATH = "autoformer_predictor_trained.pth"


# --- Function Definitions ---

class AutoformerPredictor:
    def __init__(self, input_len=60, pred_len=5, d_model=64, n_heads=8, num_encoder_layers=2, num_decoder_layers=1, dim_feedforward=256):
        self.device = 'cuda' if torch.cuda.is_available() and torch_available else 'cpu'
        self.model = None
        self.input_len = input_len
        self.pred_len = pred_len
        self.d_model = d_model

        if torch_available:
            print(f"ðŸ”§ Initializing Custom AutoformerPredictor on device: {self.device}")
            self.model = nn.Transformer(
                d_model=d_model, nhead=n_heads, num_encoder_layers=num_encoder_layers,
                num_decoder_layers=num_decoder_layers, dim_feedforward=dim_feedforward,
                activation='gelu', batch_first=False
            ).to(self.device)
            self.enc_embedding = nn.Linear(1, d_model).to(self.device)
            self.dec_embedding = nn.Linear(1, d_model).to(self.device)
            self.projection = nn.Linear(d_model, 1).to(self.device)
        else:
            print("PyTorch not available. AutoformerPredictor cannot be initialized.")

    def _generate_square_subsequent_mask(self, sz):
        mask = (torch.triu(torch.ones(sz, sz)) == 1).transpose(0, 1)
        mask = mask.float().masked_fill(mask == 0, float('-inf')).masked_fill(mask == 1, float(0.0))
        return mask.to(self.device)

    def forward(self, src, tgt):
        if not torch_available or self.model is None: return None
        src_embedded = self.enc_embedding(src)
        tgt_embedded = self.dec_embedding(tgt)
        tgt_mask = self._generate_square_subsequent_mask(tgt.size(0))
        output = self.model(src_embedded, tgt_embedded, tgt_mask=tgt_mask)
        output = self.projection(output)
        return output

    def train_model(self, train_loader, epochs=10, learning_rate=0.001):
        if not torch_available or self.model is None:
            print("Cannot train: PyTorch or model not available.")
            return
        if not train_loader:
            print("Cannot train: No data loader provided.")
            return
        print(f"Starting AutoformerPredictor training for {epochs} epochs...")
        self.model.train()
        optimizer = optim.Adam(self.model.parameters(), lr=learning_rate)
        criterion = nn.MSELoss()
        for epoch in range(epochs):
            epoch_loss = 0
            for i, (batch_src, batch_tgt_input, batch_tgt_output) in enumerate(train_loader):
                batch_src = batch_src.float().to(self.device).permute(1,0,2)
                batch_tgt_input = batch_tgt_input.float().to(self.device).permute(1,0,2)
                batch_tgt_output = batch_tgt_output.float().to(self.device).permute(1,0,2)
                optimizer.zero_grad()
                predictions = self.forward(batch_src, batch_tgt_input)
                if predictions is None:
                    print(f"Epoch {epoch+1}, Batch {i+1}: Forward pass returned None. Skipping batch.")
                    continue
                loss = criterion(predictions, batch_tgt_output)
                loss.backward()
                optimizer.step()
                epoch_loss += loss.item()
            avg_epoch_loss = epoch_loss / len(train_loader) if len(train_loader) > 0 else float('nan')
            print(f"Epoch {epoch+1}/{epochs}, Average Loss: {avg_epoch_loss:.4f}")
        print("Training finished.")
        self.save_model(AUTOFORMER_MODEL_PATH)

    def predict(self, series_data):
        if not torch_available or self.model is None:
            print("Cannot predict: PyTorch or model not available.")
            return np.array([np.nan] * self.pred_len)
        self.model.eval()
        with torch.no_grad():
            if isinstance(series_data, pd.Series): series_data = series_data.values
            if isinstance(series_data, list): series_data = np.array(series_data)
            if not isinstance(series_data, np.ndarray):
                print(f"Unsupported series data type for predict: {type(series_data)}")
                return np.array([np.nan] * self.pred_len)
            if series_data.ndim > 1: series_data = series_data.squeeze()
            if series_data.ndim == 0:
                 print(f"Series data is scalar, cannot process.")
                 return np.array([np.nan]*self.pred_len)

            mean_val = series_data[-self.input_len:].mean() if len(series_data) >= self.input_len else series_data.mean()
            std_val = series_data[-self.input_len:].std() if len(series_data) >= self.input_len else series_data.std()
            std_val = std_val if std_val > 1e-8 else 1e-8

            if len(series_data) < self.input_len:
                padding_val = series_data[0] if len(series_data) > 0 else 0
                norm_series_np = (np.concatenate([np.full(self.input_len - len(series_data), padding_val), series_data]) - mean_val) / std_val
            else:
                norm_series_np = (series_data[-self.input_len:] - mean_val) / std_val
            
            src_tensor = torch.tensor(norm_series_np, dtype=torch.float32).unsqueeze(0).unsqueeze(-1).permute(1,0,2).to(self.device)
            
            # Initialize decoder input: (S=1, N=1, F=1), using last element of src as a common practice (or zero)
            # This assumes the first decoder input is based on the last known encoder input step.
            # A zero or special <SOS> token embedding is also common.
            # For simplicity, let's use a zero tensor as the start of the target sequence.
            current_tgt_sequence_normalized = torch.zeros((1, 1, 1), dtype=torch.float32, device=self.device) # (S=1, N=1, F=1)
            
            predictions_normalized_list = []

            for _ in range(self.pred_len):
                prediction_step_output = self.forward(src_tensor, current_tgt_sequence_normalized) # (S_tgt, N, 1)
                
                if prediction_step_output is None: return np.array([np.nan] * self.pred_len)

                # Get the last predicted value from the output sequence
                last_predicted_value_normalized = prediction_step_output[-1:, :, :] # Shape (1, N, 1)
                predictions_normalized_list.append(last_predicted_value_normalized.squeeze().cpu().item())
                
                # Append the new prediction to the decoder input sequence for the next step
                # Note: self.dec_embedding is applied inside self.forward
                current_tgt_sequence_normalized = torch.cat(
                    [current_tgt_sequence_normalized, last_predicted_value_normalized], dim=0
                )
                # Optional: Keep decoder input sequence length fixed by truncating if it grows too long
                # if current_tgt_sequence_normalized.size(0) > some_max_decoder_input_len:
                #    current_tgt_sequence_normalized = current_tgt_sequence_normalized[-some_max_decoder_input_len:,:,:]


            denorm_preds = (np.array(predictions_normalized_list) * std_val) + mean_val
            return denorm_preds

    def save_model(self, path):
        if not torch_available or self.model is None: return
        torch.save({
            'model_state_dict': self.model.state_dict(),
            'enc_embedding_state_dict': self.enc_embedding.state_dict(),
            'dec_embedding_state_dict': self.dec_embedding.state_dict(),
            'projection_state_dict': self.projection.state_dict(),
            'input_len': self.input_len, 'pred_len': self.pred_len, 'd_model': self.d_model
        }, path)
        print(f"AutoformerPredictor model saved to {path}")

    def load_model(self, path):
        if not torch_available: print("Cannot load model: PyTorch not available."); return False
        if not os.path.exists(path): print(f"Model file not found at {path}. Initializing new model."); return False
        checkpoint = torch.load(path, map_location=self.device)
        self.input_len = checkpoint.get('input_len', self.input_len)
        self.pred_len = checkpoint.get('pred_len', self.pred_len)
        saved_d_model = checkpoint.get('d_model', self.d_model)
        
        # Re-initialize model parts if they were not part of the class instance before loading (e.g. if loading into a fresh class)
        # This part assumes that the architectural params like n_heads, num_layers are fixed or known.
        # For robust loading, these params should also be saved or passed to __init__ if they can vary.
        self.model = nn.Transformer(d_model=saved_d_model, nhead=8, num_encoder_layers=2, num_decoder_layers=1, dim_feedforward=256, activation='gelu', batch_first=False).to(self.device)
        self.enc_embedding = nn.Linear(1, saved_d_model).to(self.device)
        self.dec_embedding = nn.Linear(1, saved_d_model).to(self.device)
        self.projection = nn.Linear(saved_d_model, 1).to(self.device)

        self.model.load_state_dict(checkpoint['model_state_dict'])
        self.enc_embedding.load_state_dict(checkpoint['enc_embedding_state_dict'])
        self.dec_embedding.load_state_dict(checkpoint['dec_embedding_state_dict'])
        self.projection.load_state_dict(checkpoint['projection_state_dict'])
        self.model.eval()
        print(f"AutoformerPredictor model loaded from {path}")
        return True

def create_sequences(data, input_len, pred_len):
    X, y_input, y_output = [], [], []
    # data is expected to be a 1D numpy array (scaled series)
    for i in range(len(data) - input_len - pred_len + 1):
        X.append(data[i:(i + input_len)])
        # For decoder input, a common strategy is to use the target sequence shifted by one, with a start token.
        # Or, for simpler encoder-decoder, decoder might start with last known encoder value or a zero.
        # Here, y_input is the sequence the decoder tries to predict, shifted.
        # Example: Encoder sees t1..t60. Decoder input starts with t60 (or <SOS>). Target output is t61..t65.
        # During training, decoder input (teacher forcing) would be t60, t61, t62, t63, t64.
        # For this example, let's make y_input match y_output for simplicity, but shifted for teacher forcing.
        # This part needs careful alignment with how AutoformerPredictor.forward uses tgt.
        # For the current AutoformerPredictor.forward, tgt is the *target sequence* for teacher forcing.
        y_dec_inp_seq = np.concatenate(([data[i + input_len -1]], data[(i + input_len):(i + input_len + pred_len -1)])) # Start with last known, then actuals
        y_input.append(y_dec_inp_seq)
        y_output.append(data[(i + input_len):(i + input_len + pred_len)])
    return np.array(X), np.array(y_input), np.array(y_output)

def train_or_load_autoformer(df_series, input_len=60, pred_len=5, force_train=False):
    if not torch_available: print("PyTorch not available, cannot train or load Autoformer."); return None
    autoformer_forecaster = AutoformerPredictor(input_len=input_len, pred_len=pred_len)
    if not force_train and os.path.exists(AUTOFORMER_MODEL_PATH):
        print(f"Loading existing Autoformer model from {AUTOFORMER_MODEL_PATH}")
        if autoformer_forecaster.load_model(AUTOFORMER_MODEL_PATH): return autoformer_forecaster
        else: print(f"Failed to load model from {AUTOFORMER_MODEL_PATH}. Proceeding to train a new one."); force_train = True
    if force_train:
        print("Training a new AutoformerPredictor model...")
        scaler = StandardScaler()
        series_scaled = scaler.fit_transform(df_series.values.reshape(-1, 1))
        X_seq, y_seq_dec_in, y_seq_out = create_sequences(series_scaled.flatten(), input_len, pred_len)
        if len(X_seq) == 0: print("Not enough data for Autoformer training. Model remains un-trained."); return autoformer_forecaster
        X_seq = X_seq.reshape(X_seq.shape[0], X_seq.shape[1], 1)
        y_seq_dec_in = y_seq_dec_in.reshape(y_seq_dec_in.shape[0], y_seq_dec_in.shape[1], 1)
        y_seq_out = y_seq_out.reshape(y_seq_out.shape[0], y_seq_out.shape[1], 1)
        dataset = torch.utils.data.TensorDataset(torch.from_numpy(X_seq).float(), torch.from_numpy(y_seq_dec_in).float(), torch.from_numpy(y_seq_out).float())
        train_loader = torch.utils.data.DataLoader(dataset, batch_size=16, shuffle=True)
        autoformer_forecaster.train_model(train_loader, epochs=20) # epochs=20 for demo
    else: print("Autoformer training skipped. Model is un-trained if not loaded.")
    return autoformer_forecaster

def fetch_twelve_data(symbol, api_key, start_date_str=None, end_date_str=None):
    base_url = "https://api.twelvedata.com/time_series"
    params = {"symbol": symbol, "interval": "1day", "apikey": api_key, "format": "JSON", "outputsize": 5000}
    if start_date_str: params["start_date"] = start_date_str
    if end_date_str: params["end_date"] = end_date_str
    print(f"Fetching data for {symbol} from Twelve Data (interval=1day, from {start_date_str} to {end_date_str})...")
    try:
        response = requests.get(base_url, params=params, timeout=30)
        response.raise_for_status()
        data = response.json()
    except requests.exceptions.RequestException as e: print(f"Request failed for {symbol}: {e}"); return None
    except ValueError as e: print(f"Failed to parse JSON for {symbol}: {e}. Response: {response.text[:200]}..."); return None
    if isinstance(data, dict) and (data.get("status") == "error" or "values" not in data): print(f"API Error for {symbol} (Code: {data.get('code')}): {data.get('message', 'Unknown error')}"); return None
    if not isinstance(data, dict) or "values" not in data or not data["values"]: print(f"No data values for {symbol}, unexpected format, or empty list."); return None
    df = pd.DataFrame(data["values"]).rename(columns={'datetime': 'date'})
    for col in ['open', 'high', 'low', 'close', 'volume']:
        if col in df.columns: df[col] = pd.to_numeric(df[col], errors='coerce').astype('float64')
        else:
            if col in ['open', 'high', 'low', 'close']: print(f"Critical column '{col}' missing. Cannot proceed."); return None
            df[col] = 0.0
    if 'date' not in df.columns: print("Critical Error: 'date' column missing."); return None
    df.index = pd.to_datetime(df['date']); df.drop(columns=['date'], inplace=True); df.sort_index(inplace=True)
    df.dropna(subset=[col for col in ['open', 'high', 'low', 'close'] if col in df.columns], inplace=True)
    if df.empty: print(f"No data remaining for {symbol} after processing."); return None
    print(f"Successfully fetched/processed {len(df)} data points for {symbol}."); return df

def add_technical_indicators(df):
    df_feat = df.copy()
    if not hasattr(df_feat.ta, 'rsi'): print("pandas_ta not registered. Skipping TIs."); return df_feat
    print("Adding technical indicators...")
    try:
        ohlcv_cols = ['open', 'high', 'low', 'close', 'volume']
        for col in ohlcv_cols:
            if col not in df_feat.columns:
                if col in ['high', 'low', 'open', 'close']: df_feat[col] = df_feat['close']
                elif col == 'volume': df_feat[col] = 0.0
            df_feat[col] = df_feat[col].astype('float64') # Ensure float64 for pandas_ta

        c, h, l, v = 'close', 'high', 'low', 'volume'
        df_feat.ta.rsi(close=df_feat[c], length=14, append=True)
        df_feat.ta.rsi(close=df_feat[c], length=9, append=True, col_names='RSI_9')
        df_feat.ta.rsi(close=df_feat[c], length=25, append=True, col_names='RSI_25')
        df_feat.ta.macd(close=df_feat[c], fast=12, slow=26, signal=9, append=True)
        df_feat.ta.macd(close=df_feat[c], fast=5, slow=15, signal=9, append=True, col_names=('MACD_5_15_9', 'MACDh_5_15_9', 'MACDs_5_15_9'))
        for p in [10, 20, 50, 100, 200]:
            df_feat.ta.sma(close=df_feat[c], length=p, append=True)
            df_feat.ta.ema(close=df_feat[c], length=p, append=True)
        df_feat.ta.bbands(close=df_feat[c], length=20, std=2, append=True)
        if all(x in df_feat.columns for x in [h,l,c]):
            df_feat.ta.atr(high=df_feat[h], low=df_feat[l], close=df_feat[c], length=14, append=True)
            df_feat.ta.adx(high=df_feat[h], low=df_feat[l], close=df_feat[c], length=14, append=True)
            df_feat.ta.stoch(high=df_feat[h], low=df_feat[l], close=df_feat[c], append=True)
            df_feat.ta.willr(high=df_feat[h], low=df_feat[l], close=df_feat[c], append=True)
            df_feat.ta.cci(high=df_feat[h], low=df_feat[l], close=df_feat[c], append=True)
        if all(x in df_feat.columns for x in [h,l,c,v]):
            df_feat.ta.mfi(high=df_feat[h], low=df_feat[l], close=df_feat[c], volume=df_feat[v], append=True)
        df_feat.columns = df_feat.columns.str.replace('[^A-Za-z0-9_]+', '', regex=True)
        print("Technical indicators added.")
    except Exception as e: print(f"Error adding TIs: {e}\n{traceback.format_exc()}")
    return df_feat

def add_optimized_features(df, price_col='close', volume_col='volume'):
    print("Adding optimized statistical and ratio features...")
    df_new = df.copy()
    df_new['returns'] = df_new[price_col].pct_change()
    safe_price = df_new[price_col].replace(0, np.nan); safe_price_shifted = df_new[price_col].shift(1).replace(0, np.nan)
    df_new['log_returns'] = np.log(safe_price / safe_price_shifted)
    for window in [5, 10, 20, 50]:
        df_new[f'volatility_{window}'] = df_new['log_returns'].rolling(window).std()
        df_new[f'skew_{window}'] = df_new['log_returns'].rolling(window).skew()
        df_new[f'kurtosis_{window}'] = df_new['log_returns'].rolling(window).kurt()
    if volume_col in df_new.columns and price_col in df_new.columns:
        rolling_mean_volume = df_new[volume_col].rolling(20).mean().replace(0, np.nan)
        df_new['volume_ratio'] = df_new[volume_col] / rolling_mean_volume
        df_new['price_volume'] = df_new[price_col] * df_new[volume_col]
        df_new['volume_change'] = df_new[volume_col].pct_change()
    if all(col in df_new.columns for col in ['high', 'low', 'close']):
        safe_low = df_new['low'].replace(0, np.nan); safe_high = df_new['high'].replace(0, np.nan); safe_close = df_new['close'].replace(0, np.nan)
        df_new['high_low_ratio'] = df_new['high'] / safe_low
        df_new['close_to_high_ratio'] = safe_close / safe_high
        df_new['close_to_low_ratio'] = safe_close / safe_low
        df_new['intraday_range_norm'] = (df_new['high'] - df_new['low']) / safe_close
    if 'RSI_14' in df_new.columns:
        df_new['RSI_signal'] = 0; df_new.loc[df_new['RSI_14'] < 30, 'RSI_signal'] = 1; df_new.loc[df_new['RSI_14'] > 70, 'RSI_signal'] = -1
    if 'MACD_12_26_9' in df_new.columns and 'MACDs_12_26_9' in df_new.columns:
        df_new['MACD_signal_line_cross'] = (df_new['MACD_12_26_9'] > df_new['MACDs_12_26_9']).astype(int)
    return df_new

def add_wavelet_features(df, column='close', wavelet='mexh', scales_range=(1, 32), num_scales_to_extract=5):
    df_feat = df.copy(); print(f"Adding CWT for '{column}' using wavelet '{wavelet}'...")
    if pywt is None: print("PyWavelets not available."); return df_feat
    signal = df_feat[column].values
    if len(signal) < scales_range[1] + 5: print(f"Signal length {len(signal)} too short for CWT. Skipping."); return df_feat
    actual_max_scale = min(scales_range[1], len(signal) // 2 - 1)
    if actual_max_scale < scales_range[0]: print("Max scale too small. Skipping CWT."); return df_feat
    scales = np.arange(scales_range[0], actual_max_scale + 1)
    if len(scales) == 0: print("No valid scales for CWT. Skipping."); return df_feat
    try:
        coefficients, _ = pywt.cwt(signal, scales, wavelet)
        coeffs_df = pd.DataFrame(coefficients.T, index=df_feat.index, columns=[f"cwt_scale_{s}" for s in scales])
        df_feat[f'{column}_cwt_mean'] = coeffs_df.mean(axis=1); df_feat[f'{column}_cwt_std'] = coeffs_df.std(axis=1)
        s_indices_to_extract = np.linspace(0, len(scales)-1, min(num_scales_to_extract, len(scales)), dtype=int)
        for s_idx in s_indices_to_extract:
            actual_scale_val = scales[s_idx]; col_name_for_scale = f"cwt_scale_{actual_scale_val}"
            if col_name_for_scale in coeffs_df.columns: df_feat[f'{column}_cwt_energy_s{actual_scale_val}'] = coeffs_df[col_name_for_scale]**2
        print("Wavelet features added.")
    except Exception as e: print(f"Error adding wavelet features: {e}\n{traceback.format_exc()}")
    return df_feat

def add_entropy_features(df, column='close', window=40):
    df_feat = df.copy(); print(f"Adding Entropy for '{column}' (window={window})...")
    if ant is None: print("Antropy not available."); return df_feat
    if len(df_feat) < window + 15: print(f"Data too short for entropy. Skipping."); return df_feat
    try:
        sig = df_feat[column].astype(float)
        df_feat[f'{column}_entropy_sample'] = sig.rolling(window=window, min_periods=window).apply(lambda x: ant.sample_entropy(x.dropna()) if x.dropna().shape[0] >= window//2 else np.nan, raw=False)
        df_feat[f'{column}_entropy_spectral'] = sig.rolling(window=window, min_periods=window).apply(lambda x: ant.spectral_entropy(x.dropna(), sf=1.0, method='welch', nperseg=min(x.dropna().shape[0], window // 2 if window // 2 > 0 else 1)) if x.dropna().shape[0] == window else np.nan, raw=False)
        print("Entropy features added.")
    except Exception as e: print(f"Error adding entropy features: {e}\n{traceback.format_exc()}")
    return df_feat

def add_advanced_technical_features(df, price_col='close', high_col='high', low_col='low', volume_col='volume'):
    print("Adding advanced TIs...")
    df_new = df.copy()
    if not hasattr(df_new.ta, 'mom'): print("pandas_ta not registered. Skipping advanced TIs."); return df_new
    try:
        if not all(c in df_new.columns for c in [price_col, high_col, low_col]): print("Missing OHLC. Skipping adv TIs."); return df_new
        df_new[price_col] = df_new[price_col].astype(float); df_new[high_col] = df_new[high_col].astype(float); df_new[low_col] = df_new[low_col].astype(float)
        if volume_col in df_new.columns: df_new[volume_col] = df_new[volume_col].astype(float)

        df_new.ta.mom(close=df_new[price_col], append=True)
        df_new.ta.roc(close=df_new[price_col], append=True)
        df_new.ta.natr(high=df_new[high_col], low=df_new[low_col], close=df_new[price_col], append=True)
        df_new.ta.aroon(high=df_new[high_col], low=df_new[low_col], append=True)
        df_new.ta.stc(close=df_new[price_col], tclength=23, fast=50, slow=100, factor=0.5, append=True, col_names=('STC_23_50_05', 'STCD_23_50_05', 'STCK_23_50_05'))
        if volume_col in df_new.columns:
            df_new.ta.pvol(close=df_new[price_col], volume=df_new[volume_col], append=True)
            df_new.ta.cmf(high=df_new[high_col], low=df_new[low_col], close=df_new[price_col], volume=df_new[volume_col], append=True)
        df_new.columns = df_new.columns.str.replace('[^A-Za-z0-9_]+', '', regex=True)
        print("Advanced TIs added.")
    except Exception as e: print(f"Error adding advanced TIs: {e}\n{traceback.format_exc()}")
    return df_new

def add_transformer_features_conceptual(df, column='close', sequence_length=20):
    df_feat = df.copy(); print(f"Adding Transformer-inspired conceptual features for '{column}'...")
    if len(df_feat) < sequence_length + 5: print(f"Data too short. Skipping."); return df_feat
    feature_col_base = f"{column}_trans_seq"
    for col_name_suffix in ['mean', 'std', 'trend', 'volatility', 'autocorr1']: df_feat[f'{feature_col_base}_{col_name_suffix}'] = np.nan
    try:
        data_series = df_feat[column].values
        windows = np.lib.stride_tricks.sliding_window_view(data_series, sequence_length)
        results = {'mean': [], 'std': [], 'trend': [], 'volatility': [], 'autocorr1': []}
        for seq in windows:
            mean_val, std_val = np.mean(seq), np.std(seq)
            norm_seq = (seq - mean_val) / std_val if std_val > 1e-8 else np.zeros_like(seq)
            results['mean'].append(np.mean(norm_seq)); results['std'].append(np.std(norm_seq))
            if len(norm_seq) > 1:
                try: results['trend'].append(np.polyfit(np.arange(len(norm_seq)), norm_seq, 1)[0] if not np.isnan(np.polyfit(np.arange(len(norm_seq)), norm_seq, 1)[0]) else 0.0)
                except: results['trend'].append(0.0)
                diff_norm_seq = np.diff(norm_seq)
                results['volatility'].append(np.std(diff_norm_seq) if len(diff_norm_seq) > 0 else 0.0)
                if len(norm_seq) >= 2:
                    s1, s2 = norm_seq[:-1], norm_seq[1:]
                    if len(s1) >= 1 and np.std(s1) > 1e-8 and np.std(s2) > 1e-8:
                        try: results['autocorr1'].append(np.corrcoef(s1, s2)[0, 1] if not np.isnan(np.corrcoef(s1, s2)[0, 1]) else 0.0)
                        except: results['autocorr1'].append(0.0)
                    else: results['autocorr1'].append(0.0)
                else: results['autocorr1'].append(0.0)
            else: results['trend'].append(0.0); results['volatility'].append(0.0); results['autocorr1'].append(0.0)
        idx_start = sequence_length - 1
        for key, values in results.items(): df_feat.loc[df_feat.index[idx_start:idx_start+len(values)], f'{feature_col_base}_{key}'] = values
        print("Conceptual Transformer-inspired features added.")
    except Exception as e: print(f"Error conceptual Transformer features: {e}\n{traceback.format_exc()}")
    return df_feat

def add_clasp_regimes(df, column='close', period=20, n_cps=3):
    df_reg = df.copy(); print(f"Detecting regimes using ClaSPSegmentation for {column}...")
    if not sktime_available or ClaSPSegmentation is None: print("ClaSPSegmentation not available. Using simple fallback."); return detect_regimes_simple(df, column)
    if column not in df_reg.columns or df_reg[column].isnull().all(): print(f"'{column}' not found. Skipping ClaSP."); df_reg['regime'] = 0; return df_reg
    series_for_clasp = df_reg[column].dropna()
    if len(series_for_clasp) < period * 2: print(f"Series too short for ClaSP. Using simple fallback."); return detect_regimes_simple(df, column)
    try:
        clasp = ClaSPSegmentation(period_length=period, n_cps=n_cps); found_cps = clasp.fit_predict(series_for_clasp)
        df_reg['regime'] = 0; current_regime = 0; last_cp = 0
        for cp in found_cps:
            segment_indices = series_for_clasp.index[last_cp:cp]; df_reg.loc[segment_indices, 'regime'] = current_regime
            current_regime = (current_regime + 1) % (n_cps + 1); last_cp = cp
        segment_indices = series_for_clasp.index[last_cp:]; df_reg.loc[segment_indices, 'regime'] = current_regime
        df_reg['regime'] = df_reg['regime'].ffill().bfill().fillna(0)
        print(f"ClaSP regimes identified. Distribution:\n{df_reg['regime'].value_counts(normalize=True).sort_index()*100} %")
    except Exception as e: print(f"Error ClaSPSegmentation: {e}. Using simple fallback."); return detect_regimes_simple(df, column)
    return df_reg

def detect_regimes_simple(df, column='close', window=20):
    df_reg = df.copy(); print(f"Detecting regimes (simplified volatility-based) for {column}...")
    if column not in df_reg.columns: print(f"'{column}' not found. Skipping."); df_reg['regime_simple'] = 0; return df_reg
    returns = np.log(df_reg[column] / df_reg[column].shift(1)).dropna()
    rolling_vol = returns.rolling(window=window, min_periods=window//2).std()
    df_reg['regime_simple'] = 0
    if not rolling_vol.dropna().empty:
        vol_low_thresh = rolling_vol.quantile(0.33); vol_high_thresh = rolling_vol.quantile(0.67)
        df_reg.loc[rolling_vol.index[rolling_vol <= vol_low_thresh], 'regime_simple'] = 1
        df_reg.loc[rolling_vol.index[rolling_vol > vol_high_thresh], 'regime_simple'] = 2
    else: print("Not enough data for vol percentile. Defaulting simple regimes to 0.")
    df_reg['regime_simple'] = df_reg['regime_simple'].fillna(method='bfill').fillna(0)
    print(f"Simple Regimes (0:Med,1:Low,2:High): {df_reg['regime_simple'].value_counts(normalize=True).sort_index()*100} %")
    return df_reg

def balanced_target_definition(df, column='close', periods=5, lower_q_thresh=0.45, upper_q_thresh=0.55):
    df_t = df.copy(); print(f"Balanced target definition for '{column}' over {periods} periods...")
    if column not in df_t.columns: print(f"'{column}' not found for target."); df_t['target'] = 0; return df_t
    df_t[column] = pd.to_numeric(df_t[column], errors='coerce')
    df_t['future_log_return_target'] = np.log(df_t[column].shift(-periods) / df_t[column])
    valid_returns = df_t['future_log_return_target'].dropna()
    df_t['target'] = 0
    if len(valid_returns) > 20:
        lower_q_val = valid_returns.quantile(lower_q_thresh); upper_q_val = valid_returns.quantile(upper_q_thresh)
        df_t.loc[df_t['future_log_return_target'] < lower_q_val, 'target'] = 0
        df_t.loc[df_t['future_log_return_target'] > upper_q_val, 'target'] = 1
    else: print("Not enough valid returns for target balancing.")
    df_t.drop(columns=['future_log_return_target'], inplace=True, errors='ignore')
    print(f"Target distribution:\n{df_t['target'].value_counts(normalize=True, dropna=False)*100}")
    return df_t

def discover_causal_structure(df_features, target_col='target', max_feats=10, symbol=""):
    print(f"\nDiscovering causal structure for {symbol} using DoWhy...")
    if not dowhy_available or CausalModel is None: print("DoWhy not available."); return None
    df_c = df_features.copy()
    if target_col not in df_c.columns or df_c[target_col].isnull().all(): print(f"Target missing for causal discovery."); return None
    df_c[target_col] = pd.to_numeric(df_c[target_col], errors='coerce')
    cand_cols = [c for c in df_c.columns if pd.api.types.is_numeric_dtype(df_c[c]) and c != target_col and df_c[c].notnull().any()]
    if not cand_cols: print("No numeric candidate columns for causal discovery."); return None
    df_subset_for_causal = df_c[cand_cols + [target_col]].copy()
    df_subset_for_causal.replace([np.inf, -np.inf], np.nan, inplace=True)
    if cand_cols: scaler_causal = StandardScaler(); df_subset_for_causal[cand_cols] = scaler_causal.fit_transform(df_subset_for_causal[cand_cols])
    df_subset_for_causal.dropna(inplace=True)
    if df_subset_for_causal.empty or target_col not in df_subset_for_causal.columns or df_subset_for_causal[target_col].nunique() < 1: print("Not enough data after cleaning/scaling for causal discovery."); return None
    potential_causes = ['RSI_14', 'MACDh_12_26_9', 'ADX_14', 'ATR_14', f'{target_col}_cwt_mean', f'{target_col}_cwt_std', f'{target_col}_entropy_sample', 'regime_simple', 'volatility_20', 'log_returns', 'BBP_20_20', 'BBB_20_20'] # Adjusted for target_col prefix in cwt/entropy if done that way
    graph_feats = [c for c in potential_causes if c in df_subset_for_causal.columns and c != target_col and df_subset_for_causal[c].nunique() > 1]
    if not graph_feats:
        print("Predefined causal graph_feats not suitable, selecting top varying features.")
        num_to_select = min(max_feats, len(cand_cols))
        if num_to_select > 0: graph_feats = df_subset_for_causal[cand_cols].var().nlargest(num_to_select).index.tolist()
        else: print("No candidate columns for graph_feats fallback."); return None
    if not graph_feats: print("No suitable graph features for causal discovery."); return None
    final_df_for_causal_model = df_subset_for_causal[graph_feats + [target_col]].copy()
    if final_df_for_causal_model.empty or final_df_for_causal_model.shape[0] < 20 or final_df_for_causal_model[target_col].nunique() < 1: print("Final DF for causal model too small."); return None
    print(f"DoWhy using graph features: {graph_feats} for Outcome: {target_col}")
    treatment_var = graph_feats[0]; graph_str = "digraph { " + "; ".join([f'"{f}" -> "{target_col}"' for f in graph_feats]) + " }"
    print(f"Generated Causal Graph:\n{graph_str}")
    try: model = CausalModel(data=final_df_for_causal_model, treatment=treatment_var, outcome=target_col, graph=graph_str); print("DoWhy CausalModel created."); return model
    except Exception as e: print(f"DoWhy CausalModel error: {e}\n{traceback.format_exc()}"); return None

def causal_feature_selection(df_full_features, target_col='target', max_causal_feats=10):
    print("Performing Causal Feature Selection with DoWhy...")
    if not dowhy_available or CausalModel is None: print("DoWhy not available. Skipping."); return []
    df_clean = df_full_features.copy(); df_clean[target_col] = pd.to_numeric(df_clean[target_col], errors='coerce')
    numeric_features_for_causal = [col for col in df_clean.columns if col != target_col and pd.api.types.is_numeric_dtype(df_clean[col]) and df_clean[col].nunique() > 1]
    if not numeric_features_for_causal: print("No numeric features for Causal FS."); return []
    scaler_cfs = StandardScaler(); df_clean[numeric_features_for_causal] = scaler_cfs.fit_transform(df_clean[numeric_features_for_causal])
    df_clean.replace([np.inf, -np.inf], np.nan, inplace=True); df_clean.dropna(subset=numeric_features_for_causal + [target_col], inplace=True)
    if df_clean.empty or target_col not in df_clean.columns or df_clean[target_col].nunique() < 1: print("Data empty after cleaning for Causal FS."); return []
    final_features_for_causal = [f for f in numeric_features_for_causal if f in df_clean.columns and df_clean[f].nunique() >1]
    if not final_features_for_causal : print("No features remaining for Causal FS."); return []
    print(f"Starting causal effect estimation for {len(final_features_for_causal)} features.")
    feature_effects = {}
    for feature_as_treatment in final_features_for_causal:
        try:
            current_data_for_model = df_clean[final_features_for_causal + [target_col]] # Use consistent data subset
            model = CausalModel(data=current_data_for_model, treatment=feature_as_treatment, outcome=target_col, graph=None) # Let DoWhy build graph
            identified_estimand = model.identify_effect(proceed_when_unidentifiable=True)
            estimate = model.estimate_effect(identified_estimand, method_name="backdoor.linear_regression", test_significance=False, method_params={'force_univariate': True})
            if estimate is not None and hasattr(estimate, 'value') and not np.isnan(estimate.value): feature_effects[feature_as_treatment] = abs(estimate.value)
        except Exception as e: pass # Suppress errors for individual features to continue loop
    if not feature_effects: print("Causal ranking yielded no valid effects.")
    else:
        sorted_features = sorted(feature_effects.items(), key=lambda x: x[1], reverse=True)
        print(f"Causal features by effect (top {min(5, len(sorted_features))} of {len(sorted_features)}): {sorted_features[:min(5, len(sorted_features))]}")
        return sorted_features[:max_causal_feats]
    return []

def prepare_ml_data(df, target_col='target', test_split_size=0.15, min_test_samples=50):
    print("Preparing ML data...")
    if target_col not in df.columns: print(f"Target '{target_col}' missing."); return None, None, None, None, None
    cols_to_drop_base = ['open', 'high', 'low', 'close', 'volume', 'returns']
    cols_to_drop_dynamic = [c for c in df.columns if 'target_' in c and c != target_col] + [c for c in df.columns if 'future_return' in c]
    all_cols_to_drop = list(set(cols_to_drop_base + cols_to_drop_dynamic))
    X = df.drop(columns=[c for c in all_cols_to_drop if c in df.columns] + [target_col], errors='ignore')
    y = df[target_col].copy()
    if y.isnull().all(): print("Target all NaN."); return None, None, None, None, None
    valid_target_mask = y.notna(); X = X.loc[valid_target_mask]; y = y.loc[valid_target_mask]
    if X.empty or y.empty: print("X or y empty after target NaN filter."); return None, None, None, None, None
    X.replace([np.inf, -np.inf], np.nan, inplace=True)
    if X.isnull().any().any(): X = X.fillna(X.mean()) # Simple mean imputation for any remaining NaNs in X
    if X.isnull().any().any(): X.dropna(axis=0, how='any', inplace=True); y = y.loc[X.index] # Drop if mean impute fails
    if X.empty or y.empty: print("X or y empty after internal NaN handling."); return None, None, None, None, None
    print(f"Data shape post-NaN handling in prepare_ml_data: X={X.shape}, y={y.shape}")
    if len(X) < min_test_samples * 2: print(f"Not enough data ({len(X)} rows) for robust split."); return None, None, None, None, None
    n_samples = len(X); test_size_abs = max(min_test_samples, int(n_samples * test_split_size))
    if n_samples - test_size_abs < min_test_samples : test_size_abs = n_samples - min_test_samples
    if test_size_abs < 1: print(f"Cannot make meaningful split."); return None,None,None,None,None
    train_size = n_samples - test_size_abs
    X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]; y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]
    if X_train.empty or y_train.empty or X_test.empty or y_test.empty: print("Train/Test set empty after split."); return None, None, None, None, None
    print(f"Train shapes: X_train={X_train.shape}, y_train={y_train.shape}; Test shapes: X_test={X_test.shape}, y_test={y_test.shape}")
    numeric_cols_xtrain = X_train.select_dtypes(include=np.number).columns
    if not numeric_cols_xtrain.empty:
        scaler = StandardScaler(); X_train_scaled_np = scaler.fit_transform(X_train[numeric_cols_xtrain])
        X_train_scaled = pd.DataFrame(X_train_scaled_np, columns=numeric_cols_xtrain, index=X_train.index)
        for col in X_train.columns.difference(numeric_cols_xtrain): X_train_scaled[col] = X_train[col]
        numeric_cols_xtest = X_test.select_dtypes(include=np.number).columns
        cols_to_scale_in_test = [col for col in numeric_cols_xtrain if col in numeric_cols_xtest]
        if cols_to_scale_in_test:
            X_test_scaled_np = scaler.transform(X_test[cols_to_scale_in_test])
            X_test_scaled = pd.DataFrame(X_test_scaled_np, columns=cols_to_scale_in_test, index=X_test[cols_to_scale_in_test].index)
            for col in X_test.columns.difference(cols_to_scale_in_test): X_test_scaled[col] = X_test[col]
        else: X_test_scaled = X_test.copy()
        return X_train_scaled, X_test_scaled, y_train, y_test, scaler
    else: print("No numeric columns in X_train for scaling."); return X_train, X_test, y_train, y_test, None

def lgbm_objective(trial, X_train, y_train, X_val, y_val, base_params):
    params = {
        'objective': base_params['objective'], 'metric': base_params['metric'], 'num_class': base_params.get('num_class'),
        'boosting_type': 'gbdt', 'n_estimators': trial.suggest_int('n_estimators', 100, 1000, step=50),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1, log=True),
        'num_leaves': trial.suggest_int('num_leaves', 10, 50), 'max_depth': trial.suggest_int('max_depth', -1, 10),
        'min_child_samples': trial.suggest_int('min_child_samples', 5, 50),
        'feature_fraction': trial.suggest_float('feature_fraction', 0.5, 1.0),
        'bagging_fraction': trial.suggest_float('bagging_fraction', 0.5, 1.0),
        'bagging_freq': trial.suggest_int('bagging_freq', 0, 7),
        'reg_alpha': trial.suggest_float('reg_alpha', 1e-3, 1.0, log=True),
        'reg_lambda': trial.suggest_float('reg_lambda', 1e-3, 1.0, log=True),
        'random_state': 42, 'verbose': -1, 'n_jobs': -1, 'class_weight': base_params.get('class_weight', 'balanced')
    }
    if params['num_class'] is None: del params['num_class']
    model = lgb.LGBMClassifier(**params)
    model.fit(X_train, y_train, eval_set=[(X_val, y_val)], eval_metric=base_params['metric'], callbacks=[lgb.early_stopping(30, verbose=False)])
    y_proba_val = model.predict_proba(X_val)
    return log_loss(y_val, y_proba_val) # Minimize log_loss

def optimize_lgbm_hyperparameters(X_train, y_train, base_params, n_trials=50, validation_ratio=0.2):
    if not optuna_available: print("Optuna not available. Skipping HPO."); return optimized_lightgbm_params()
    print(f"Optimizing LightGBM hyperparameters with Optuna ({n_trials} trials)...")
    # Ensure enough samples for split
    if len(X_train) * validation_ratio < 1 or len(X_train) * (1-validation_ratio) <1 :
        print("Not enough samples in X_train for Optuna validation split. Using default params.")
        return optimized_lightgbm_params()

    # Using iloc for splitting after ensuring indices are reset if X_train is not using default integer index
    X_train_reset = X_train.reset_index(drop=True)
    y_train_reset = y_train.reset_index(drop=True)
    
    # Calculate split point based on indices
    split_idx = int(len(X_train_reset) * (1 - validation_ratio))
    if split_idx == 0 or split_idx == len(X_train_reset): # Avoid empty train or val set
        print("Validation split for Optuna resulted in empty train or validation set. Using default params.")
        return optimized_lightgbm_params()

    X_train_opt, X_val_opt = X_train_reset.iloc[:split_idx], X_train_reset.iloc[split_idx:]
    y_train_opt, y_val_opt = y_train_reset.iloc[:split_idx], y_train_reset.iloc[split_idx:]

    study = optuna.create_study(direction='minimize')
    study.optimize(lambda trial: lgbm_objective(trial, X_train_opt, y_train_opt, X_val_opt, y_val_opt, base_params), n_trials=n_trials,
                   callbacks=[lambda st, tr: print(f"Optuna Trial {tr.number} finished: value={tr.value:.4f}") if tr.number % 10 == 0 else None])
    print(f"Best trial for LightGBM: Value={study.best_value}, Params={study.best_params}")
    final_params = {**base_params, **study.best_params}
    return final_params

def train_lightgbm_model(X_train, y_train, X_test, y_test, optimized_params=None):
    print("Training LightGBM model...")
    if X_train.empty or y_train.empty: print("X_train or y_train empty. Skipping LGBM."); return None, None
    y_train_squeezed = y_train.squeeze(); y_test_squeezed = y_test.squeeze()
    unique_labels_train = sorted(y_train_squeezed.unique()); num_classes = len(unique_labels_train)
    if num_classes <= 1: print(f"Only {num_classes} class(es) in y_train. Skipping LGBM."); return None, None
    base_params = {'objective': 'multiclass' if num_classes > 2 else 'binary', 'metric': 'multi_logloss' if num_classes > 2 else 'binary_logloss', 'random_state': 42, 'verbose': -1, 'n_jobs': -1, 'class_weight': 'balanced'}
    if base_params['objective'] == 'multiclass': base_params['num_class'] = num_classes
    
    params_to_use = {**base_params, **(optimized_params if optimized_params else optimized_lightgbm_params())} # Ensure optimized_params is dict
    if not optimized_params: print("Using default (or non-Optuna optimized) LightGBM parameters.")
    else: print("Using Optuna-optimized (or provided) LightGBM parameters.")

    label_map = {label: i for i, label in enumerate(unique_labels_train)}
    y_train_mapped = y_train_squeezed.map(label_map); y_test_mapped = y_test_squeezed.map(label_map).fillna(-1).astype(int)
    model = lgb.LGBMClassifier(**params_to_use)
    valid_eval_indices = y_test_mapped != -1
    eval_set_data = (X_test.loc[valid_eval_indices], y_test_mapped[valid_eval_indices]) if valid_eval_indices.any() else None
    if eval_set_data: model.fit(X_train, y_train_mapped, eval_set=[eval_set_data], eval_metric=params_to_use.get('metric', 'logloss'), callbacks=[lgb.early_stopping(30, verbose=False)])
    else: print("Warning: No valid eval set. Fitting on full training data."); model.fit(X_train, y_train_mapped)
    if not valid_eval_indices.any(): print("No valid test samples for evaluation."); return model, None
    y_pred_mapped = model.predict(X_test.loc[valid_eval_indices]); y_proba = model.predict_proba(X_test.loc[valid_eval_indices]) # Predict only on valid indices for consistency
    acc = accuracy_score(y_test_mapped[valid_eval_indices], y_pred_mapped) # y_pred_mapped already subsetted
    print(f"\nðŸŽ¯ Accuracy on mapped test data: {acc:.4f}")
    if params_to_use['objective'] == 'binary' and y_proba.shape[1] == 2:
        try: auc = roc_auc_score(y_test_mapped[valid_eval_indices], y_proba[:, 1]); print(f"ðŸ“Š AUC: {auc:.4f}")
        except ValueError as e_auc: print(f"AUC Calc Error: {e_auc}")
    print("\nClassification Report (on mapped and valid test labels):")
    try: print(classification_report(y_test_mapped[valid_eval_indices], y_pred_mapped, labels=sorted(y_train_mapped.unique()), zero_division=0))
    except Exception as e_cr: print(f"Classification Report Error: {e_cr}")
    feat_imp_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': model.feature_importances_}).sort_values(by='Importance', ascending=False)
    print("\nTop 10 features:\n", feat_imp_df.head(10)); return model, feat_imp_df

def plot_feature_importance(feature_importance_df, top_n=20, symbol_for_plot=""):
    if feature_importance_df is None or feature_importance_df.empty: print("No feature importance to plot."); return
    plt.figure(figsize=(12, max(6, min(top_n, len(feature_importance_df)) * 0.45)))
    sns.barplot(x='Importance', y='Feature', data=feature_importance_df.head(top_n), palette="viridis")
    plt.title(f'Top {top_n} Feature Importances for {symbol_for_plot} (LightGBM)', fontsize=16); plt.tight_layout(); plt.show()

def export_lgbm_to_onnx(lgbm_model, X_sample_df, file_path="lgbm_model.onnx", target_opset=12):
    print(f"\nExporting LGBM model to ONNX: {file_path} (opset={target_opset})")
    # Corrected check for ONNX libraries
    if not all([onnx_available, skl2onnx_available, onnxmltools_available, (FloatTensorType is not None)]):
        print("One or more ONNX libraries (onnx, skl2onnx, onnxmltools) or FloatTensorType not available. Skipping ONNX export."); return None
    if lgbm_model is None or X_sample_df is None or X_sample_df.empty: print("Model or sample data empty for ONNX. Skipping."); return None
    try:
        initial_type = [('float_input', FloatTensorType([None, X_sample_df.shape[1]]))]
        converted_model = onnxmltools.convert_lightgbm(lgbm_model, initial_types=initial_type, target_opset=target_opset)
        with open(file_path, "wb") as f: f.write(converted_model.SerializeToString())
        print(f"Model exported to ONNX: {file_path}"); onnx.checker.check_model(file_path); print("ONNX model check OK."); return file_path
    except Exception as e:
        print(f"Error exporting LGBM to ONNX: {e}. Fallback to pickle.")
        try: 
            import pickle
            pkl_path = file_path.replace('.onnx', '.pkl')
            with open(pkl_path, 'wb') as pf: pickle.dump(lgbm_model, pf)
            print(f"Model saved as pickle: {pkl_path}")
            return pkl_path
        except Exception as ep: print(f"Pickle save error: {ep}"); return None

def simple_feature_selection_fallback(X_train, y_train, max_features=20):
    print("Using simple variance-based feature selection fallback...")
    if X_train.empty: return pd.DataFrame()
    X_train_numeric = X_train.select_dtypes(include=np.number)
    if X_train_numeric.empty: print("No numeric features for variance selection."); selected_features = X_train.columns[:max_features].tolist(); return pd.DataFrame({'Feature': selected_features})
    variance_scores = X_train_numeric.var().sort_values(ascending=False)
    num_features_to_select = min(max_features, len(variance_scores))
    selected_features = variance_scores.head(num_features_to_select).index.tolist()
    return pd.DataFrame({'Feature': selected_features, 'Score': variance_scores.head(num_features_to_select).values})

def prioritized_feature_selection(X_train, y_train, causal_ranking, max_features=25): # Increased max_features
    print("Prioritized feature selection: Causal Ranking + Mutual Information...")
    if X_train.empty or y_train.empty: return pd.DataFrame()
    top_causal_features = []; num_causal_to_select = min(len(causal_ranking) if causal_ranking else 0, max_features // 2)
    if causal_ranking:
        for feat, score in causal_ranking[:num_causal_to_select]:
            if feat in X_train.columns: top_causal_features.append(feat)
    remaining_slots = max_features - len(top_causal_features); stat_selected_features = []
    if remaining_slots > 0:
        features_for_stat_selection = [f for f in X_train.columns if f not in top_causal_features]
        if features_for_stat_selection:
            X_remaining_for_stat = X_train[features_for_stat_selection]; y_train_squeezed = y_train.squeeze()
            if y_train_squeezed.nunique() > 1 and not X_remaining_for_stat.empty:
                X_remaining_numeric = X_remaining_for_stat.select_dtypes(include=np.number)
                if not X_remaining_numeric.empty:
                    num_stat_to_select = min(remaining_slots, X_remaining_numeric.shape[1])
                    try:
                        selector_mi = SelectKBest(mutual_info_classif, k=num_stat_to_select); selector_mi.fit(X_remaining_numeric, y_train_squeezed)
                        stat_selected_features = X_remaining_numeric.columns[selector_mi.get_support()].tolist()
                    except Exception as e_mi: print(f"Error in mutual_info_classif: {e_mi}.")
                else: print("No numeric features for MI selection.")
            else: print("Not enough target variance or no remaining features for MI.")
        else: print("No remaining features for MI selection.")
    final_selected_features = list(dict.fromkeys(top_causal_features + stat_selected_features))
    if not final_selected_features and not X_train.empty: print("No features from prioritized selection, using fallback."); return simple_feature_selection_fallback(X_train, y_train, max_features)
    print(f"Prioritized features: {len(final_selected_features)}. Top 5: {final_selected_features[:5]}"); return pd.DataFrame({'Feature': final_selected_features})

def optimized_lightgbm_params():
    return {'boosting_type': 'gbdt', 'num_leaves': 31, 'learning_rate': 0.05, 'n_estimators': 200, 'feature_fraction': 0.8, 'bagging_fraction': 0.8, 'bagging_freq': 5, 'min_child_samples': 20, 'reg_alpha': 0.01, 'reg_lambda': 0.01, 'random_state': 42, 'verbose': -1, 'n_jobs': -1, 'class_weight': 'balanced'}

def run_patchtst_foundation_forecast(symbol_name: str, historical_data_df: pd.DataFrame, prediction_length: int, context_length: int = 96, model_checkpoint: str = "ibm-research/patchtst-etth1-pretrain", patch_length: int = 16 ):
    print(f"\n---  Forecasting for {symbol_name} using PatchTST Foundation Model ({model_checkpoint}) ---")
    if not torch_available or 'PatchTSTForPrediction' not in globals():
        print("Torch or PatchTST from transformers not available. Skipping."); return {"status": "failed", "reason": "Torch/Transformers missing", "method": "PatchTST (Skipped)", "forecast": np.array([np.nan] * prediction_length)}
    if 'close' not in historical_data_df.columns:
        print(f"'close' column missing for {symbol_name}. Skipping."); return {"status": "failed", "reason": "'close' missing", "method": f"PatchTST ({model_checkpoint}) (Skipped)", "forecast": np.array([np.nan] * prediction_length)}
    close_prices = historical_data_df['close'].values.astype(np.float32)
    if len(close_prices) < context_length:
        print(f"Not enough data ({len(close_prices)}) for PatchTST context ({context_length}). Skipping."); return {"status": "failed", "reason": "Insufficient data", "method": f"PatchTST ({model_checkpoint}) (Skipped)", "forecast": np.array([np.nan] * prediction_length), "last_price": close_prices[-1] if len(close_prices) > 0 else np.nan}
    try:
        config = PatchTSTConfig.from_pretrained(model_checkpoint, prediction_length=prediction_length, context_length=context_length, patch_length=patch_length, num_input_channels=1) # Ensure univariate
        model = PatchTSTForPrediction.from_pretrained(model_checkpoint, config=config) # Pass config
        device = "cuda" if torch.cuda.is_available() else "cpu"; model.to(device)
        
        past_values_np = close_prices[-context_length:]
        past_values_tensor = torch.tensor(past_values_np, dtype=torch.float32).unsqueeze(0).to(device) # Shape: (1, context_length) for univariate
        # For models expecting (batch_size, num_channels, context_length) or (batch_size, context_length, num_channels)
        # PatchTST often expects (batch_size, context_length) for univariate, or (batch_size, num_features, context_length)
        # The generate function of PatchTSTForPrediction specifically asks for `past_values` of shape (batch_size, sequence_length) for univariate
        
        with torch.no_grad():
            generated_outputs = model.generate(past_values=past_values_tensor) # prediction_length is part of config now

        # Output shape from generate is often (batch_size, num_patches, patch_length) for PatchTST's own generation
        # Or (batch_size, prediction_length, num_variates) for prediction head
        # For PatchTSTForPrediction's generate, it should directly give (batch_size, prediction_length) for univariate.
        forecast_values = generated_outputs.sequences.cpu().numpy().squeeze() # .sequences is common for generate output

        if forecast_values.ndim == 0: forecast_values = np.array([forecast_values.item()] * prediction_length)
        elif forecast_values.shape[0] != prediction_length: # Check if length matches
             print(f"Warning: PatchTST generated {forecast_values.shape[0]} steps, expected {prediction_length}. Adjusting/Padding...")
             if forecast_values.shape[0] < prediction_length:
                forecast_values = np.pad(forecast_values, (0, prediction_length - forecast_values.shape[0]), 'edge')
             else: # Too many steps
                forecast_values = forecast_values[:prediction_length]
        
        last_actual_price = close_prices[-1]; avg_prediction = np.nanmean(forecast_values)
        direction = "N/A"; magnitude = float('nan'); confidence_label = "N/A"; forecast_std = np.nanstd(forecast_values)
        if not np.isnan(avg_prediction) and not np.isnan(last_actual_price):
            if avg_prediction > last_actual_price: direction = "ðŸ“ˆ UP"
            elif avg_prediction < last_actual_price: direction = "ðŸ“‰ DOWN"
            else: direction = "âž¡ï¸ FLAT"
            if last_actual_price != 0: magnitude = abs((avg_prediction - last_actual_price) / last_actual_price * 100)
            else: magnitude = float('inf') if avg_prediction != 0 else 0.0
        if np.isnan(forecast_std) or np.isnan(last_actual_price): confidence_label = "âšª Unknown (NaN values)"
        elif forecast_std == 0 and not np.isnan(last_actual_price): confidence_label = "ðŸŸ¡ Medium (Flat Forecast)"
        elif abs(last_actual_price) > 1e-9 and (forecast_std / abs(last_actual_price)) < 0.01 : confidence_label = "ðŸŸ¢ High (Low Rel Std)"
        elif abs(last_actual_price) > 1e-9 and (forecast_std / abs(last_actual_price)) < 0.05 : confidence_label = "ðŸŸ¡ Medium (Mod Rel Std)"
        else: confidence_label = "ðŸ”´ Low (High Rel Std)"
        print(f"âœ… PatchTST Forecast: {forecast_values}"); print(f"ðŸ“Š Analysis: {direction} {magnitude:.1f}% - Confidence: {confidence_label} (Forecast_Std: {forecast_std:.4f})")
        return {"status": "Completed", "forecast": forecast_values, "method": f"PatchTST ({model_checkpoint})", "last_price": last_actual_price, "avg_forecast": avg_prediction, "direction": direction, "magnitude": magnitude, "confidence_metric_value": forecast_std, "confidence_label": confidence_label, "advanced_analysis": True}
    except Exception as e:
        print(f"Error PatchTST forecast for {symbol_name}: {e}"); error_trace = traceback.format_exc(); print(error_trace)
        return {"status": "failed", "reason": str(e), "trace": error_trace, "method": f"PatchTST ({model_checkpoint}) (Error)", "forecast": np.array([np.nan] * prediction_length), "last_price": historical_data_df['close'].iloc[-1] if not historical_data_df.empty and 'close' in historical_data_df else np.nan}

# --- MAIN WORKFLOW FUNCTION ---
def run_full_workflow(symbol=DEFAULT_SYMBOL, start_date_str=START_DATE, end_date_str=END_DATE, api_key_val=API_KEY, force_train_autoformer=False, run_optuna_lgbm=False, use_foundation_model=True):
    print(f"\n{'='*40}\nðŸš€ ENHANCED WORKFLOW FOR: {symbol}\n{'='*40}")
    default_return = {"symbol": symbol, "status": "Workflow Started", "raw_data_shape": (0,0), "featured_data_shape": (0,0), "X_train_shape": (0,0), "X_test_shape": (0,0), "selected_features_count": 0, "selected_feature_names": [], "scaler_object": None, "ml_model_object": None, "lgbm_feature_importance": None, "causal_model_object": None, "causal_feature_ranking": [], "onnx_model_path": None, "forecasting_results": None, "lgbm_optimized_params": None }
    df_raw = fetch_twelve_data(symbol, api_key_val, start_date_str=start_date_str, end_date_str=end_date_str)
    if df_raw is None or df_raw.empty: default_return["status"] = "Data Fetching Failed"; return default_return
    default_return["raw_data_shape"] = df_raw.shape
    price_c, high_c, low_c, vol_c, target_col = 'close', 'high', 'low', 'volume', 'target'
    print(f"\n--- ðŸ”§ Feature Engineering: {symbol} ---")
    df_f = df_raw.copy()
    df_f = add_technical_indicators(df_f); df_f = add_optimized_features(df_f, price_col=price_c, volume_col=vol_c)
    df_f = add_wavelet_features(df_f, column=price_c); df_f = add_entropy_features(df_f, column=price_c, window=40)
    df_f = add_advanced_technical_features(df_f, price_col=price_c, high_col=high_c, low_col=low_c, volume_col=vol_c)
    df_f = add_transformer_features_conceptual(df_f, column=price_c, sequence_length=20)
    if 'RSI_14' in df_f.columns and 'ADX_14' in df_f.columns: df_f['RSI_ADX_interaction'] = df_f['RSI_14'] * df_f['ADX_14'] / 100.0
    if 'ATR_14' in df_f.columns and 'volatility_20' in df_f.columns: volatility_safe = df_f['volatility_20'].replace(0, np.nan); df_f['ATR_vol_ratio'] = df_f['ATR_14'] / volatility_safe
    default_return["featured_data_shape"] = df_f.shape
    print(f"\n--- Data Cleaning (Inf/NaN Handling & Imputation): {symbol} ---")
    df_f.replace([np.inf, -np.inf], np.nan, inplace=True); numeric_cols_to_impute = df_f.select_dtypes(include=np.number).columns
    if not numeric_cols_to_impute.empty:
        nan_counts_before = df_f[numeric_cols_to_impute].isnull().sum().sum()
        if nan_counts_before > 0: print(f"NaNs before imputation: {nan_counts_before}"); df_f[numeric_cols_to_impute] = df_f[numeric_cols_to_impute].interpolate(method='linear', limit_direction='both', axis=0).fillna(method='bfill').fillna(method='ffill').fillna(0); print(f"NaNs after imputation: {df_f[numeric_cols_to_impute].isnull().sum().sum()}")
        else: print("No NaNs in numeric features to impute.")
    else: print("No numeric columns for imputation.")
    print(f"\n--- ðŸ“Š Regime Detection: {symbol} ---"); df_f = detect_regimes_simple(df_f, column=price_c) # Or add_clasp_regimes
    print(f"\n--- ðŸŽ¯ Target Definition: {symbol} ---"); df_f = balanced_target_definition(df_f, column=price_c, periods=5)
    print(f"\n--- Causal Discovery & Ranking: {symbol} ---"); causal_model_obj = None; causal_feat_ranking = []
    if df_f is not None and not df_f.empty and target_col in df_f.columns and df_f[target_col].nunique(dropna=True) > 1:
        causal_model_obj = discover_causal_structure(df_f.copy(), target_col=target_col, symbol=symbol)
        if causal_model_obj: causal_feat_ranking = causal_feature_selection(df_f.copy(), target_col=target_col)
    else: print("Skipping Causal Discovery.")
    default_return["causal_model_object"] = causal_model_obj; default_return["causal_feature_ranking"] = causal_feat_ranking if causal_feat_ranking else []
    print(f"\n--- ML Preparation & Feature Selection: {symbol} ---"); X_tr, X_te, y_tr, y_te, scaler_obj = None, None, None, None, None; sel_feat_names = []; current_status_ml = "ML Prep Incomplete"
    if df_f is None or df_f.empty or target_col not in df_f.columns or df_f[target_col].isnull().all(): current_status_ml = "ML Prep Failed - Data/Target issue"
    elif df_f[target_col].nunique(dropna=True) <= 1: current_status_ml = f"ML Prep Skipped - Target has <=1 unique non-NaN values"
    else:
        X_tr_p, X_te_p, y_tr_p, y_te_p, scaler_obj_p = prepare_ml_data(df_f.copy(), target_col=target_col)
        default_return["scaler_object"] = scaler_obj_p
        if X_tr_p is not None and not X_tr_p.empty and y_tr_p is not None and not y_tr_p.empty and y_tr_p.nunique() > 1:
            scores_df = prioritized_feature_selection(X_tr_p.copy(), y_tr_p.copy(), causal_feat_ranking, max_features=25)
            sel_feat_names_intermediate = scores_df['Feature'].tolist() if scores_df is not None and not scores_df.empty else X_tr_p.columns.tolist()[:25]
            X_tr_sel = X_tr_p[[c for c in sel_feat_names_intermediate if c in X_tr_p.columns]]; X_te_sel = X_te_p[[c for c in sel_feat_names_intermediate if c in X_te_p.columns]]
            if not X_tr_sel.empty:
                numeric_cols_xtr_sel = X_tr_sel.select_dtypes(include=np.number).columns
                if not numeric_cols_xtr_sel.empty:
                    variance_selector = VarianceThreshold(threshold=0.001); variance_selector.fit(X_tr_sel[numeric_cols_xtr_sel])
                    sel_feat_names_vt = X_tr_sel[numeric_cols_xtr_sel].columns[variance_selector.get_support()].tolist()
                    sel_feat_names = sel_feat_names_vt if sel_feat_names_vt else sel_feat_names_intermediate
                    non_numeric_sel_intermed = [f for f in sel_feat_names_intermediate if f not in numeric_cols_xtr_sel and f in X_tr_p.columns]
                    sel_feat_names = list(dict.fromkeys(sel_feat_names + non_numeric_sel_intermed))
                else: sel_feat_names = X_tr_sel.columns.tolist()
                X_tr = X_tr_p[sel_feat_names].copy(); X_te = X_te_p[sel_feat_names].copy(); y_tr, y_te = y_tr_p, y_te_p
                if not X_tr.empty and scaler_obj_p is not None:
                    final_numeric_cols_tr = X_tr.select_dtypes(include=np.number).columns
                    if not final_numeric_cols_tr.empty:
                        final_scaler = StandardScaler(); X_tr[final_numeric_cols_tr] = final_scaler.fit_transform(X_tr[final_numeric_cols_tr])
                        default_return["scaler_object"] = final_scaler
                        if not X_te.empty: final_numeric_cols_te = [c for c in final_numeric_cols_tr if c in X_te.columns]; X_te[final_numeric_cols_te] = final_scaler.transform(X_te[final_numeric_cols_te]) if final_numeric_cols_te else None
                    current_status_ml = "ML Prep Completed"
                elif X_tr.empty: current_status_ml = "ML Prep Failed - X_tr empty post-selection"
                else: current_status_ml = "ML Prep Completed (No re-scale post-selection)"
            else: current_status_ml = "ML Prep Failed - X_tr_sel empty"
        else: current_status_ml = "ML Prep Failed - Initial data prep insufficient"
    default_return["status"] = current_status_ml; default_return["X_train_shape"] = X_tr.shape if X_tr is not None else (0,0); default_return["X_test_shape"] = X_te.shape if X_te is not None else (0,0); default_return["selected_feature_names"] = sel_feat_names; default_return["selected_features_count"] = len(sel_feat_names)
    lgbm_model_obj, lgbm_feat_imp, onnx_file_path, lgbm_final_params = None, None, None, None
    if current_status_ml == "ML Prep Completed" and X_tr is not None and not X_tr.empty and y_tr is not None and not y_tr.empty and y_tr.nunique() > 1:
        print(f"\n--- Model Training & Eval: {symbol} ---")
        try:
            if run_optuna_lgbm:
                unique_labels_tr = sorted(y_tr.squeeze().unique()); num_classes_tr = len(unique_labels_tr)
                optuna_base_params = {'objective': 'multiclass' if num_classes_tr > 2 else 'binary', 'metric': 'multi_logloss' if num_classes_tr > 2 else 'binary_logloss', 'num_class': num_classes_tr if num_classes_tr > 2 else None, 'class_weight': 'balanced'}
                lgbm_final_params = optimize_lgbm_hyperparameters(X_tr, y_tr.squeeze(), optuna_base_params, n_trials=30)
            else: lgbm_final_params = optimized_lightgbm_params()
            default_return["lgbm_optimized_params"] = lgbm_final_params
            lgbm_model_obj, lgbm_feat_imp = train_lightgbm_model(X_tr, y_tr, X_te, y_te, optimized_params=lgbm_final_params)
            if lgbm_model_obj:
                default_return["status"] = "Completed"; default_return["ml_model_object"] = lgbm_model_obj; default_return["lgbm_feature_importance"] = lgbm_feat_imp
                if lgbm_feat_imp is not None and not lgbm_feat_imp.empty: plot_feature_importance(lgbm_feat_imp, top_n=20, symbol_for_plot=symbol)
                if X_te is not None and not X_te.empty and not X_te.iloc[[0]].empty:
                    onnx_sample_data = X_te.iloc[[0]][sel_feat_names] if sel_feat_names and all(f in X_te.columns for f in sel_feat_names) else X_te.iloc[[0]]
                    onnx_file_path = export_lgbm_to_onnx(lgbm_model_obj, onnx_sample_data, file_path=f"{symbol}_lgbm_model.onnx")
                    default_return["onnx_model_path"] = onnx_file_path
            else: default_return["status"] = "Completed (LGBM Training Failed/Skipped)"
        except Exception as e_train: print(f"LGBM Training error: {e_train}\n{traceback.format_exc()}"); default_return["status"] = "Completed (LGBM Training Error)"
    else: print(f"âš ï¸ Skipping LGBM Training for {symbol}: {current_status_ml} or invalid data."); default_return["status"] = current_status_ml if "Failed" in current_status_ml or "Skipped" in current_status_ml else "ML Training Skipped"

    if use_foundation_model:
        print(f"\n--- Foundation Model Forecasting (PatchTST): {symbol} ---")
        foundation_forecast_output = run_patchtst_foundation_forecast(symbol_name=symbol, historical_data_df=df_raw, prediction_length=5, context_length=96, model_checkpoint="ibm-research/patchtst-etth1-pretrain", patch_length=16 )
        default_return["forecasting_results"] = foundation_forecast_output
    else:
        print(f"\n--- Autoformer Model Initialization & Forecasting: {symbol} ---")
        autoformer_predictor_instance = train_or_load_autoformer(df_raw['close'], force_train=force_train_autoformer)
        if autoformer_predictor_instance:
            forecast_output = ultimate_forecasting_workflow(symbol, df_raw.copy(), autoformer_predictor_instance, prediction_length=5) # Ensure this function exists and is compatible
            default_return["forecasting_results"] = forecast_output
        else: print("Custom Autoformer instance unavailable. Skipping forecasting."); default_return["forecasting_results"] = {"status": "failed", "reason": "Custom Autoformer instance unavailable"}
    print(f"\n{'='*40}\nWORKFLOW FOR {symbol} CONCLUDED (Final Status: {default_return['status']})\n{'='*40}")
    return default_return

# --- MAIN EXECUTION BLOCK ---
if __name__ == "__main__":
    print("ðŸš€ Starting ENHANCED ML Trading System Workflow...")
    if API_KEY == "YOUR_API_KEY_HERE" or not API_KEY: print("="*30 + " IMPORTANT: SET YOUR TWELVE_DATA_API_KEY " + "="*30)
    stocks_to_analyze = ["MSFT", "AAPL"]
    all_run_results = {}
    FORCE_TRAIN_AUTOFORMER_FLAG = False
    RUN_OPTUNA_FOR_LGBM_FLAG = False # Set True for HPO
    USE_FOUNDATION_MODEL_FLAG = True # Set True to use PatchTST, False for custom Autoformer (if ultimate_forecasting_workflow is defined for it)

    for i, stock_symbol in enumerate(stocks_to_analyze):
        print(f"\nðŸŽ¯ Processing {stock_symbol} ({i+1}/{len(stocks_to_analyze)})...")
        try:
            stock_run_data = run_full_workflow(symbol=stock_symbol, api_key_val=API_KEY, force_train_autoformer=FORCE_TRAIN_AUTOFORMER_FLAG, run_optuna_lgbm=RUN_OPTUNA_FOR_LGBM_FLAG, use_foundation_model=USE_FOUNDATION_MODEL_FLAG)
            all_run_results[stock_symbol] = stock_run_data
        except Exception as e_main_loop: print(f"FATAL ERROR processing {stock_symbol}: {e_main_loop}\n{traceback.format_exc()}"); all_run_results[stock_symbol] = {"symbol": stock_symbol, "status": f"Workflow Crashed: {e_main_loop}"}
        if i < len(stocks_to_analyze) - 1: print(f"Pausing..."); time.sleep(5) # Reduced sleep
    print("\n\n--- ðŸ“ˆ DETAILED RESULTS SUMMARY ---")
    for symbol_res_key, res_item in all_run_results.items():
        print(f"\n--- Results for: {symbol_res_key} ---")
        if res_item and isinstance(res_item, dict):
            print(f"  Overall Status: {res_item.get('status', 'N/A')}")
            # ... (rest of summary print logic from previous version) ...
            forecast_details = res_item.get('forecasting_results')
            if forecast_details and isinstance(forecast_details, dict) and forecast_details.get('status', '').lower() == 'completed':
                print(f"  Forecasting Method: {forecast_details.get('method', 'N/A')}")
                print(f"  Forecast Values (next {len(forecast_details.get('forecast', []))} days): {forecast_details.get('forecast', 'N/A')}")
                print(f"  Predicted Direction: {forecast_details.get('direction', 'N/A')}")
                mag = forecast_details.get('magnitude'); print(f"  Predicted Magnitude: {mag:.1f}%" if isinstance(mag, (int, float)) else f"N/A ({mag})")
                conf_val = forecast_details.get('confidence_metric_value'); conf_lab = forecast_details.get('confidence_label', 'N/A')
                print(f"  Forecast Confidence: {conf_lab} (Metric Val: {conf_val:.4f})" if isinstance(conf_val, (int,float)) else f"N/A ({conf_lab})")
            elif forecast_details: print(f"  Forecasting Status: Failed ({forecast_details.get('reason', 'Unknown')})")
            else: print("  Forecasting Data: N/A")
        else: print(f"  No valid results for {symbol_res_key}")
    print(f"\nðŸŽ‰ Enhanced Analysis Run Complete!")
    print("\n" + "="*60 + "\nðŸ’¡ Recommendation for Further Forecasting Enhancement...\n" + "="*60 + "\n1. Fully implement/train custom Autoformer OR ensure PatchTST (or other foundation model) is fine-tuned for your specific financial data.\n2. Explore other pre-trained models from Hugging Face or specialized forecasting libraries.\n" + "="*60)
```
