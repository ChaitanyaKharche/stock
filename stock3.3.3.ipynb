```python
import traceback
import os
import time
import requests
import numpy as np
import pandas as pd
import pandas_ta
import yfinance as yf
from datetime import datetime, timedelta
import pywt
import antropy as ant
import lightgbm as lgb
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import accuracy_score, classification_report, log_loss, roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.feature_selection import SelectKBest, mutual_info_classif, f_classif, VarianceThreshold
from tsfm_public.toolkit.dataset import ForecastDFDataset
import gc




# Foundational Model Imports
from transformers import PatchTSTConfig, PatchTSTForPrediction, Trainer, TrainingArguments, EarlyStoppingCallback, set_seed

# --- Global Flags and Initializations for Optional Libraries ---
optuna_available = False
torch_available = False
sktime_available = False
ClaSPSegmentation = None
dowhy_available = False
CausalModel = None
nx = None
# ONNX related flags and types
onnx_available = False
skl2onnx_available = False
onnxmltools_available = False
FloatTensorType = None # Crucial to initialize globally for the check later
SKIP_CAUSAL_ANALYSIS_FOR_DEBUGGING = True

if hasattr(pd.DataFrame, 'ta') is False and pandas_ta is not None:
    try:
        pd.DataFrame.ta = pandas_ta.Core(df=None)
        print("pandas_ta DataFrame accessor registered globally.")
    except Exception as e:
        print(f"Could not globally register pandas_ta accessor: {e}")

try:
    import optuna
    optuna_available = True
    print("Optuna imported successfully.")
except ImportError:
    print("Optuna not found. LightGBM hyperparameter optimization with Optuna will be skipped.")

try:
    import torch
    import torch.nn as nn
    import torch.optim as optim
    torch_available = True
    print("PyTorch imported successfully.")
    if torch.cuda.is_available():
        print(f"PyTorch CUDA available: True, Version: {torch.version.cuda}")
        print(f"Using PyTorch on GPU: {torch.cuda.get_device_name(0)}")
    else:
        print("PyTorch CUDA available: False.")
    print("Neural ODE features are DISABLED by configuration (per original script).")
except ImportError:
    print("PyTorch not found. Custom Transformer and Foundation Model features will be SKIPPED.")

try:
    import sktime
    from sktime.annotation.clasp import ClaSPSegmentation
    sktime_available = True
    print(f"Sktime {sktime.__version__} imported successfully (ClaSPSegmentation ready).")
except ImportError as e_sktime_import:
    print(f"Sktime modules import failed: {e_sktime_import}. ClaSP Regime detection will be skipped.")
except Exception as e_sktime_general:
    print(f"A general error occurred during sktime import or setup: {e_sktime_general}")

try:
    import dowhy
    from dowhy import CausalModel
    import networkx as nx
    dowhy_available = True
    print(f"DoWhy {dowhy.__version__} and NetworkX {nx.__version__} imported successfully.")
except ImportError:
    print("DoWhy or NetworkX not found. Causal Discovery will be skipped.")

try:
    import onnx
    onnx_available = True
    import onnxruntime as ort # ort is imported but not directly used in a flag
    import skl2onnx
    skl2onnx_available = True # Set flag here
    from skl2onnx.common.data_types import FloatTensorType # Assigns to global FloatTensorType
    import onnxmltools
    onnxmltools_available = True # Set flag here
    print("ONNX, ONNXRuntime, skl2onnx, and onnxmltools imported successfully.")
    if hasattr(onnxmltools, '__version__'):
          print(f"Onnxmltools version: {onnxmltools.__version__}")
except ImportError as e:
    # If any import fails, the respective *_available flag will remain False from its global init
    print(f"One or more ONNX components not found: {e}. ONNX features will be skipped.")

print("\nAll libraries and modules conditional imports attempted.")

# --- Constants ---
TWELVE_DATA_API_KEY = "b6dbb92e551a46f2b20de27540aeef0a" # Replace with your actual key
API_KEY = TWELVE_DATA_API_KEY
DEFAULT_SYMBOL = "MSFT"
START_DATE = (datetime.now() - timedelta(days=3*365)).strftime('%Y-%m-%d')
END_DATE = datetime.now().strftime('%Y-%m-%d')
AUTOFORMER_MODEL_PATH = "autoformer_predictor_trained.pth"

# --- Function Definitions (AutoformerPredictor and its helpers) ---
class AutoformerPredictor:
    def __init__(self, input_len=60, pred_len=5, d_model=64, n_heads=8, num_encoder_layers=2, num_decoder_layers=1, dim_feedforward=256):
        self.device = 'cuda' if torch.cuda.is_available() and torch_available else 'cpu'
        self.model = None
        self.input_len = input_len
        self.pred_len = pred_len
        self.d_model = d_model

        if torch_available:
            print(f"üîß Initializing Custom AutoformerPredictor on device: {self.device}")
            self.model = nn.Transformer(
                d_model=d_model, nhead=n_heads, num_encoder_layers=num_encoder_layers,
                num_decoder_layers=num_decoder_layers, dim_feedforward=dim_feedforward,
                activation='gelu', batch_first=False
            ).to(self.device)
            self.enc_embedding = nn.Linear(1, d_model).to(self.device)
            self.dec_embedding = nn.Linear(1, d_model).to(self.device)
            self.projection = nn.Linear(d_model, 1).to(self.device)
        else:
            print("PyTorch not available. AutoformerPredictor cannot be initialized.")

    def _generate_square_subsequent_mask(self, sz):
        mask = (torch.triu(torch.ones(sz, sz)) == 1).transpose(0, 1)
        mask = mask.float().masked_fill(mask == 0, float('-inf')).masked_fill(mask == 1, float(0.0))
        return mask.to(self.device)

    def forward(self, src, tgt):
        if not torch_available or self.model is None: return None
        src_embedded = self.enc_embedding(src)
        tgt_embedded = self.dec_embedding(tgt)
        tgt_mask = self._generate_square_subsequent_mask(tgt.size(0))
        output = self.model(src_embedded, tgt_embedded, tgt_mask=tgt_mask)
        output = self.projection(output)
        return output

    def train_model(self, train_loader, epochs=10, learning_rate=0.001):
        if not torch_available or self.model is None: print("Cannot train: PyTorch or model not available."); return
        if not train_loader: print("Cannot train: No data loader provided."); return
        print(f"Starting AutoformerPredictor training for {epochs} epochs...")
        self.model.train()
        optimizer = optim.Adam(self.model.parameters(), lr=learning_rate)
        criterion = nn.MSELoss()
        for epoch in range(epochs):
            epoch_loss = 0
            for i, (batch_src, batch_tgt_input, batch_tgt_output) in enumerate(train_loader):
                batch_src = batch_src.float().to(self.device).permute(1,0,2)
                batch_tgt_input = batch_tgt_input.float().to(self.device).permute(1,0,2)
                batch_tgt_output = batch_tgt_output.float().to(self.device).permute(1,0,2)
                optimizer.zero_grad()
                predictions = self.forward(batch_src, batch_tgt_input)
                if predictions is None: continue
                loss = criterion(predictions, batch_tgt_output); loss.backward(); optimizer.step()
                epoch_loss += loss.item()
            avg_epoch_loss = epoch_loss / len(train_loader) if len(train_loader) > 0 else float('nan')
            print(f"Epoch {epoch+1}/{epochs}, Average Loss: {avg_epoch_loss:.4f}")
        print("Training finished."); self.save_model(AUTOFORMER_MODEL_PATH)

    def predict(self, series_data):
        if not torch_available or self.model is None: print("Cannot predict: PyTorch or model not available."); return np.array([np.nan] * self.pred_len)
        self.model.eval()
        with torch.no_grad():
            if isinstance(series_data, pd.Series): series_data = series_data.values
            if isinstance(series_data, list): series_data = np.array(series_data)
            if not isinstance(series_data, np.ndarray): print(f"Unsupported series data type for predict: {type(series_data)}"); return np.array([np.nan] * self.pred_len)
            if series_data.ndim > 1: series_data = series_data.squeeze()
            if series_data.ndim == 0: print(f"Series data is scalar, cannot process."); return np.array([np.nan]*self.pred_len)

            mean_val = series_data[-self.input_len:].mean() if len(series_data) >= self.input_len else series_data.mean()
            std_val = series_data[-self.input_len:].std() if len(series_data) >= self.input_len else series_data.std()
            std_val = std_val if std_val > 1e-8 else 1e-8
            if len(series_data) < self.input_len:
                padding_val = series_data[0] if len(series_data) > 0 else 0
                norm_series_np = (np.concatenate([np.full(self.input_len - len(series_data), padding_val), series_data]) - mean_val) / std_val
            else: norm_series_np = (series_data[-self.input_len:] - mean_val) / std_val
            src_tensor = torch.tensor(norm_series_np, dtype=torch.float32).unsqueeze(0).unsqueeze(-1).permute(1,0,2).to(self.device)
            current_tgt_sequence_normalized = torch.zeros((1, 1, 1), dtype=torch.float32, device=self.device)
            predictions_normalized_list = []
            for _ in range(self.pred_len):
                prediction_step_output = self.forward(src_tensor, current_tgt_sequence_normalized)
                if prediction_step_output is None: return np.array([np.nan] * self.pred_len)
                last_predicted_value_normalized = prediction_step_output[-1:, :, :]
                predictions_normalized_list.append(last_predicted_value_normalized.squeeze().cpu().item())
                current_tgt_sequence_normalized = torch.cat([current_tgt_sequence_normalized, last_predicted_value_normalized], dim=0)
            denorm_preds = (np.array(predictions_normalized_list) * std_val) + mean_val
            return denorm_preds

    def save_model(self, path):
        if not torch_available or self.model is None: return
        torch.save({'model_state_dict': self.model.state_dict(), 'enc_embedding_state_dict': self.enc_embedding.state_dict(), 'dec_embedding_state_dict': self.dec_embedding.state_dict(), 'projection_state_dict': self.projection.state_dict(), 'input_len': self.input_len, 'pred_len': self.pred_len, 'd_model': self.d_model}, path)
        print(f"AutoformerPredictor model saved to {path}")

    def load_model(self, path):
        if not torch_available: print("Cannot load model: PyTorch not available."); return False
        if not os.path.exists(path): print(f"Model file not found: {path}. New model needed."); return False
        checkpoint = torch.load(path, map_location=self.device)
        self.input_len = checkpoint.get('input_len', self.input_len); self.pred_len = checkpoint.get('pred_len', self.pred_len)
        saved_d_model = checkpoint.get('d_model', self.d_model)
        self.model = nn.Transformer(d_model=saved_d_model, nhead=8, num_encoder_layers=2, num_decoder_layers=1, dim_feedforward=256, activation='gelu', batch_first=False).to(self.device)
        self.enc_embedding = nn.Linear(1, saved_d_model).to(self.device); self.dec_embedding = nn.Linear(1, saved_d_model).to(self.device); self.projection = nn.Linear(saved_d_model, 1).to(self.device)
        self.model.load_state_dict(checkpoint['model_state_dict']); self.enc_embedding.load_state_dict(checkpoint['enc_embedding_state_dict']); self.dec_embedding.load_state_dict(checkpoint['dec_embedding_state_dict']); self.projection.load_state_dict(checkpoint['projection_state_dict'])
        self.model.eval(); print(f"AutoformerPredictor model loaded from {path}"); return True

def create_sequences(data, input_len, pred_len):
    X, y_input, y_output = [], [], []
    for i in range(len(data) - input_len - pred_len + 1):
        X.append(data[i:(i + input_len)])
        y_dec_inp_seq = np.concatenate(([data[i + input_len -1]], data[(i + input_len):(i + input_len + pred_len -1)]))
        y_input.append(y_dec_inp_seq)
        y_output.append(data[(i + input_len):(i + input_len + pred_len)])
    return np.array(X), np.array(y_input), np.array(y_output)

def train_or_load_autoformer(df_series, input_len=60, pred_len=5, force_train=False):
    if not torch_available: print("PyTorch not available for Autoformer."); return None
    autoformer_forecaster = AutoformerPredictor(input_len=input_len, pred_len=pred_len)
    if not force_train and os.path.exists(AUTOFORMER_MODEL_PATH):
        print(f"Loading Autoformer model from {AUTOFORMER_MODEL_PATH}")
        if autoformer_forecaster.load_model(AUTOFORMER_MODEL_PATH): return autoformer_forecaster
        else: print(f"Failed to load Autoformer. Training new one."); force_train = True
    if force_train:
        print("Training new AutoformerPredictor model...")
        scaler = StandardScaler(); series_scaled = scaler.fit_transform(df_series.values.reshape(-1, 1))
        X_seq, y_seq_dec_in, y_seq_out = create_sequences(series_scaled.flatten(), input_len, pred_len)
        if len(X_seq) == 0: print("Not enough data for Autoformer training. Model un-trained."); return autoformer_forecaster
        X_seq = X_seq.reshape(X_seq.shape[0], X_seq.shape[1], 1); y_seq_dec_in = y_seq_dec_in.reshape(y_seq_dec_in.shape[0], y_seq_dec_in.shape[1], 1); y_seq_out = y_seq_out.reshape(y_seq_out.shape[0], y_seq_out.shape[1], 1)
        dataset = torch.utils.data.TensorDataset(torch.from_numpy(X_seq).float(), torch.from_numpy(y_seq_dec_in).float(), torch.from_numpy(y_seq_out).float())
        train_loader = torch.utils.data.DataLoader(dataset, batch_size=16, shuffle=True)
        autoformer_forecaster.train_model(train_loader, epochs=20)
    else: print("Autoformer training skipped. Model un-trained if not loaded.")
    return autoformer_forecaster

# --- Data Fetching and Feature Engineering Functions (mostly unchanged, minor type hints/robustness) ---
def fetch_twelve_data(symbol: str, api_key: str, start_date_str: str = None, end_date_str: str = None) -> pd.DataFrame | None:
    base_url = "https://api.twelvedata.com/time_series"
    params = {"symbol": symbol, "interval": "1day", "apikey": api_key, "format": "JSON", "outputsize": 5000}
    if start_date_str: params["start_date"] = start_date_str
    if end_date_str: params["end_date"] = end_date_str
    print(f"Fetching data for {symbol} from Twelve Data (interval=1day, from {start_date_str} to {end_date_str})...")
    try:
        response = requests.get(base_url, params=params, timeout=30)
        response.raise_for_status()
        data = response.json()
    except requests.exceptions.RequestException as e: print(f"Request failed for {symbol}: {e}"); return None
    except ValueError as e: print(f"Failed to parse JSON for {symbol}: {e}. Response: {response.text[:200]}..."); return None
    if isinstance(data, dict) and (data.get("status") == "error" or "values" not in data): print(f"API Error for {symbol}: {data.get('message', 'Unknown error')}"); return None
    if not isinstance(data, dict) or "values" not in data or not data["values"]: print(f"No data values for {symbol}, or unexpected format."); return None
    df = pd.DataFrame(data["values"]).rename(columns={'datetime': 'date'})
    for col in ['open', 'high', 'low', 'close', 'volume']:
        if col in df.columns: df[col] = pd.to_numeric(df[col], errors='coerce').astype('float64')
        else:
            if col in ['open', 'high', 'low', 'close']: print(f"Critical column '{col}' missing."); return None
            df[col] = 0.0 # Default for volume
    if 'date' not in df.columns: print("Critical 'date' column missing."); return None
    df.index = pd.to_datetime(df['date']); df.drop(columns=['date'], inplace=True); df.sort_index(inplace=True)
    df.dropna(subset=[col for col in ['open', 'high', 'low', 'close'] if col in df.columns], inplace=True)
    if df.empty: print(f"No data remaining for {symbol}."); return None
    print(f"Successfully fetched/processed {len(df)} data points for {symbol}."); return df

def add_technical_indicators(df: pd.DataFrame) -> pd.DataFrame:
    df_feat = df.copy()
    if not hasattr(df_feat.ta, 'rsi'): print("pandas_ta not registered. Skipping TIs."); return df_feat
    print("Adding technical indicators...")
    try:
        for col in ['open', 'high', 'low', 'close', 'volume']: # Ensure float64
            if col in df_feat.columns: df_feat[col] = df_feat[col].astype('float64')
            else: # Add dummy if essential and missing (though fetch should handle this)
                if col in ['high', 'low', 'open', 'close'] and 'close' in df_feat: df_feat[col] = df_feat['close']
                elif col == 'volume': df_feat[col] = 0.0

        c, h, l, v = 'close', 'high', 'low', 'volume'
        required_cols_for_all_ta = [c,h,l,v] # Check these are present
        if not all(col in df_feat.columns for col in required_cols_for_all_ta):
            print("Warning: Not all OHLCV columns present, some TIs might fail or be inaccurate.")

        df_feat.ta.rsi(close=df_feat[c], length=14, append=True); df_feat.ta.rsi(close=df_feat[c], length=9, append=True, col_names='RSI_9'); df_feat.ta.rsi(close=df_feat[c], length=25, append=True, col_names='RSI_25')
        df_feat.ta.macd(close=df_feat[c], fast=12, slow=26, signal=9, append=True); df_feat.ta.macd(close=df_feat[c], fast=5, slow=15, signal=9, append=True, col_names=('MACD_5_15_9', 'MACDh_5_15_9', 'MACDs_5_15_9'))
        for p in [10, 20, 50, 100, 200]: df_feat.ta.sma(close=df_feat[c], length=p, append=True); df_feat.ta.ema(close=df_feat[c], length=p, append=True)
        df_feat.ta.bbands(close=df_feat[c], length=20, std=2, append=True)
        if all(x in df_feat.columns for x in [h,l,c]):
            df_feat.ta.atr(high=df_feat[h], low=df_feat[l], close=df_feat[c], length=14, append=True); df_feat.ta.adx(high=df_feat[h], low=df_feat[l], close=df_feat[c], length=14, append=True)
            df_feat.ta.stoch(high=df_feat[h], low=df_feat[l], close=df_feat[c], append=True); df_feat.ta.willr(high=df_feat[h], low=df_feat[l], close=df_feat[c], append=True); df_feat.ta.cci(high=df_feat[h], low=df_feat[l], close=df_feat[c], append=True)
        if all(x in df_feat.columns for x in [h,l,c,v]): df_feat.ta.mfi(high=df_feat[h], low=df_feat[l], close=df_feat[c], volume=df_feat[v], append=True)
        df_feat.columns = df_feat.columns.str.replace('[^A-Za-z0-9_]+', '', regex=True) # Sanitize column names
        print("Technical indicators added.")
    except Exception as e: print(f"Error adding TIs: {e}\n{traceback.format_exc()}"); return df # Return original on error
    return df_feat

def add_optimized_features(df: pd.DataFrame, price_col='close', volume_col='volume') -> pd.DataFrame:
    print("Adding optimized statistical and ratio features..."); df_new = df.copy()
    df_new['returns'] = df_new[price_col].pct_change()
    safe_price = df_new[price_col].replace(0, np.nan); safe_price_shifted = df_new[price_col].shift(1).replace(0, np.nan)
    df_new['log_returns'] = np.log(safe_price / safe_price_shifted)
    for window in [5, 10, 20, 50]:
        df_new[f'volatility_{window}'] = df_new['log_returns'].rolling(window).std(); df_new[f'skew_{window}'] = df_new['log_returns'].rolling(window).skew(); df_new[f'kurtosis_{window}'] = df_new['log_returns'].rolling(window).kurt()
    if volume_col in df_new.columns and price_col in df_new.columns and df_new[volume_col].isnull().sum() < len(df_new): # Check if volume exists and not all NaN
        rolling_mean_volume = df_new[volume_col].rolling(20).mean().replace(0, np.nan)
        df_new['volume_ratio'] = df_new[volume_col] / rolling_mean_volume; df_new['price_volume'] = df_new[price_col] * df_new[volume_col]; df_new['volume_change'] = df_new[volume_col].pct_change()
    if all(col in df_new.columns for col in ['high', 'low', 'close']):
        safe_low = df_new['low'].replace(0, np.nan); safe_high = df_new['high'].replace(0, np.nan); safe_close = df_new['close'].replace(0, np.nan)
        df_new['high_low_ratio'] = df_new['high'] / safe_low; df_new['close_to_high_ratio'] = safe_close / safe_high; df_new['close_to_low_ratio'] = safe_close / safe_low; df_new['intraday_range_norm'] = (df_new['high'] - df_new['low']) / safe_close
    if 'RSI_14' in df_new.columns: df_new['RSI_signal'] = 0; df_new.loc[df_new['RSI_14'] < 30, 'RSI_signal'] = 1; df_new.loc[df_new['RSI_14'] > 70, 'RSI_signal'] = -1
    if 'MACD_12_26_9' in df_new.columns and 'MACDs_12_26_9' in df_new.columns: df_new['MACD_signal_line_cross'] = (df_new['MACD_12_26_9'] > df_new['MACDs_12_26_9']).astype(int)
    return df_new

def add_wavelet_features(df: pd.DataFrame, column='close', wavelet='mexh', scales_range=(1, 32), num_scales_to_extract=5) -> pd.DataFrame:
    df_feat = df.copy(); print(f"Adding CWT for '{column}' using wavelet '{wavelet}'...");
    if pywt is None: print("PyWavelets not available."); return df_feat
    signal = df_feat[column].values
    if len(signal) < scales_range[1] + 5: print(f"Signal length {len(signal)} too short for CWT. Skipping."); return df_feat
    actual_max_scale = min(scales_range[1], len(signal) // 2 - 1)
    if actual_max_scale < scales_range[0]: print("Max scale too small after constraint. Skipping CWT."); return df_feat
    scales = np.arange(scales_range[0], actual_max_scale + 1)
    if len(scales) == 0: print("No valid scales for CWT. Skipping."); return df_feat
    try:
        coefficients, _ = pywt.cwt(signal, scales, wavelet)
        coeffs_df = pd.DataFrame(coefficients.T, index=df_feat.index, columns=[f"cwt_scale_{s}" for s in scales])
        df_feat[f'{column}_cwt_mean'] = coeffs_df.mean(axis=1); df_feat[f'{column}_cwt_std'] = coeffs_df.std(axis=1)
        s_indices_to_extract = np.linspace(0, len(scales)-1, min(num_scales_to_extract, len(scales)), dtype=int)
        for s_idx in s_indices_to_extract:
            actual_scale_val = scales[s_idx]; col_name_for_scale = f"cwt_scale_{actual_scale_val}"
            if col_name_for_scale in coeffs_df.columns: df_feat[f'{column}_cwt_energy_s{actual_scale_val}'] = coeffs_df[col_name_for_scale]**2
        print("Wavelet features added.")
    except Exception as e: print(f"Error adding wavelet features: {e}\n{traceback.format_exc()}"); return df # Return original on error
    return df_feat

def add_entropy_features(df: pd.DataFrame, column='close', window=40) -> pd.DataFrame:
    df_feat = df.copy(); print(f"Adding Entropy for '{column}' (window={window})...");
    if ant is None: print("Antropy not available."); return df_feat
    if len(df_feat) < window + 15: print(f"Data too short for entropy features. Skipping."); return df_feat
    try:
        sig = df_feat[column].astype(float)
        df_feat[f'{column}_entropy_sample'] = sig.rolling(window=window, min_periods=window).apply(lambda x: ant.sample_entropy(x.dropna()) if x.dropna().shape[0] >= window//2 and x.dropna().shape[0] > 1 else np.nan, raw=False)
        df_feat[f'{column}_entropy_spectral'] = sig.rolling(window=window, min_periods=window).apply(lambda x: ant.spectral_entropy(x.dropna(), sf=1.0, method='welch', nperseg=min(x.dropna().shape[0], window // 2 if window // 2 > 0 else 1)) if x.dropna().shape[0] == window and x.dropna().shape[0] > 1 else np.nan, raw=False)
        print("Entropy features added.")
    except Exception as e: print(f"Error adding entropy features: {e}\n{traceback.format_exc()}"); return df # Return original on error
    return df_feat

def add_advanced_technical_features(df: pd.DataFrame, price_col='close', high_col='high', low_col='low', volume_col='volume') -> pd.DataFrame:
    print("Adding advanced TIs..."); df_new = df.copy()
    if not hasattr(df_new.ta, 'mom'): print("pandas_ta not registered. Skipping advanced TIs."); return df_new
    try:
        if not all(c in df_new.columns for c in [price_col, high_col, low_col]): print("Missing OHLC for adv TIs. Skipping."); return df_new
        for col in [price_col, high_col, low_col]: df_new[col] = df_new[col].astype(float) # Ensure float
        if volume_col in df_new.columns: df_new[volume_col] = df_new[volume_col].astype(float)

        df_new.ta.mom(close=df_new[price_col], append=True); df_new.ta.roc(close=df_new[price_col], append=True)
        df_new.ta.natr(high=df_new[high_col], low=df_new[low_col], close=df_new[price_col], append=True)
        df_new.ta.aroon(high=df_new[high_col], low=df_new[low_col], append=True)
        df_new.ta.stc(close=df_new[price_col], tclength=23, fast=50, slow=100, factor=0.5, append=True, col_names=('STC_23_50_05', 'STCD_23_50_05', 'STCK_23_50_05'))
        if volume_col in df_new.columns and df_new[volume_col].isnull().sum() < len(df_new):
            df_new.ta.pvol(close=df_new[price_col], volume=df_new[volume_col], append=True)
            df_new.ta.cmf(high=df_new[high_col], low=df_new[low_col], close=df_new[price_col], volume=df_new[volume_col], append=True)
        df_new.columns = df_new.columns.str.replace('[^A-Za-z0-9_]+', '', regex=True) # Final sanitize
        print("Advanced TIs added.")
    except Exception as e: print(f"Error adding advanced TIs: {e}\n{traceback.format_exc()}"); return df # Return original on error
    return df_new

def add_transformer_features_conceptual(df: pd.DataFrame, column='close', sequence_length=20) -> pd.DataFrame:
    df_feat = df.copy(); print(f"Adding Transformer-inspired conceptual features for '{column}'...");
    if len(df_feat) < sequence_length + 5: print(f"Data too short for conceptual Transformer features. Skipping."); return df_feat
    feature_col_base = f"{column}_trans_seq"
    for col_suffix in ['mean', 'std', 'trend', 'volatility', 'autocorr1']: df_feat[f'{feature_col_base}_{col_suffix}'] = np.nan
    try:
        data_series = df_feat[column].values; windows = np.lib.stride_tricks.sliding_window_view(data_series, sequence_length)
        results = {key: [np.nan] * (sequence_length -1) for key in ['mean', 'std', 'trend', 'volatility', 'autocorr1']} # Pre-pad with NaNs

        for seq in windows:
            mean_val, std_val = np.mean(seq), np.std(seq)
            norm_seq = (seq - mean_val) / std_val if std_val > 1e-8 else np.zeros_like(seq)
            results['mean'].append(np.mean(norm_seq)); results['std'].append(np.std(norm_seq))
            current_trend, current_vol, current_ac = 0.0, 0.0, 0.0
            if len(norm_seq) > 1:
                try: current_trend = np.polyfit(np.arange(len(norm_seq)), norm_seq, 1)[0] if not np.isnan(np.polyfit(np.arange(len(norm_seq)), norm_seq, 1)[0]) else 0.0
                except: pass # Keep 0.0
                diff_norm_seq = np.diff(norm_seq); current_vol = np.std(diff_norm_seq) if len(diff_norm_seq) > 0 else 0.0
                if len(norm_seq) >= 2:
                    s1, s2 = norm_seq[:-1], norm_seq[1:]
                    if len(s1) >= 1 and np.std(s1) > 1e-8 and np.std(s2) > 1e-8:
                        try: current_ac = np.corrcoef(s1, s2)[0, 1] if not np.isnan(np.corrcoef(s1, s2)[0, 1]) else 0.0
                        except: pass # Keep 0.0
            results['trend'].append(current_trend); results['volatility'].append(current_vol); results['autocorr1'].append(current_ac)
        
        for key, values in results.items(): # Assign all at once, ensuring length match
             if len(values) == len(df_feat): df_feat[f'{feature_col_base}_{key}'] = values
             else: print(f"Warning: Length mismatch for conceptual feature {key}. Expected {len(df_feat)}, got {len(values)}")

        print("Conceptual Transformer-inspired features added.")
    except Exception as e: print(f"Error in conceptual Transformer features: {e}\n{traceback.format_exc()}"); return df
    return df_feat

def add_clasp_regimes(df: pd.DataFrame, column='close', period=20, n_cps=3) -> pd.DataFrame:
    df_reg = df.copy(); print(f"Detecting regimes using ClaSPSegmentation for {column}...");
    if not sktime_available or ClaSPSegmentation is None: print("ClaSPSegmentation not available. Using simple fallback."); return detect_regimes_simple(df, column)
    if column not in df_reg.columns or df_reg[column].isnull().all(): print(f"'{column}' not found for ClaSP. Defaulting regime."); df_reg['regime_clasp'] = 0; return df_reg
    series_for_clasp = df_reg[column].dropna()
    if len(series_for_clasp) < period * 3: print(f"Series length {len(series_for_clasp)} too short for ClaSP (period {period}). Using simple fallback."); return detect_regimes_simple(df, column)
    try:
        clasp = ClaSPSegmentation(period_length=period, n_cps=n_cps); found_cps = clasp.fit_predict(series_for_clasp)
        df_reg['regime_clasp'] = 0; current_regime = 0; last_cp_idx = 0 # Use index for series_for_clasp
        for cp_val_idx in found_cps: # cp_val_idx are indices relative to series_for_clasp
            # Map series_for_clasp indices back to df_reg indices
            original_indices_segment = series_for_clasp.index[last_cp_idx:cp_val_idx]
            df_reg.loc[original_indices_segment, 'regime_clasp'] = current_regime
            current_regime = (current_regime + 1) % (n_cps + 1); last_cp_idx = cp_val_idx
        original_indices_last_segment = series_for_clasp.index[last_cp_idx:]
        df_reg.loc[original_indices_last_segment, 'regime_clasp'] = current_regime
        df_reg['regime_clasp'] = df_reg['regime_clasp'].ffill().bfill().fillna(0) # Fill NaNs in original df
        print(f"ClaSP regimes identified. Distribution:\n{df_reg['regime_clasp'].value_counts(normalize=True).sort_index()*100} %")
    except Exception as e: print(f"Error ClaSPSegmentation: {e}. Using simple fallback."); return detect_regimes_simple(df, column)
    return df_reg

def detect_regimes_simple(df: pd.DataFrame, column='close', window=20) -> pd.DataFrame:
    df_reg = df.copy(); print(f"Detecting regimes (simplified volatility-based) for {column}...");
    if column not in df_reg.columns: print(f"'{column}' not found. Skipping simple regimes."); df_reg['regime_simple'] = 0; return df_reg
    returns = np.log(df_reg[column].replace(0, np.nan) / df_reg[column].shift(1).replace(0, np.nan)).dropna()
    if returns.empty: print("No valid returns for simple regime detection. Skipping."); df_reg['regime_simple'] = 0; return df_reg
    rolling_vol = returns.rolling(window=window, min_periods=window//2 if window//2 > 0 else 1).std()
    df_reg['regime_simple'] = 0 # Default: medium volatility
    if not rolling_vol.dropna().empty:
        vol_low_thresh = rolling_vol.quantile(0.33); vol_high_thresh = rolling_vol.quantile(0.67)
        # Align indices properly after rolling operation
        df_reg.loc[rolling_vol.index[rolling_vol <= vol_low_thresh], 'regime_simple'] = 1  # Low vol
        df_reg.loc[rolling_vol.index[rolling_vol > vol_high_thresh], 'regime_simple'] = 2   # High vol
    else: print("Not enough data for vol percentile. Defaulting simple regimes to 0.")
    df_reg['regime_simple'] = df_reg['regime_simple'].fillna(method='bfill').fillna(method='ffill').fillna(0) # Ensure all NaNs are filled
    print(f"Simple Regimes (0:Med,1:Low,2:High):\n{df_reg['regime_simple'].value_counts(normalize=True, dropna=False).sort_index()*100} %"); return df_reg

def balanced_target_definition(df: pd.DataFrame, column='close', periods=5, lower_q_thresh=0.45, upper_q_thresh=0.55) -> pd.DataFrame:
    df_t = df.copy(); print(f"Balanced target definition for '{column}' over {periods} periods...");
    if column not in df_t.columns: print(f"'{column}' not found for target. Defaulting target."); df_t['target'] = 0; return df_t
    df_t[column] = pd.to_numeric(df_t[column], errors='coerce').replace(0, np.nan) # Handle zeros before log
    df_t['future_log_return_target'] = np.log(df_t[column].shift(-periods) / df_t[column])
    valid_returns = df_t['future_log_return_target'].dropna()
    df_t['target'] = 0 # Default class (e.g., Hold/Neutral for binary if middle band is also 0)
    if len(valid_returns) > 20:
        lower_q_val = valid_returns.quantile(lower_q_thresh); upper_q_val = valid_returns.quantile(upper_q_thresh)
        df_t.loc[df_t['future_log_return_target'] < lower_q_val, 'target'] = 0 # Example: Assign 0 for "down/neutral"
        df_t.loc[df_t['future_log_return_target'] > upper_q_val, 'target'] = 1 # Example: Assign 1 for "up"
    else: print("Not enough valid returns for quantile-based target balancing. Default target used.")
    df_t.drop(columns=['future_log_return_target'], inplace=True, errors='ignore')
    print(f"Target distribution:\n{df_t['target'].value_counts(normalize=True, dropna=False)*100}"); return df_t

def discover_causal_structure(df_features: pd.DataFrame, target_col='target', max_feats=10, symbol="") -> CausalModel | None:
    print(f"\nDiscovering causal structure for {symbol} using DoWhy...");
    if not dowhy_available or CausalModel is None: print("DoWhy not available."); return None
    df_c = df_features.copy()
    if target_col not in df_c.columns or df_c[target_col].isnull().all(): print(f"Target '{target_col}' missing for causal discovery."); return None
    df_c[target_col] = pd.to_numeric(df_c[target_col], errors='coerce')
    cand_cols = [c for c in df_c.columns if pd.api.types.is_numeric_dtype(df_c[c]) and c != target_col and df_c[c].notnull().any() and df_c[c].var() > 1e-6] # Added variance check
    if not cand_cols: print("No numeric candidate columns with variance for causal discovery."); return None
    
    df_subset_for_causal = df_c[cand_cols + [target_col]].copy(); df_subset_for_causal.replace([np.inf, -np.inf], np.nan, inplace=True)
    if cand_cols: scaler_causal = StandardScaler(); df_subset_for_causal[cand_cols] = scaler_causal.fit_transform(df_subset_for_causal[cand_cols])
    df_subset_for_causal.dropna(inplace=True)
    if df_subset_for_causal.empty or target_col not in df_subset_for_causal.columns or df_subset_for_causal[target_col].nunique() < 1: print("Not enough data post-cleaning/scaling for causal discovery."); return None
    
    # Use column name prefix from feature engineering if it exists
    cwt_mean_col = f"{price_c}_cwt_mean" if f"{price_c}_cwt_mean" in df_subset_for_causal.columns else 'close_cwt_mean' # Fallback
    cwt_std_col = f"{price_c}_cwt_std" if f"{price_c}_cwt_std" in df_subset_for_causal.columns else 'close_cwt_std'
    entropy_sample_col = f"{price_c}_entropy_sample" if f"{price_c}_entropy_sample" in df_subset_for_causal.columns else 'close_entropy_sample'

    potential_causes = ['RSI_14', 'MACDh_12_26_9', 'ADX_14', 'ATR_14', cwt_mean_col, cwt_std_col, entropy_sample_col, 'regime_simple', 'volatility_20', 'log_returns']
    # Add default bband names from pandas_ta (e.g. BBL_20_2.0, BBM_20_2.0, BBU_20_2.0, BBB_20_2.0, BBP_20_2.0)
    # Need to sanitize them: BBL_20_20, BBM_20_20 etc.
    potential_causes.extend(['BBP_20_20', 'BBB_20_20']) # Ensure these names match sanitized names after add_technical_indicators
    
    graph_feats = [c for c in potential_causes if c in df_subset_for_causal.columns and c != target_col and df_subset_for_causal[c].nunique() > 1]
    if not graph_feats:
        print("Predefined causal graph_feats not suitable, selecting top varying features (after scaling).")
        num_to_select = min(max_feats, len(cand_cols));
        if num_to_select > 0: graph_feats = df_subset_for_causal[cand_cols].var().nlargest(num_to_select).index.tolist()
        else: print("No candidate columns for graph_feats fallback."); return None
    if not graph_feats: print("No suitable graph features for causal discovery."); return None
    
    final_df_for_causal_model = df_subset_for_causal[graph_feats + [target_col]].copy()
    if final_df_for_causal_model.empty or final_df_for_causal_model.shape[0] < 20 or final_df_for_causal_model[target_col].nunique() < 1: print("Final DF for causal model too small or target has no variation."); return None
    print(f"DoWhy using graph features: {graph_feats} for Outcome: {target_col}"); treatment_var = graph_feats[0]
    graph_str = "digraph { " + "; ".join([f'"{f}" -> "{target_col}"' for f in graph_feats]) + " }"
    print(f"Generated Causal Graph:\n{graph_str}")
    try: model = CausalModel(data=final_df_for_causal_model, treatment=treatment_var, outcome=target_col, graph=graph_str); print("DoWhy CausalModel created."); return model
    except Exception as e: print(f"DoWhy CausalModel error: {e}\n{traceback.format_exc()}"); return None

def discover_causal_structure(df_features: pd.DataFrame, target_col='target', price_c='close', max_feats=10, symbol="") -> CausalModel | None:
    print(f"\nDiscovering causal structure for {symbol} using DoWhy...");
    if not dowhy_available or CausalModel is None:
        print("DoWhy not available.")
        return None
    
    df_c = df_features.copy()
    
    if target_col not in df_c.columns or df_c[target_col].isnull().all():
        print(f"Target '{target_col}' missing for causal discovery.")
        return None
    df_c[target_col] = pd.to_numeric(df_c[target_col], errors='coerce')
    cand_cols = [c for c in df_c.columns if pd.api.types.is_numeric_dtype(df_c[c]) and c != target_col and df_c[c].notnull().any() and df_c[c].var() > 1e-6] # Added variance check
    if not cand_cols:
        print("No numeric candidate columns with variance for causal discovery.")
        return None
    
    df_subset_for_causal = df_c[cand_cols + [target_col]].copy()
    df_subset_for_causal.replace([np.inf, -np.inf], np.nan, inplace=True)
    if cand_cols:
        scaler_causal = StandardScaler()
        df_subset_for_causal[cand_cols] = scaler_causal.fit_transform(df_subset_for_causal[cand_cols])
    df_subset_for_causal.dropna(inplace=True)
    if df_subset_for_causal.empty or target_col not in df_subset_for_causal.columns or df_subset_for_causal[target_col].nunique() < 1:
        print("Not enough data post-cleaning/scaling for causal discovery.")
        return None
    
    # Use column name prefix from feature engineering if it exists
    cwt_mean_col = f"{price_c}_cwt_mean" if f"{price_c}_cwt_mean" in df_subset_for_causal.columns else 'close_cwt_mean' # Fallback
    cwt_std_col = f"{price_c}_cwt_std" if f"{price_c}_cwt_std" in df_subset_for_causal.columns else 'close_cwt_std'
    entropy_sample_col = f"{price_c}_entropy_sample" if f"{price_c}_entropy_sample" in df_subset_for_causal.columns else 'close_entropy_sample'

    potential_causes = ['RSI_14', 'MACDh_12_26_9', 'ADX_14', 'ATR_14', cwt_mean_col, cwt_std_col, entropy_sample_col, 'regime_simple', 'volatility_20', 'log_returns']
    # Add default bband names from pandas_ta (e.g. BBL_20_2.0, BBM_20_2.0, BBU_20_2.0, BBB_20_2.0, BBP_20_2.0)
    # Need to sanitize them: BBL_20_20, BBM_20_20 etc.
    potential_causes.extend(['BBP_20_20', 'BBB_20_20']) # Ensure these names match sanitized names after add_technical_indicators
    
    graph_feats = [c for c in potential_causes if c in df_subset_for_causal.columns and c != target_col and df_subset_for_causal[c].nunique() > 1]
    if not graph_feats:
        print("Predefined causal graph_feats not suitable, selecting top varying features (after scaling).")
        num_to_select = min(max_feats, len(cand_cols));
        if num_to_select > 0:
            graph_feats = df_subset_for_causal[cand_cols].var().nlargest(num_to_select).index.tolist()
        else:
            print("No candidate columns for graph_feats fallback.")
            return None
    if not graph_feats:
        print("No suitable graph features for causal discovery.")
        return None
    
    final_df_for_causal_model = df_subset_for_causal[graph_feats + [target_col]].copy()
    if final_df_for_causal_model.empty or final_df_for_causal_model.shape[0] < 20 or final_df_for_causal_model[target_col].nunique() < 1:
        print("Final DF for causal model too small or target has no variation.")
        return None
    print(f"DoWhy using graph features: {graph_feats} for Outcome: {target_col}")
    treatment_var = graph_feats[0]
    graph_str = "digraph { " + "; ".join([f'"{f}" -> "{target_col}"' for f in graph_feats]) + " }"
    print(f"Generated Causal Graph:\n{graph_str}")
    try:
        model = CausalModel(data=final_df_for_causal_model, treatment=treatment_var, outcome=target_col, graph=graph_str)
        print("DoWhy CausalModel created.")
        return model
    except Exception as e:
        print(f"DoWhy CausalModel error: {e}\n{traceback.format_exc()}")
        return None

def prepare_ml_data(df: pd.DataFrame, target_col='target', test_split_size=0.15, min_test_samples=50):
    print("Preparing ML data...");
    if target_col not in df.columns: print(f"Target '{target_col}' missing."); return None, None, None, None, None
    cols_to_drop_base = ['open', 'high', 'low', 'close', 'volume', 'returns'] # Keep log_returns
    cols_to_drop_dynamic = [c for c in df.columns if 'target_' in c and c != target_col] + [c for c in df.columns if 'future_return' in c]
    all_cols_to_drop = list(set(cols_to_drop_base + cols_to_drop_dynamic))
    X = df.drop(columns=[c for c in all_cols_to_drop if c in df.columns] + [target_col], errors='ignore')
    y = df[target_col].copy()
    if y.isnull().all(): print("Target is all NaN."); return None, None, None, None, None
    valid_target_mask = y.notna(); X = X.loc[valid_target_mask]; y = y.loc[valid_target_mask]
    if X.empty or y.empty: print("X or y empty after target NaN filter."); return None, None, None, None, None
    X.replace([np.inf, -np.inf], np.nan, inplace=True)
    if X.isnull().any().any(): X = X.fillna(X.median()) # Use median for robustness to outliers
    if X.isnull().any().any(): # If median fails (e.g. all-NaN column), drop rows with any NaNs
        print(f"Warning: NaNs still present after median imputation. Dropping rows with NaNs in X. Nulls per col:\n{X.isnull().sum()[X.isnull().sum()>0]}")
        X.dropna(axis=0, how='any', inplace=True); y = y.loc[X.index]
    if X.empty or y.empty: print("X or y empty after internal NaN handling."); return None, None, None, None, None
    print(f"Data shape post-NaN handling in prepare_ml_data: X={X.shape}, y={y.shape}")
    if len(X) < min_test_samples * 2: print(f"Not enough data ({len(X)} rows) for robust split."); return None, None, None, None, None
    n_samples = len(X); test_size_abs = max(min_test_samples, int(n_samples * test_split_size))
    if n_samples - test_size_abs < min_test_samples: test_size_abs = n_samples - min_test_samples
    if test_size_abs < 1 and n_samples > 0 : test_size_abs = 1 # Ensure at least one test sample if possible
    elif test_size_abs < 1 : print(f"Cannot make meaningful split."); return None,None,None,None,None

    train_size = n_samples - test_size_abs
    if train_size < 1: print(f"Train size too small ({train_size}). Cannot split."); return None,None,None,None,None

    X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]; y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]
    if X_train.empty or y_train.empty or X_test.empty or y_test.empty: print("Train/Test set empty post-split."); return None, None, None, None, None
    print(f"Train shapes: X_train={X_train.shape}, y_train={y_train.shape}; Test shapes: X_test={X_test.shape}, y_test={y_test.shape}")
    numeric_cols_xtrain = X_train.select_dtypes(include=np.number).columns
    if not numeric_cols_xtrain.empty:
        scaler = StandardScaler(); X_train_scaled_np = scaler.fit_transform(X_train[numeric_cols_xtrain])
        X_train_scaled = pd.DataFrame(X_train_scaled_np, columns=numeric_cols_xtrain, index=X_train.index)
        for col in X_train.columns.difference(numeric_cols_xtrain): X_train_scaled[col] = X_train[col]
        numeric_cols_xtest = X_test.select_dtypes(include=np.number).columns
        cols_to_scale_in_test = [col for col in numeric_cols_xtrain if col in numeric_cols_xtest] # Use cols from train scaler
        if cols_to_scale_in_test:
            X_test_scaled_np = scaler.transform(X_test[cols_to_scale_in_test])
            X_test_scaled = pd.DataFrame(X_test_scaled_np, columns=cols_to_scale_in_test, index=X_test[cols_to_scale_in_test].index)
            for col in X_test.columns.difference(cols_to_scale_in_test): X_test_scaled[col] = X_test[col]
        else: X_test_scaled = X_test.copy(); print("No common numeric cols to scale in X_test or X_test has no numeric.")
        return X_train_scaled, X_test_scaled, y_train, y_test, scaler
    else: print("No numeric columns in X_train for scaling."); return X_train, X_test, y_train, y_test, None

# ... (lgbm_objective, optimize_lgbm_hyperparameters, train_lightgbm_model, plot_feature_importance are mostly fine, ensure robust validation split in optimize_lgbm) ...
# Optuna validation split in optimize_lgbm_hyperparameters needs care for time series.
# Current TimeSeriesSplit in optimize_lgbm_hyperparameters is for a single validation set from training data.

def lgbm_objective(trial, X_train, y_train, X_val, y_val, base_params):
    params = {**base_params, **{'boosting_type': 'gbdt', 'n_estimators': trial.suggest_int('n_estimators', 100, 1000, step=50), 'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.1, log=True), 'num_leaves': trial.suggest_int('num_leaves', 10, 50), 'max_depth': trial.suggest_int('max_depth', -1, 10), 'min_child_samples': trial.suggest_int('min_child_samples', 5, 50), 'feature_fraction': trial.suggest_float('feature_fraction', 0.5, 1.0), 'bagging_fraction': trial.suggest_float('bagging_fraction', 0.5, 1.0), 'bagging_freq': trial.suggest_int('bagging_freq', 0, 7), 'reg_alpha': trial.suggest_float('reg_alpha', 1e-3, 1.0, log=True), 'reg_lambda': trial.suggest_float('reg_lambda', 1e-3, 1.0, log=True)}}
    if params.get('num_class') is None and 'num_class' in params: del params['num_class'] # Clean up if None
    model = lgb.LGBMClassifier(**params)
    model.fit(X_train, y_train, eval_set=[(X_val, y_val)], eval_metric=base_params['metric'], callbacks=[lgb.early_stopping(30, verbose=False)])
    y_proba_val = model.predict_proba(X_val)
    return log_loss(y_val, y_proba_val)

def optimize_lgbm_hyperparameters(X_train: pd.DataFrame, y_train: pd.Series, base_params: dict, n_trials=30, validation_ratio=0.2) -> dict: # Reduced n_trials default
    if not optuna_available: print("Optuna not available. Using default HPs."); return optimized_lightgbm_params()
    print(f"Optimizing LightGBM HPs with Optuna ({n_trials} trials)...")
    if len(X_train) * validation_ratio < 1 or len(X_train) * (1-validation_ratio) < 1: print("Too few samples for Optuna val split. Using defaults."); return optimized_lightgbm_params()
    
    # Create a single validation split from the end of the training data (time-series aware)
    n_val_samples = int(len(X_train) * validation_ratio)
    if n_val_samples == 0 and len(X_train) > 1: n_val_samples = 1 # Ensure at least 1 sample if possible
    elif n_val_samples == 0 : print("Cannot create val set for Optuna. Using defaults."); return optimized_lightgbm_params()
    
    n_train_opt_samples = len(X_train) - n_val_samples
    if n_train_opt_samples == 0 : print("Train set for Optuna is empty. Using defaults."); return optimized_lightgbm_params()

    X_train_opt, X_val_opt = X_train.iloc[:n_train_opt_samples], X_train.iloc[n_train_opt_samples:]
    y_train_opt, y_val_opt = y_train.iloc[:n_train_opt_samples], y_train.iloc[n_train_opt_samples:]

    study = optuna.create_study(direction='minimize') # We want to minimize log_loss
    study.optimize(lambda trial: lgbm_objective(trial, X_train_opt, y_train_opt, X_val_opt, y_val_opt, base_params), n_trials=n_trials, show_progress_bar=True)
    print(f"Best Optuna trial for LightGBM: Value={study.best_value:.4f}, Params={study.best_params}")
    final_params = {**base_params, **study.best_params} # Combine with base params like objective, metric
    return final_params

# train_lightgbm_model, plot_feature_importance, export_lgbm_to_onnx, simple_feature_selection_fallback, prioritized_feature_selection, optimized_lightgbm_params
# These functions from the last provided script were largely okay, with the ONNX check being the main thing to ensure is fixed by global flag initializations.
# For brevity, I will assume they are mostly as per your last "complete script with fixes" and focus on the run_patchtst_foundation_forecast fix.

def train_lightgbm_model(X_train, y_train, X_test, y_test, optimized_params=None): # Accept optimized_params
    print("Training LightGBM model...")
    if X_train.empty or y_train.empty: print("X_train or y_train is empty. Skipping LightGBM training."); return None, None
    y_train_squeezed = y_train.squeeze(); y_test_squeezed = y_test.squeeze()
    unique_labels_train = sorted(y_train_squeezed.unique()); num_classes = len(unique_labels_train)
    if num_classes <= 1: print(f"Only {num_classes} class(es) in y_train. Skipping LightGBM training."); return None, None
    base_params = {'objective': 'multiclass' if num_classes > 2 else 'binary', 'metric': 'multi_logloss' if num_classes > 2 else 'binary_logloss', 'random_state': 42, 'verbose': -1, 'n_jobs': -1, 'class_weight': 'balanced'}
    if base_params['objective'] == 'multiclass': base_params['num_class'] = num_classes
    
    current_params = {**base_params, **(optimized_params if isinstance(optimized_params, dict) else optimized_lightgbm_params())}
    if optimized_params and isinstance(optimized_params, dict): print("Using Optuna-optimized (or provided) LightGBM parameters.")
    else: print("Using default (or non-Optuna optimized) LightGBM parameters.")

    label_map = {label: i for i, label in enumerate(unique_labels_train)}
    y_train_mapped = y_train_squeezed.map(label_map); y_test_mapped = y_test_squeezed.map(label_map).fillna(-1).astype(int)
    model = lgb.LGBMClassifier(**current_params)
    valid_eval_indices = y_test_mapped != -1
    # Ensure X_test is also indexed by valid_eval_indices for eval_set
    eval_set_data = (X_test[valid_eval_indices], y_test_mapped[valid_eval_indices]) if valid_eval_indices.any() else None

    if eval_set_data: model.fit(X_train, y_train_mapped, eval_set=[eval_set_data], eval_metric=current_params.get('metric', 'logloss'), callbacks=[lgb.early_stopping(30, verbose=False)])
    else: print("Warning: No valid eval set. Fitting on full training data."); model.fit(X_train, y_train_mapped)
    
    if not valid_eval_indices.any(): print("No valid test samples for evaluation after mapping."); return model, None
    
    # Predict only on the part of X_test that has valid mapped labels
    y_pred_mapped_on_valid = model.predict(X_test[valid_eval_indices])
    y_proba_on_valid = model.predict_proba(X_test[valid_eval_indices])
    
    acc = accuracy_score(y_test_mapped[valid_eval_indices], y_pred_mapped_on_valid)
    print(f"\nüéØ Accuracy on mapped test data: {acc:.4f}")
    if current_params['objective'] == 'binary' and y_proba_on_valid.shape[1] == 2:
        try: auc = roc_auc_score(y_test_mapped[valid_eval_indices], y_proba_on_valid[:, 1]); print(f"üìä AUC: {auc:.4f}")
        except ValueError as e_auc: print(f"AUC Calc Error: {e_auc}")
    print("\nClassification Report (on mapped and valid test labels):")
    try: print(classification_report(y_test_mapped[valid_eval_indices], y_pred_mapped_on_valid, labels=sorted(y_train_mapped.unique()), zero_division=0))
    except Exception as e_cr: print(f"Classification Report Error: {e_cr}")
    feat_imp_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': model.feature_importances_}).sort_values(by='Importance', ascending=False)
    print("\nTop 10 features:\n", feat_imp_df.head(10)); return model, feat_imp_df

def plot_feature_importance(feature_importance_df, top_n=20, symbol_for_plot="", min_bar_height=0.05):
    """
    Enhanced feature importance plotting with minimum bar visibility
    """
    import matplotlib.pyplot as plt
    import seaborn as sns
    import numpy as np
    
    # Ensure minimum visibility for low-importance features
    plot_data = feature_importance_df.head(top_n).copy()
    max_importance = plot_data['Importance'].max()
    min_threshold = max_importance * 0.02  # 2% of max importance as minimum
    plot_data['Importance'] = np.maximum(plot_data['Importance'], min_threshold)
    
    plt.figure(figsize=(14, max(8, min(top_n, len(plot_data)) * 0.5)))
    
    # Create color gradient based on original importance values
    colors = plt.cm.viridis(np.linspace(0, 1, len(plot_data)))
    
    ax = sns.barplot(
        x='Importance', 
        y='Feature', 
        data=plot_data, 
        palette=colors,
        orient='h'
    )
    
    # Add value labels on bars
    for i, (idx, row) in enumerate(plot_data.iterrows()):
        original_val = feature_importance_df.iloc[i]['Importance']
        ax.text(row['Importance'] + max_importance * 0.01, i, 
                f'{original_val:.2f}', 
                va='center', fontsize=9, fontweight='bold')
    
    plt.title(f'Top {top_n} Feature Importances for {symbol_for_plot} (LightGBM)', 
              fontsize=16, fontweight='bold', pad=20)
    plt.xlabel('Importance', fontsize=12, fontweight='bold')
    plt.ylabel('Feature', fontsize=12, fontweight='bold')
    plt.grid(axis='x', alpha=0.3)
    plt.tight_layout()
    plt.show()


def export_lgbm_to_onnx(lgbm_model, X_sample_df, file_path="lgbm_model.onnx", target_opset=12):
    print(f"\nExporting LGBM model to ONNX: {file_path} (opset={target_opset})")
    if not all([onnx_available, skl2onnx_available, onnxmltools_available, (FloatTensorType is not None)]): # Corrected check
        print("One or more ONNX libraries missing or FloatTensorType not imported. Skipping ONNX export."); return None
    if lgbm_model is None or X_sample_df is None or X_sample_df.empty: print("Model or sample data empty for ONNX. Skipping."); return None
    try:
        initial_type = [('float_input', FloatTensorType([None, X_sample_df.shape[1]]))]
        converted_model = onnxmltools.convert_lightgbm(lgbm_model, initial_types=initial_type, target_opset=target_opset)
        with open(file_path, "wb") as f: f.write(converted_model.SerializeToString())
        print(f"Model exported to ONNX: {file_path}"); onnx.checker.check_model(file_path); print("ONNX model check OK."); return file_path
    except Exception as e: # Outer except for ONNX export
        print(f"Error exporting LGBM to ONNX: {e}. Fallback to pickle.")
        try: # Inner try for pickle fallback
            import pickle
            pkl_path = file_path.replace('.onnx', '.pkl')
            with open(pkl_path, 'wb') as pf:
                pickle.dump(lgbm_model, pf)
            print(f"Model saved as pickle: {pkl_path}")
            return pkl_path
        except Exception as ep: # Inner except for pickle fallback
            print(f"Pickle save error: {ep}")
            return None

# simple_feature_selection_fallback and prioritized_feature_selection are mostly fine
def simple_feature_selection_fallback(X_train, y_train, max_features=20): # Same as before
    print("Using simple variance-based feature selection fallback...");
    if X_train.empty: return pd.DataFrame(); X_train_numeric = X_train.select_dtypes(include=np.number)
    if X_train_numeric.empty: print("No numeric features for variance selection."); return pd.DataFrame({'Feature': X_train.columns[:max_features].tolist()})
    variance_scores = X_train_numeric.var().sort_values(ascending=False); num_features_to_select = min(max_features, len(variance_scores)); selected_features = variance_scores.head(num_features_to_select).index.tolist()
    return pd.DataFrame({'Feature': selected_features, 'Score': variance_scores.head(num_features_to_select).values})

def prioritized_feature_selection(X_train, y_train, causal_ranking, max_features=25):
    print("Prioritized feature selection: Causal Ranking + Mutual Information...");
    if X_train.empty or y_train.empty:
        print("X_train or y_train is empty in prioritized_feature_selection. Returning empty DataFrame.")
        return pd.DataFrame() # Early exit if data is empty

    top_causal_features = [] # Initialize top_causal_features as an empty list

    num_causal_to_select = 0
    if causal_ranking: # Check if causal_ranking has items
        num_causal_to_select = min(len(causal_ranking), max_features // 2)

    if causal_ranking and num_causal_to_select > 0:
        print(f"Selecting up to {num_causal_to_select} features from causal ranking.")
        for feat, score in causal_ranking[:num_causal_to_select]:
            if feat in X_train.columns:
                top_causal_features.append(feat)
            else:
                print(f"Causal feature '{feat}' not in X_train.columns. Skipping.")
    else:
        print("No causal ranking provided or num_causal_to_select is 0.")
    
    print(f"Features selected from causal ranking: {top_causal_features}")

    remaining_slots = max_features - len(top_causal_features)
    stat_selected_features = []

    if remaining_slots > 0:
        features_for_stat_selection = [f for f in X_train.columns if f not in top_causal_features]
        if features_for_stat_selection:
            X_remaining_for_stat = X_train[features_for_stat_selection]
            y_train_squeezed = y_train.squeeze()
            
            if y_train_squeezed.nunique() > 1 and not X_remaining_for_stat.empty:
                X_remaining_numeric = X_remaining_for_stat.select_dtypes(include=np.number)
                if not X_remaining_numeric.empty:
                    num_stat_to_select = min(remaining_slots, X_remaining_numeric.shape[1])
                    if num_stat_to_select > 0: # Ensure k is positive for SelectKBest
                        try:
                            print(f"Selecting up to {num_stat_to_select} features using Mutual Information from {X_remaining_numeric.shape[1]} numeric features.")
                            selector_mi = SelectKBest(mutual_info_classif, k=num_stat_to_select)
                            selector_mi.fit(X_remaining_numeric, y_train_squeezed)
                            stat_selected_features = X_remaining_numeric.columns[selector_mi.get_support()].tolist()
                            print(f"Features selected from Mutual Information: {stat_selected_features}")
                        except Exception as e_mi:
                            print(f"Error in MI based feature selection: {e_mi}. Proceeding without these stat features.")
                    else:
                        print("No statistical features to select (num_stat_to_select is 0).")
                else:
                    print("No numeric features remaining for MI based statistical selection.")
            else:
                print("Not enough target variance or no remaining features for MI based selection.")
        else:
            print("No features remaining for MI based statistical selection (features_for_stat_selection is empty).")
    else:
        print("No remaining slots for statistical feature selection.")

    final_selected_features = list(dict.fromkeys(top_causal_features + stat_selected_features)) # Combine and keep unique, preserving order

    if not final_selected_features and not X_train.empty:
        print("No features from prioritized selection, falling back to simple variance-based selection.");
        simple_fallback_df = simple_feature_selection_fallback(X_train, y_train, max_features)
        if simple_fallback_df is not None and 'Feature' in simple_fallback_df.columns:
             final_selected_features = simple_fallback_df['Feature'].tolist()
        else:
            final_selected_features = X_train.columns[:max_features].tolist() # Last resort

    print(f"Total prioritized features selected: {len(final_selected_features)}. Top 5: {final_selected_features[:5] if final_selected_features else 'None'}")
    return pd.DataFrame({'Feature': final_selected_features})

def configure_extended_context(base_context=512, data_length=750):
    """
    Dynamically configure context window based on available data
    """
    # Calculate optimal context length
    max_possible_context = int(data_length * 0.7)  # Use 70% of data for context
    
    extended_contexts = {
        'short_term': min(256, max_possible_context),
        'medium_term': min(512, max_possible_context), 
        'long_term': min(1024, max_possible_context),
        'adaptive': min(base_context * 2, max_possible_context)
    }
    
    print(f"Context window options for {data_length} data points:")
    for name, length in extended_contexts.items():
        print(f"  {name}: {length} days")
    
    return extended_contexts

# Usage in your main workflow
#context_configs = configure_extended_context(data_length=len(processed_df))
#selected_context = context_configs['adaptive']

def add_multitimeframe_features(df, price_col='close'):
    """
    Add multi-timeframe technical indicators
    """
    import pandas as pd
    import numpy as np
    
    # Weekly aggregation (5-day)
    df[f'{price_col}_weekly_mean'] = df[price_col].rolling(5).mean()
    df[f'{price_col}_weekly_std'] = df[price_col].rolling(5).std()
    df[f'{price_col}_weekly_max'] = df[price_col].rolling(5).max()
    df[f'{price_col}_weekly_min'] = df[price_col].rolling(5).min()
    
    # Monthly aggregation (21-day)
    df[f'{price_col}_monthly_mean'] = df[price_col].rolling(21).mean()
    df[f'{price_col}_monthly_std'] = df[price_col].rolling(21).std()
    df[f'{price_col}_monthly_trend'] = df[price_col].rolling(21).apply(
        lambda x: np.polyfit(range(len(x)), x, 1)[0] if len(x) == 21 else np.nan
    )
    
    # Quarterly aggregation (63-day)
    df[f'{price_col}_quarterly_mean'] = df[price_col].rolling(63).mean()
    df[f'{price_col}_quarterly_volatility'] = df[price_col].rolling(63).std() / df[price_col].rolling(63).mean()
    
    # Cross-timeframe ratios
    df[f'{price_col}_weekly_monthly_ratio'] = df[f'{price_col}_weekly_mean'] / df[f'{price_col}_monthly_mean']
    df[f'{price_col}_monthly_quarterly_ratio'] = df[f'{price_col}_monthly_mean'] / df[f'{price_col}_quarterly_mean']
    
    print(f"Added {9} multi-timeframe features")
    return df



def detect_volatility_regimes(returns, window=21, threshold=1.5):
    """
    Detect high/low volatility regimes using rolling volatility
    """
    import pandas as pd
    import numpy as np
    
    # Calculate rolling volatility
    rolling_vol = returns.rolling(window).std()
    vol_median = rolling_vol.median()
    
    # Define regimes based on volatility threshold
    high_vol_threshold = vol_median * threshold
    low_vol_threshold = vol_median / threshold
    
    regimes = pd.Series(index=returns.index, dtype=int)
    regimes[rolling_vol >= high_vol_threshold] = 2  # High volatility
    regimes[rolling_vol <= low_vol_threshold] = 0   # Low volatility  
    regimes[(rolling_vol > low_vol_threshold) & (rolling_vol < high_vol_threshold)] = 1  # Normal
    
    # Forward fill regime assignments
    regimes = regimes.fillna(method='ffill').fillna(1)
    
    regime_counts = regimes.value_counts().sort_index()
    print(f"Volatility regime distribution:")
    print(f"  Low volatility (0): {regime_counts.get(0, 0)} periods")
    print(f"  Normal volatility (1): {regime_counts.get(1, 0)} periods") 
    print(f"  High volatility (2): {regime_counts.get(2, 0)} periods")
    
    return regimes

def add_regime_features(df, returns_col='log_returns'):
    """
    Add regime-based features to dataframe
    """
    regimes = detect_volatility_regimes(df[returns_col])
    
    df['volatility_regime'] = regimes
    df['regime_0'] = (regimes == 0).astype(int)
    df['regime_1'] = (regimes == 1).astype(int) 
    df['regime_2'] = (regimes == 2).astype(int)
    
    # Regime-specific moving averages
    for window in [10, 20, 50]:
        df[f'sma_{window}_regime_adj'] = df['close'].rolling(window).mean() * (1 + 0.1 * regimes)
    
    print(f"Added 7 regime-based features")
    return df




def optimized_lightgbm_params(): # Same as before
    return {'boosting_type': 'gbdt', 'num_leaves': 31, 'learning_rate': 0.05, 'n_estimators': 200, 'feature_fraction': 0.8, 'bagging_fraction': 0.8, 'bagging_freq': 5, 'min_child_samples': 20, 'reg_alpha': 0.01, 'reg_lambda': 0.01, 'random_state': 42, 'verbose': -1, 'n_jobs': -1, 'class_weight': 'balanced'}



def enhanced_patchtst_finetune(log_returns, context_length=512, prediction_length=5, min_sequences=20):
    """
    Enhanced PatchTST fine-tuning with intelligent data padding
    """
    import numpy as np
    from sklearn.preprocessing import StandardScaler

    sequence_length = context_length + prediction_length

    # Enhanced data sufficiency logic
    if len(log_returns) < sequence_length:
        print(f"Insufficient raw data ({len(log_returns)} < {sequence_length}). Applying intelligent padding...")

        # Calculate statistical properties
        mean_return = np.mean(log_returns)
        std_return = np.std(log_returns)

        # Apply regime-aware padding
        recent_volatility = np.std(log_returns[-min(30, len(log_returns)):])
        volatility_multiplier = recent_volatility / std_return if std_return > 0 else 1.0

        # Create synthetic historical context
        padding_length = sequence_length - len(log_returns)

        # Use autoregressive approach for better synthetic data
        synthetic_returns = []
        last_return = log_returns[0] if len(log_returns) > 0 else 0

        for _ in range(padding_length):
            # AR(1) with regime-aware noise
            next_return = 0.1 * last_return + np.random.normal(
                mean_return, std_return * volatility_multiplier
            )
            synthetic_returns.append(next_return)
            last_return = next_return

        padded_returns = np.concatenate([synthetic_returns, log_returns])
        print(f"Applied intelligent padding: {padding_length} synthetic points added")

    elif len(log_returns) >= sequence_length * 2:  # Reduced from *5 to *2
        padded_returns = log_returns
        print(f"Sufficient data for fine-tuning: {len(log_returns)} points available")

    else:
        # Intermediate case: use available data with minimal padding
        shortage = sequence_length * 2 - len(log_returns)

        # Statistical padding for small shortage
        mean_return = np.mean(log_returns)
        std_return = np.std(log_returns)

        padding = np.random.normal(mean_return, std_return, shortage)
        padded_returns = np.concatenate([padding, log_returns])
        print(f"Applied minimal padding: {shortage} points added")
    # Create sequences for training
    sequences = []
    targets = []

    for i in range(len(padded_returns) - sequence_length + 1):
        seq = padded_returns[i:i + context_length]
        target = padded_returns[i + context_length:i + sequence_length]
        sequences.append(seq)
        targets.append(target)

    if len(sequences) >= min_sequences:
        print(f"Created {len(sequences)} training sequences for fine-tuning")
        return np.array(sequences), np.array(targets), True
    else:
        print(f"Still insufficient sequences ({len(sequences)} < {min_sequences}). Using pre-trained weights.")
        return None, None, False

    
# --- FOUNDATION MODEL FORECASTING ---
def run_patchtst_foundation_forecast(
    symbol_name: str,
    historical_data_df: pd.DataFrame,
    prediction_length: int = 5,
    model_checkpoint: str = "ibm-research/patchtst-etth1-pretrain",
    fine_tune_epochs: int = 10,
    enable_fine_tuning: bool = True,
    configured_context_length: int | None = None # <-- ADDED for dynamic context
):
    """
    Complete PatchTST implementation with domain-specific fine-tuning for financial data.
    """
    print(f"\n--- Forecasting for {symbol_name} using PatchTST ({model_checkpoint}) ---")

    # Early validation
    if not torch_available or 'PatchTSTForPrediction' not in globals():
        print("Torch or PatchTST not available. Skipping.")
        return {"status": "failed", "reason": "Dependencies missing"}

    if 'close' not in historical_data_df.columns:
        print(f"'close' column missing for {symbol_name}. Skipping.")
        return {"status": "failed", "reason": "'close' missing"}

    # Ensure minimum data requirements (using a base config context length for this check)
    # temp_config_for_check = PatchTSTConfig.from_pretrained(model_checkpoint) # Get default context
    # required_min_data = (configured_context_length if configured_context_length is not None else temp_config_for_check.context_length) + prediction_length + 50 # at least 50 for fine-tuning
    # Using a fixed reasonable minimum for now, as config.context_length might be large
    required_min_data_points = (configured_context_length if configured_context_length is not None else 512) + prediction_length + 50


    if len(historical_data_df) < required_min_data_points: # Ensure enough data for context, prediction, and some buffer
        print(f"Insufficient data: {len(historical_data_df)} < {required_min_data_points} required points (context + prediction + buffer). Skipping.")
        return {"status": "failed", "reason": "insufficient historical data"}

    try:
        # === STEP 1: PROPER CONFIG AND MODEL SETUP ===
        print("Loading PatchTST configuration...")
        base_config = PatchTSTConfig.from_pretrained(model_checkpoint) # Load base to get some defaults like d_model

        # Determine effective context length
        effective_context_length = base_config.context_length
        if configured_context_length is not None:
            effective_context_length = configured_context_length
        # Ensure context_length is not too large for the available data
        effective_context_length = min(effective_context_length, len(historical_data_df) - prediction_length -1) # -1 to be safe

        if effective_context_length <= 0:
             print(f"Effective context length ({effective_context_length}) is too small. Adjusting or skipping.")
             # Fallback or error, e.g., use a small default or return error
             if len(historical_data_df) > prediction_length + 10 : # Arbitrary small context if possible
                 effective_context_length = max(10, len(historical_data_df) // 4) # Ensure some minimal context
             else:
                 return {"status": "failed", "reason": f"Cannot determine a valid context length with data size {len(historical_data_df)} and pred_len {prediction_length}"}


        financial_config = PatchTSTConfig(
            context_length=effective_context_length,
            prediction_length=prediction_length,
            patch_length=16,
            patch_stride=8,
            num_input_channels=1,
            d_model=base_config.d_model,
            num_attention_heads=base_config.num_attention_heads,
            num_hidden_layers=base_config.num_hidden_layers,
            ffn_dim=base_config.ffn_dim,
            dropout=0.1,
            head_dropout=0.1,
            scaling="std",
            loss="mse"
        )
        print(f"Financial config - Context: {financial_config.context_length}, Prediction: {financial_config.prediction_length}")

        model = PatchTSTForPrediction(financial_config)

        # === STEP 2: SELECTIVE WEIGHT TRANSFER ===
        print("Loading and transferring compatible weights...")
        try:
            pretrained_model = PatchTSTForPrediction.from_pretrained(model_checkpoint)
            pretrained_dict = pretrained_model.state_dict()
            model_dict = model.state_dict()
            compatible_weights = {}
            for k, v in pretrained_dict.items():
                if k in model_dict and v.size() == model_dict[k].size():
                    if not any(skip_layer in k for skip_layer in ['input_embedding', 'projection', 'head', 'value_embedding']): # Added value_embedding
                        compatible_weights[k] = v
            model.load_state_dict(compatible_weights, strict=False)
            print(f"Transferred {len(compatible_weights)}/{len(model_dict)} compatible weights")
        except Exception as e:
            print(f"Weight transfer warning: {e}. Using random initialization for some layers.")

        device = "cuda" if torch.cuda.is_available() else "cpu"
        model.to(device)

        # === STEP 3: DATA PREPARATION FOR FINANCIAL DOMAIN ===
        close_prices = historical_data_df['close'].values.astype(np.float32)
        
        # Financial-specific preprocessing
        price_std = np.std(close_prices)
        price_mean = np.mean(close_prices)
        outlier_mask = np.abs(close_prices - price_mean) < 3 * price_std # Stricter outlier removal if needed: 2.5 * price_std
        close_prices_cleaned = close_prices[outlier_mask]

        # Check data sufficiency AGAIN after outlier removal for log transform and differencing
        if len(close_prices_cleaned) < financial_config.context_length + financial_config.prediction_length + 2: # +2 for diff and one original price
            print(f"Insufficient data ({len(close_prices_cleaned)}) after outlier removal for context/prediction. Skipping.")
            return {"status": "failed", "reason": "insufficient clean data for context/prediction"}

        log_prices = np.log(close_prices_cleaned) # Use cleaned prices
        log_returns = np.diff(log_prices) # This results in len(log_prices) - 1 elements

        # One more check for log_returns length
        if len(log_returns) < financial_config.context_length + financial_config.prediction_length:
            print(f"Insufficient log_returns data ({len(log_returns)}) for context/prediction. Required: {financial_config.context_length + financial_config.prediction_length}. Skipping.")
            return {"status": "failed", "reason": "insufficient log_returns data for context/prediction"}


        # === STEP 4: DOMAIN-SPECIFIC FINE-TUNING ===
        finetuned_this_run = False # Flag to track if fine-tuning actually happened
        if enable_fine_tuning and fine_tune_epochs > 0:
            print(f"Attempting fine-tuning on {symbol_name} financial data for {fine_tune_epochs} epochs...")

            # ---> USE enhanced_patchtst_finetune FOR DATA PREPARATION <---
            X_sequences, y_sequences, finetune_data_ok = enhanced_patchtst_finetune(
                log_returns=log_returns,  # Pass the 1D log returns array
                context_length=financial_config.context_length,
                prediction_length=financial_config.prediction_length,
                min_sequences=20  # Minimum number of sequences to proceed with training
            )

            if finetune_data_ok:
                finetuned_this_run = True
                train_inputs = torch.tensor(X_sequences, dtype=torch.float32).unsqueeze(-1).to(device)
                train_targets = torch.tensor(y_sequences, dtype=torch.float32).unsqueeze(-1).to(device)

                dataset = torch.utils.data.TensorDataset(train_inputs, train_targets)
                # Adjust batch size: ensure it's not larger than dataset size
                batch_size = min(16, len(X_sequences) // 2 if len(X_sequences) >= 4 else 1)
                if batch_size == 0 and len(X_sequences) > 0: batch_size = 1
                
                if batch_size > 0 :
                    train_loader = torch.utils.data.DataLoader(
                        dataset,
                        batch_size=batch_size,
                        shuffle=True,
                        drop_last=True if len(X_sequences) > batch_size else False # Avoid error if drop_last and not enough samples
                    )

                    model.train()
                    optimizer = torch.optim.AdamW(model.parameters(), lr=1e-4, weight_decay=0.01)
                    scheduler = torch.optim.lr_scheduler.CosineAnnealingLR(optimizer, T_max=fine_tune_epochs)
                    best_loss = float('inf')
                    patience = 3
                    patience_counter = 0

                    for epoch in range(fine_tune_epochs):
                        epoch_loss = 0
                        num_batches = 0
                        if not train_loader: # Should not happen if batch_size > 0
                            print("Train loader is empty, cannot train.")
                            break

                        for batch_idx, (batch_inputs_data, batch_targets_data) in enumerate(train_loader):
                            # batch_inputs_data and batch_targets_data are already on device from above
                            optimizer.zero_grad()
                            outputs = model(past_values=batch_inputs_data, future_values=batch_targets_data)
                            loss = outputs.loss
                            pred_std = torch.std(outputs.prediction_outputs)
                            if pred_std > 0.1: loss += 0.1 * pred_std # Volatility penalty
                            loss.backward()
                            torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1.0)
                            optimizer.step()
                            epoch_loss += loss.item()
                            num_batches += 1
                        
                        if num_batches > 0:
                            avg_loss = epoch_loss / num_batches
                            scheduler.step()
                            print(f"Epoch {epoch+1}/{fine_tune_epochs} | Loss: {avg_loss:.6f} | LR: {scheduler.get_last_lr()[0]:.6f}")
                            if avg_loss < best_loss:
                                best_loss = avg_loss
                                patience_counter = 0
                            else:
                                patience_counter += 1
                                if patience_counter >= patience:
                                    print(f"Early stopping at epoch {epoch+1}")
                                    break
                        else:
                            print(f"Epoch {epoch+1}/{fine_tune_epochs} | No batches processed. Skipping scheduler step.")
                            # If no batches, likely means training data was too small for even one batch with drop_last=True
                            # or train_loader was unexpectedly empty.
                            break # Break from epoch loop if no batches can be processed.

                    if num_batches > 0 : print(f"Fine-tuning completed. Best loss: {best_loss:.6f if best_loss != float('inf') else 'N/A'}")
                else:
                    print(f"Fine-tuning skipped for {symbol_name}: Not enough sequences ({len(X_sequences)}) to form a batch.")
                    finetuned_this_run = False # Ensure flag is false
            else:
                print(f"Fine-tuning skipped for {symbol_name} due to insufficient sequences from enhanced_patchtst_finetune.")
        else:
            print(f"Fine-tuning disabled or fine_tune_epochs is 0 for {symbol_name}.")


        # === STEP 5: FINANCIAL-OPTIMIZED FORECASTING ===
        model.eval()
        
        # Prepare input for forecasting using log returns. Context length from the model's config.
        current_model_context_length = model.config.context_length # Use context_length from the *actual model instance*
        
        if len(log_returns) >= current_model_context_length:
            past_returns_for_forecast = log_returns[-current_model_context_length:]
        else:
            # This case should ideally be rare if data checks above are robust.
            # If it happens, pad with mean return.
            print(f"Warning: log_returns ({len(log_returns)}) shorter than model context ({current_model_context_length}) for forecast input. Padding.")
            mean_log_return = np.mean(log_returns) if len(log_returns) > 0 else 0
            padding_needed_forecast = current_model_context_length - len(log_returns)
            past_returns_for_forecast = np.concatenate([
                np.full(padding_needed_forecast, mean_log_return),
                log_returns
            ])

        returns_mean = np.mean(past_returns_for_forecast)
        returns_std = np.std(past_returns_for_forecast)
        returns_std = max(returns_std, 1e-8)
        normalized_returns_for_forecast = (past_returns_for_forecast - returns_mean) / returns_std
        past_tensor = torch.tensor(normalized_returns_for_forecast, dtype=torch.float32).view(1, current_model_context_length, 1).to(device)

        with torch.no_grad():
            outputs = model(past_values=past_tensor) # For inference, only past_values
            forecast_returns_normalized = outputs.prediction_outputs.cpu().numpy().squeeze()

        if forecast_returns_normalized.ndim == 0: forecast_returns_normalized = np.array([forecast_returns_normalized])
        elif forecast_returns_normalized.ndim > 1: forecast_returns_normalized = forecast_returns_normalized.flatten()

        # Ensure correct number of predictions, matching financial_config.prediction_length
        target_prediction_len = financial_config.prediction_length
        if len(forecast_returns_normalized) >= target_prediction_len:
            forecast_returns_normalized = forecast_returns_normalized[:target_prediction_len]
        else:
            last_val = forecast_returns_normalized[-1] if len(forecast_returns_normalized) > 0 else 0
            padding = np.full(target_prediction_len - len(forecast_returns_normalized), last_val)
            forecast_returns_normalized = np.concatenate([forecast_returns_normalized, padding])

        forecast_returns = (forecast_returns_normalized * returns_std) + returns_mean
        
        # Convert back to price levels using the last log_price from cleaned data
        last_log_price_cleaned = log_prices[-1] # log_prices was derived from close_prices_cleaned
        forecast_log_prices = last_log_price_cleaned + np.cumsum(forecast_returns)
        forecast_prices = np.exp(forecast_log_prices)
        forecast_prices = np.maximum(forecast_prices, 0.01)

        # === STEP 6: FINANCIAL ANALYSIS ===
        last_actual_price_cleaned = close_prices_cleaned[-1] # Use last price from cleaned series for consistency
        price_change = forecast_prices[-1] - last_actual_price_cleaned
        magnitude_pct = (price_change / last_actual_price_cleaned) * 100 if last_actual_price_cleaned != 0 else 0
        direction = "üìà UP" if price_change > 0 else "üìâ DOWN" if price_change < 0 else "Ìö°Î≥¥ HOLD"

        # Volatility based confidence
        forecast_price_returns = np.diff(forecast_prices) / forecast_prices[:-1] if len(forecast_prices) > 1 else np.array([0])
        forecast_volatility = np.std(forecast_price_returns)
        
        historical_close_prices_for_vol = close_prices_cleaned[-(30 + 1):] # Need 31 points for 30 returns
        if len(historical_close_prices_for_vol) > 1:
             historical_returns_for_vol = np.diff(historical_close_prices_for_vol) / historical_close_prices_for_vol[:-1]
             historical_volatility = np.std(historical_returns_for_vol)
        else:
             historical_volatility = forecast_volatility # Fallback if not enough hist data for vol

        vol_ratio = forecast_volatility / max(historical_volatility, 1e-8)
        confidence = "üî¥ Low"
        if vol_ratio < 1.5 and abs(magnitude_pct) > 0.5 : confidence = "üü¢ High"
        elif vol_ratio < 2.0 and abs(magnitude_pct) > 0.2 : confidence = "üü° Medium"
        
        atr_val = np.nan
        if 'ATR_14' in historical_data_df.columns and not historical_data_df['ATR_14'].empty:
            atr_val = historical_data_df['ATR_14'].iloc[-1]
        elif len(historical_data_df) >= 15 : # Min length for ATR calc
            try: # Simplified ATR calculation if not present
                high_col, low_col, close_col_df = historical_data_df['high'], historical_data_df['low'], historical_data_df['close']
                tr1 = pd.DataFrame(high_col - low_col)
                tr2 = pd.DataFrame(abs(high_col - close_col_df.shift(1)))
                tr3 = pd.DataFrame(abs(low_col - close_col_df.shift(1)))
                tr = pd.concat([tr1, tr2, tr3], axis=1).max(axis=1)
                atr_val = tr.rolling(14).mean().iloc[-1]
            except Exception as e_atr:
                print(f"Could not calculate ATR fallback: {e_atr}")


        return {
            "status": "success", "forecast": forecast_prices.tolist(), "last_price": float(last_actual_price_cleaned),
            "direction": direction, "magnitude": float(magnitude_pct), "confidence": confidence,
            "atr_threshold": float(atr_val) if not np.isnan(atr_val) else None,
            "method": f"PatchTST {'Fine-tuned' if finetuned_this_run else 'Pre-trained'} ({fine_tune_epochs} epochs attempted)",
            "model_info": {
                "context_length": financial_config.context_length, "prediction_length": financial_config.prediction_length,
                "fine_tuned_actually": finetuned_this_run
            }
        }

    except Exception as e:
        print(f"Error in PatchTST forecasting for {symbol_name}: {e}")
        import traceback
        traceback.print_exc()
        return {"status": "failed", "reason": f"PatchTST error: {str(e)}", "forecast": None}

def integrate_gap_specific_prediction(symbol: str, multi_tf_data: dict) -> dict:
    """
    Simple gap prediction integration for your notebook.
    """
    daily_data = multi_tf_data.get("daily", pd.DataFrame())
    
    if daily_data.empty:
        return {"status": "failed", "reason": "No daily data available"}
    
    # Simple gap analysis based on recent volatility and momentum
    if len(daily_data) >= 5:
        recent_returns = daily_data['close'].pct_change().tail(5)
        recent_volatility = recent_returns.std()
        momentum = recent_returns.mean()
        
        # Simple gap probability calculation
        gap_probability = min(abs(momentum) * 100 + recent_volatility * 50, 100)
        direction = "UP" if momentum > 0 else "DOWN"
        confidence = "HIGH" if gap_probability > 70 else "MEDIUM" if gap_probability > 40 else "LOW"
        
        return {
            "status": "success",
            "symbol": symbol,
            "gap_prediction": {
                "gap_probability": round(gap_probability, 1),
                "expected_direction": direction,
                "confidence": confidence,
                "momentum": round(momentum, 4),
                "volatility": round(recent_volatility, 4)
            }
        }
    else:
        return {
            "status": "failed", 
            "reason": "Insufficient data for gap analysis"
        }

def simple_ma_fallback(prices, length=5):
    """Moving average fallback for errors"""
    if len(prices) < 5:
        return np.full(length, prices[-1] if len(prices) > 0 else np.nan)
    ma_val = np.mean(prices[-5:])
    return np.full(length, ma_val)


def simple_ma_fclose(prices, length=5):
    """Moving average fallback"""
    if len(prices) < length:
        return np.full(length, np.nan)
    ma_val = np.mean(prices[-length:])
    return np.full(length, ma_val)


# --- MAIN WORKFLOW FUNCTION ---
def run_full_workflow(symbol=DEFAULT_SYMBOL, start_date_str=START_DATE, end_date_str=END_DATE, api_key_val=API_KEY, force_train_autoformer=False, run_optuna_lgbm=False, use_foundation_model=True):
    print(f"\n{'='*40}\nüöÄ ENHANCED WORKFLOW FOR:processed_dfprocessed_df {symbol}\n{'='*40}")
    # Initialize default_return dictionary
    default_return = {"symbol": symbol, "status": "Workflow Started", "raw_data_shape": (0,0), "featured_data_shape": (0,0), "X_train_shape": (0,0), "X_test_shape": (0,0), "selected_features_count": 0, "selected_feature_names": [], "scaler_object": None, "ml_model_object": None, "lgbm_feature_importance": None, "causal_model_object": None, "causal_feature_ranking": [], "onnx_model_path": None, "forecasting_results": None, "lgbm_optimized_params": None }
    price_c, target_col = 'close', 'target' # Define globally used column names

    df_raw = fetch_twelve_data(symbol, api_key_val, start_date_str=start_date_str, end_date_str=end_date_str)
    if df_raw is None or df_raw.empty: default_return["status"] = "Data Fetching Failed"; return default_return
    default_return["raw_data_shape"] = df_raw.shape

    # ---> START CONTEXT CONFIGURATION <---
    print(f"\n--- ‚öôÔ∏è Configuring Context Windows for {symbol} ---")
    dynamic_context_length = 512 # Default fallback context length
    
    if not df_raw.empty:
     context_configs = configure_extended_context(data_length=len(df_raw))
     dynamic_context_length = context_configs['adaptive'] # Or your preferred key like 'long_term'
     print(f"Dynamically selected context length for {symbol}: {dynamic_context_length} days")
    else:
     print(f"Using default context length ({dynamic_context_length}) for {symbol} due to empty df_raw.")
    
    # Ensure "forecasting_results" is a dictionary before assignment
    if default_return.get("forecasting_results") is None: # Check if the value is None
     default_return["forecasting_results"] = {}   # Initialize as an empty dictionary
    
    # Now it's safe to assign the key
    default_return["forecasting_results"]["configured_context_length"] = dynamic_context_length
     # ---> END CONTEXT CONFIGURATION <---


    
    print(f"\n--- üîß Feature Engineering: {symbol} ---")
    df_f = df_raw.copy()
    df_f = add_technical_indicators(df_f); df_f = add_optimized_features(df_f, price_col=price_c, volume_col='volume')
    df_f = add_wavelet_features(df_f, column=price_c); df_f = add_entropy_features(df_f, column=price_c, window=40)
    df_f = add_advanced_technical_features(df_f, price_col=price_c, high_col='high', low_col='low', volume_col='volume')
    df_f = add_transformer_features_conceptual(df_f, column=price_c, sequence_length=20)
    df_f = add_multitimeframe_features(df_f, price_col=price_c)


    if 'log_returns' in df_f.columns:
     df_f = add_regime_features(df_f, returns_col='log_returns')
    else:
        print(f"Skipping regime features for {symbol} as 'log_returns' column is missing.")
    
    if 'RSI_14' in df_f.columns and 'ADX_14' in df_f.columns: df_f['RSI_ADX_interaction'] = df_f['RSI_14'] * df_f['ADX_14'] / 100.0
    if 'ATR_14' in df_f.columns and 'volatility_20' in df_f.columns: volatility_safe = df_f['volatility_20'].replace(0, np.nan); df_f['ATR_vol_ratio'] = df_f['ATR_14'] / volatility_safe
    default_return["featured_data_shape"] = df_f.shape
    
    print(f"\n--- Data Cleaning (Inf/NaN Handling & Imputation): {symbol} ---")
    df_f.replace([np.inf, -np.inf], np.nan, inplace=True); numeric_cols_to_impute = df_f.select_dtypes(include=np.number).columns
    if not numeric_cols_to_impute.empty:
        nan_counts_before = df_f[numeric_cols_to_impute].isnull().sum().sum()
        if nan_counts_before > 0: print(f"NaNs before imputation: {nan_counts_before}"); df_f[numeric_cols_to_impute] = df_f[numeric_cols_to_impute].interpolate(method='linear', limit_direction='both', axis=0).fillna(method='bfill').fillna(method='ffill').fillna(0); print(f"NaNs after imputation: {df_f[numeric_cols_to_impute].isnull().sum().sum()}")
        else: print("No NaNs in numeric features to impute.")
    else: print("No numeric columns for imputation.")
    
    print(f"\n--- üìä Regime Detection: {symbol} ---"); df_f = detect_regimes_simple(df_f, column=price_c)
    print(f"\n--- üéØ Target Definition: {symbol} ---"); df_f = balanced_target_definition(df_f, column=price_c, periods=5)

    causal_model_obj = None
    causal_feat_ranking = []

    print(f"\n--- Causal Discovery & Ranking: {symbol} ---"); causal_model_obj = None; causal_feat_ranking = []
    if not SKIP_CAUSAL_ANALYSIS_FOR_DEBUGGING:
        # This print statement below is redundant if the one above it is kept, but not harmful.
        # You might want to remove one of them for cleaner output.
        print(f"\n--- Causal Discovery & Ranking (Actual Process): {symbol} ---") # Clarified for distinction
        if df_f is not None and not df_f.empty and target_col in df_f.columns and df_f[target_col].nunique(dropna=True) > 1:
            causal_model_obj = discover_causal_structure(df_f.copy(), target_col=target_col, price_c=price_c, symbol=symbol, max_feats=10)
            if causal_model_obj:
                # Assuming causal_feature_selection is defined elsewhere and you want to call it.
                # If it's not defined or you don't have it, this line will error later.
                # For now, focusing on the IndentationError.
                # causal_feat_ranking = causal_feature_selection(df_f.copy(), target_col=target_col, max_causal_feats=10)
                print(f"Causal model object created for {symbol}.") # Placeholder if causal_feature_selection is an issue
            else:
                print(f"Causal model object was not created for {symbol}, skipping feature ranking.")
        else:
            # This block executes if the conditions for causal discovery (df_f checks) are not met.
            print(f"Skipping Causal Discovery for {symbol} due to data/target issues (data empty, target missing, or <2 unique target values).")
    else:
        # This block executes if SKIP_CAUSAL_ANALYSIS_FOR_DEBUGGING is True.
        print(f"\n--- SKIPPING Causal Discovery & Ranking for {symbol} (DEBUG MODE) ---")
        # These lines correctly assign the (likely None/empty) initial values to default_return
        default_return["causal_model_object"] = causal_model_obj
        default_return["causal_feature_ranking"] = causal_feat_ranking
    
    print(f"\n--- ML Preparation & Feature Selection: {symbol} ---"); X_tr, X_te, y_tr, y_te, scaler_obj = None, None, None, None, None; sel_feat_names = []; current_status_ml = "ML Prep Incomplete"
    if df_f is None or df_f.empty or target_col not in df_f.columns or df_f[target_col].isnull().all(): current_status_ml = "ML Prep Failed - Data/Target empty or all NaN"
    elif df_f[target_col].nunique(dropna=True) <= 1: current_status_ml = f"ML Prep Skipped - Target has <=1 unique non-NaN values ({df_f[target_col].nunique(dropna=True)})"
    else:
        X_tr_p, X_te_p, y_tr_p, y_te_p, scaler_obj_p = prepare_ml_data(df_f.copy(), target_col=target_col)
        if X_tr_p is not None and not X_tr_p.empty and y_tr_p is not None and not y_tr_p.empty and y_tr_p.nunique() > 1:
            scores_df = prioritized_feature_selection(X_tr_p.copy(), y_tr_p.copy(), causal_feat_ranking, max_features=25)
            sel_feat_names_intermediate = scores_df['Feature'].tolist() if scores_df is not None and not scores_df.empty else X_tr_p.columns.tolist()[:25]
            X_tr_sel = X_tr_p[[c for c in sel_feat_names_intermediate if c in X_tr_p.columns]]; X_te_sel = X_te_p[[c for c in sel_feat_names_intermediate if c in X_te_p.columns]]
            if not X_tr_sel.empty:
                numeric_cols_xtr_sel = X_tr_sel.select_dtypes(include=np.number).columns
                if not numeric_cols_xtr_sel.empty:
                    variance_selector = VarianceThreshold(threshold=0.001); variance_selector.fit(X_tr_sel[numeric_cols_xtr_sel])
                    sel_feat_names_vt = X_tr_sel[numeric_cols_xtr_sel].columns[variance_selector.get_support()].tolist()
                    sel_feat_names = sel_feat_names_vt if sel_feat_names_vt else sel_feat_names_intermediate
                    non_numeric_sel_intermed = [f for f in sel_feat_names_intermediate if f not in numeric_cols_xtr_sel and f in X_tr_p.columns]
                    sel_feat_names = list(dict.fromkeys(sel_feat_names + non_numeric_sel_intermed))[:25] # Ensure max_features not exceeded
                else: sel_feat_names = X_tr_sel.columns.tolist()[:25]
                X_tr = X_tr_p[sel_feat_names].copy(); X_te = X_te_p[sel_feat_names].copy(); y_tr, y_te = y_tr_p.copy(), y_te_p.copy() # Ensure copies
                if not X_tr.empty:
                    final_numeric_cols_tr = X_tr.select_dtypes(include=np.number).columns
                    if not final_numeric_cols_tr.empty:
                        final_scaler = StandardScaler(); X_tr[final_numeric_cols_tr] = final_scaler.fit_transform(X_tr[final_numeric_cols_tr]); default_return["scaler_object"] = final_scaler
                        if not X_te.empty: final_numeric_cols_te = [c for c in final_numeric_cols_tr if c in X_te.columns]; X_te[final_numeric_cols_te] = final_scaler.transform(X_te[final_numeric_cols_te]) if final_numeric_cols_te else None
                    current_status_ml = "ML Prep Completed"
                else: current_status_ml = "ML Prep Failed - X_tr empty post-selection"
            else: current_status_ml = "ML Prep Failed - X_tr_sel empty"
        else: current_status_ml = "ML Prep Failed - Initial data prep insufficient or target invariant"
    default_return["status"] = current_status_ml; default_return["X_train_shape"] = X_tr.shape if X_tr is not None else (0,0); default_return["X_test_shape"] = X_te.shape if X_te is not None else (0,0); default_return["selected_feature_names"] = sel_feat_names; default_return["selected_features_count"] = len(sel_feat_names)
    
    lgbm_model_obj, lgbm_feat_imp, onnx_file_path, lgbm_final_params = None, None, None, None
    if current_status_ml == "ML Prep Completed" and X_tr is not None and not X_tr.empty and y_tr is not None and not y_tr.empty and y_tr.nunique() > 1:
        print(f"\n--- Model Training & Eval: {symbol} ---")
        try:
            if run_optuna_lgbm:
                unique_labels_tr = sorted(y_tr.squeeze().unique()); num_classes_tr = len(unique_labels_tr)
                optuna_base_params = {'objective': 'multiclass' if num_classes_tr > 2 else 'binary', 'metric': 'multi_logloss' if num_classes_tr > 2 else 'binary_logloss', 'num_class': num_classes_tr if num_classes_tr > 2 else None, 'class_weight': 'balanced', 'random_state': 42, 'verbose': -1, 'n_jobs': -1}
                lgbm_final_params = optimize_lgbm_hyperparameters(X_tr, y_tr.squeeze(), optuna_base_params, n_trials=30)
            else: lgbm_final_params = optimized_lightgbm_params()
            default_return["lgbm_optimized_params"] = lgbm_final_params
            lgbm_model_obj, lgbm_feat_imp = train_lightgbm_model(X_tr, y_tr, X_te, y_te, optimized_params=lgbm_final_params)
            if lgbm_model_obj:
                default_return["status"] = "Completed"; default_return["ml_model_object"] = lgbm_model_obj; default_return["lgbm_feature_importance"] = lgbm_feat_imp
                if lgbm_feat_imp is not None and not lgbm_feat_imp.empty: plot_feature_importance(lgbm_feat_imp, top_n=20, symbol_for_plot=symbol)
                if X_te is not None and not X_te.empty and not X_te.iloc[[0]].empty:
                    onnx_sample_df = X_te.iloc[[0]][sel_feat_names] if sel_feat_names and all(f in X_te.columns for f in sel_feat_names) else X_te.iloc[[0]]
                    onnx_file_path = export_lgbm_to_onnx(lgbm_model_obj, onnx_sample_df, file_path=f"{symbol}_lgbm_model.onnx")
                    default_return["onnx_model_path"] = onnx_file_path
            else: default_return["status"] = "Completed (LGBM Training Failed/Skipped)"
        except Exception as e_train: print(f"LGBM Training error: {e_train}\n{traceback.format_exc()}"); default_return["status"] = "Completed (LGBM Training Error)"
    else: print(f"‚ö†Ô∏è Skipping LGBM Training for {symbol}: {current_status_ml} or invalid data."); default_return["status"] = current_status_ml if "Failed" in current_status_ml or "Skipped" in current_status_ml else "ML Training Skipped"

    if use_foundation_model:
        print(f"\n--- Foundation Model Forecasting (PatchTST): {symbol} ---")
        
        # Use the enhanced PatchTST with fine-tuning
        foundation_forecast_output = run_patchtst_foundation_forecast(
            symbol_name=symbol, 
            historical_data_df=df_raw, 
            prediction_length=5,
            fine_tune_epochs=15,  # Enable fine-tuning
            enable_fine_tuning=True,
            configured_context_length=dynamic_context_length
        )
        
        # Add gap-specific prediction
        gap_specific_prediction = integrate_gap_specific_prediction(symbol, {"daily": df_raw})
        
        default_return["forecasting_results"] = foundation_forecast_output
        default_return["gap_specific_analysis"] = gap_specific_prediction

        
    else: # Fallback to custom Autoformer (which needs training to be effective)
        print(f"\n--- Autoformer Model Initialization & Forecasting: {symbol} ---")
        autoformer_predictor_instance = train_or_load_autoformer(df_raw['close'], force_train=force_train_autoformer)
        if autoformer_predictor_instance:
            # This 'ultimate_forecasting_workflow' was the one for the custom Autoformer. Ensure it's defined and compatible.
            # For now, let's assume it's similar to how PatchTST output is structured.
            # forecast_output = ultimate_forecasting_workflow(symbol, df_raw.copy(), autoformer_predictor_instance, prediction_length=5)
            # Simplified call for custom Autoformer using its predict method directly:
            if hasattr(autoformer_predictor_instance, 'predict'):
                custom_fcst_vals = autoformer_predictor_instance.predict(df_raw['close'])
                # Basic dict structure for custom autoformer output
                default_return["forecasting_results"] = {
                    "status": "Completed" if custom_fcst_vals is not None else "Failed",
                    "forecast": custom_fcst_vals, "method": "Custom AutoformerPredictor (Untrained/Loaded)",
                    "last_price": df_raw['close'].iloc[-1] if not df_raw.empty else np.nan
                    # Add more analysis here like direction, magnitude, confidence for custom model
                }
            else: default_return["forecasting_results"] = {"status": "failed", "reason": "Custom Autoformer predict method missing"}
        else: print("Custom Autoformer instance unavailable. Skipping forecasting."); default_return["forecasting_results"] = {"status": "failed", "reason": "Custom Autoformer instance unavailable"}
    print(f"\n{'='*40}\nWORKFLOW FOR {symbol} CONCLUDED (Final Status: {default_return['status']})\n{'='*40}")
    return default_return

# --- MAIN EXECUTION BLOCK ---
if __name__ == "__main__":
    print("üöÄ Starting ENHANCED ML Trading System Workflow...")
    if API_KEY == "YOUR_API_KEY_HERE" or not API_KEY: print("="*30 + " IMPORTANT: SET YOUR TWELVE_DATA_API_KEY " + "="*30)
    stocks_to_analyze = ["MSFT", "AAPL"]
    all_run_results = {}
    FORCE_TRAIN_AUTOFORMER_FLAG = False # Set True for initial/re-training of custom Autoformer
    RUN_OPTUNA_FOR_LGBM_FLAG = False    # Set True for LightGBM HPO
    USE_FOUNDATION_MODEL_FLAG = True  # Set True to use PatchTST, False for custom Autoformer

    for i, stock_symbol in enumerate(stocks_to_analyze):
        print(f"\nüéØ Processing {stock_symbol} ({i+1}/{len(stocks_to_analyze)})...")
        try:
            stock_run_data = run_full_workflow(symbol=stock_symbol, api_key_val=API_KEY, force_train_autoformer=FORCE_TRAIN_AUTOFORMER_FLAG, run_optuna_lgbm=RUN_OPTUNA_FOR_LGBM_FLAG, use_foundation_model=USE_FOUNDATION_MODEL_FLAG)
            all_run_results[stock_symbol] = stock_run_data
        except Exception as e_main_loop: print(f"FATAL ERROR processing {stock_symbol}: {e_main_loop}\n{traceback.format_exc()}"); all_run_results[stock_symbol] = {"symbol": stock_symbol, "status": f"Workflow Crashed: {e_main_loop}"}
        if i < len(stocks_to_analyze) - 1: print(f"Pausing..."); time.sleep(5)
    print("\n\n--- üìà DETAILED RESULTS SUMMARY ---")
    for symbol_res_key, res_item in all_run_results.items():
        print(f"\n--- Results for: {symbol_res_key} ---")
        if res_item and isinstance(res_item, dict):
            print(f"  Overall Status: {res_item.get('status', 'N/A')}")
            forecast_details = res_item.get('forecasting_results')
            if forecast_details and isinstance(forecast_details, dict) and forecast_details.get('status', '').lower() == 'completed':
                print(f"  Forecasting Method: {forecast_details.get('method', 'N/A')}")
                fc_vals = forecast_details.get('forecast'); print(f"  Forecast Values (next {len(fc_vals) if fc_vals is not None else 0} days): {fc_vals}")
                print(f"  Predicted Direction: {forecast_details.get('direction', 'N/A')}")
                mag = forecast_details.get('magnitude'); print(f"  Predicted Magnitude: {mag:.1f}%" if isinstance(mag, (int, float)) and not np.isnan(mag) else f"N/A ({mag})")
                conf_val = forecast_details.get('confidence_metric_value'); conf_lab = forecast_details.get('confidence_label', 'N/A')
                print(f"  Forecast Confidence: {conf_lab} (Metric Val: {conf_val:.4f})" if isinstance(conf_val, (int,float)) and not np.isnan(conf_val) else f"N/A ({conf_lab})")
            elif forecast_details: print(f"  Forecasting Status: {forecast_details.get('status', 'Failed')} ({forecast_details.get('reason', 'Unknown')})")
            else: print("  Forecasting Data: N/A")
        else: print(f"  No valid results for {symbol_res_key}")
    print(f"\nüéâ Enhanced Analysis Run Complete!")
    print("\n" + "="*60 + "\nüí° Recommendation for Further Forecasting Enhancement...\n" + "="*60 + "\n1. The PatchTST model loaded might have uninitialized weights for its prediction head; it requires FINE-TUNING on relevant financial data for reliable forecasts.\n2. Alternatively, fully implement and train the custom AutoformerPredictor.\n3. Explore other pre-trained models from Hugging Face or specialized forecasting libraries.\n" + "="*60)
```
